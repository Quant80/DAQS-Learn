## 🏗️ Lesson 1 — Introduction to Functions

### Why are Functions Required?

When you go to school every morning, you always do the same steps: get up, wash, brush teeth, wear uniform, eat, and go. Your mum does not explain all these steps each morning — she just says **"Get ready for school"** and you know what to do. That phrase is like a **function call**.

In coding, if a group of statements is **repeatedly required**, it is not good to write them separately each time. Instead, define them once in a **function block** and call them whenever needed.

### What is a Function?

> A **function** is a group of statements or a block of code that performs a certain task.
> **NB:** A function is a block of code which only runs when it is **called**.

**Advantages of using functions:**
- Maintaining code is easier
- Organise and manage code into logical pieces
- **Code re-usability** — write once, use many times

### Types of Functions

```
┌────────────────────────────────────────────────────────────────┐
│                    TYPES OF FUNCTIONS                          │
├─────────────────────────────┬──────────────────────────────────┤
│  Pre-defined (Built-in)     │  User-defined                    │
├─────────────────────────────┼──────────────────────────────────┤
│  Created by Python itself   │  Created by the developer        │
│  Installed with Python      │  Written to meet requirements    │
│  Examples: print(), input() │  Example: def add():  ...        │
│  type(), id(), len(), etc.  │  (what we focus on this chapter) │
└─────────────────────────────┴──────────────────────────────────┘
```

### Anatomy of a Function

```
  def  function_name ( parameters ) :
  ┌──┐  ┌──────────┐  ┌──────────┐
  │  │  │          │  │          │
  keyword  name    parentheses  colon (mandatory)

      # body of function (indented)
      statement 1
      statement 2
      return value   ← optional
```

**Syntax to define a function:**
```python
def function_name(parameters):
    # body of function
    return value
```

**Syntax to call a function:**
```python
function_name(parameters)
```

## 📥 Lesson 2 — Function with Arguments

### What is a Parametric Function?

A function that has **parameter(s)** in its definition as input is called a **parametric function** (or function with argument).

### What is an Argument?

An **argument** is a variable or value that is sent to the function as input.

```
┌────────────────────────────────────────────────────────────┐
│          Formal vs Actual Arguments                        │
│                                                            │
│   def add( a , b ):   ← a, b are FORMAL arguments         │
│               │   │      (parameter names in definition)   │
│               ▼   ▼                                        │
│      add(  7  ,  3  )  ← 7, 3 are ACTUAL arguments        │
│                            (values passed in the call)     │
└────────────────────────────────────────────────────────────┘
```

**Ex 7.6** — Parametric function:

````python
# Ex 7.6 — Parametric Function
# a and b are formal arguments; 7 and 3 are actual arguments
def add(a, b):
    c = a + b
    print(c)

add(7, 3)
````

### Types of Arguments

In Python, there are **5 types** of arguments based on how values are sent:

```
┌───────────────────────────────────────────────────────────────────┐
│                    5 TYPES OF ARGUMENTS                           │
├───┬───────────────────────────┬───────────────────────────────────┤
│ 1 │ Positional                │ Matched by order: add(5, 2)       │
│ 2 │ Keyword                   │ Matched by name: add(b=2, a=5)    │
│ 3 │ Positional + Keyword      │ Mixed: add(5, b=2)                │
│ 4 │ Variable Length (*args)   │ Any number: detail("John", 20)    │
│ 5 │ Keyword Variable (**kwargs)│ Any key-value: detail(name="Jo") │
└───┴───────────────────────────┴───────────────────────────────────┘
```

### 1. Positional Arguments

Actual arguments are received by formal arguments **in the same order** they are sent. Both **number and position** must match.

**Ex 7.7** — Positional arguments (order matters):

````python
# Ex 7.7 — Positional Arguments
def sub(a, b):
    c = a - b
    print(c)

sub(5, 2)   # a=5, b=2 → output: 3
````

````python
# Ex 7.8 — Positional arguments TypeError (too many arguments)
def sub(a, b):
    c = a - b
    print(c)

# sub(10, 20, 30)  # TypeError: sub() takes 2 positional arguments but 3 were given
sub(10, 20)  # correct: exactly 2 arguments
````

````python
# Ex 7.9 — Order matters with positional arguments
def sub(a, b):
    c = a - b
    print(c)

sub(5, 2)   # 5-2 = 3
sub(2, 5)   # 2-5 = -3 (different result — order changed!)
````

### 2. Keyword Arguments

Send values using the **name of the parameter** as a key. Order no longer matters since each value is identified by name.

Syntax: `function_name(keyword = actual_argument)`

**Ex 7.10** — Keyword arguments:

````python
# Ex 7.10 — Keyword arguments (order doesn't matter)
def sub(a, b):
    c = a - b
    print(c)

sub(b=2, a=5)   # same as sub(5, 2) even though b comes first
````

### 4. Default Arguments

Assign a **default value** to a parameter in the function definition. This value is used when the caller does not pass that argument.

<div style='border: 2px solid #fd7e14; border-radius: 12px; padding: 15px; background: #fff8f0; margin: 15px 0;'>
<strong style='color: #fd7e14;'>⚠️ Rule:</strong> Default arguments must come <strong>after</strong> all non-default arguments in the function definition. Placing a non-default argument after a default one causes a <code>SyntaxError</code>.
</div>

**Ex 7.13** — TypeError when a required argument is missing:

````python
# Ex 7.13 — TypeError: missing required argument
def sub(a, b):
    c = a - b
    print(c)

# sub(a=5)  # TypeError: sub() missing 1 required positional argument: 'b'
sub(a=5, b=2)  # correct: both arguments provided
````

````python
# Ex 7.14 — Default arguments: b defaults to 0 if not provided
def sub(a, b=0):
    c = a - b
    print(c)

sub(a=5, b=2)  # b provided: 5 - 2 = 3
sub(a=5)       # b omitted: 5 - 0 = 5 (default used)
````

````python
# Ex 7.15 — SyntaxError: non-default argument after default argument
# def add(a=3, b):  # SyntaxError! b (non-default) comes after a (default)
#     print(a + b)

# Correct: default arguments must come LAST
def add(b, a=3):
    print(a + b)

add(2)       # b=2, a uses default 3 → 5
add(2, a=5)  # b=2, a=5 → 7
````

### 5. Variable Length Arguments (`*args`)

When the programmer does not know how many values will be passed, use a **variable length argument** written with a `*` before the variable name. Internally, the provided values are stored as a **tuple**.

**Syntax:**
```python
def function_name(x, *y):
    # body of function
    return value
```

Here `x` is a regular formal argument and `*y` is the variable length argument — it can receive any number of additional values.

**Ex 7.16** — Variable length argument (`*args`):

````python
# Ex 7.16 — Variable length argument (*args)
def detail(name, *others):
    print(name, others)

detail("John")              # others = () empty tuple
detail("John", 20)          # others = (20,)
detail("John", 20, 3.7)     # others = (20, 3.7)
````

```python
def function_name(**x):
```

### Function Calling Another Function

It is possible in Python for a function to **call another function** inside its body.

**Syntax:**
```python
def function_one():
    # body of function one
    return value

def function_two():
    function_one()   # calling function_one from function_two
    return value

function_two()
```

**Ex 7.19** — Function calling another function:

````python
# Ex 7.19 — Function calling another function
def add(a, b):
    return(a + b)

def average(a, b):
    sum = add(a, b)   # calling add() from inside average()
    avg = sum / 2
    print("Average =", avg)

average(7, 3)
````

### Nested Functions

A **nested function** is a function defined **inside** another function. The inner function must be called from within the outer function.

**Syntax:**
```python
def function_one():
    # body of function one
    def function_two():
        # body of function two
        return ...
    function_two()   # must call inner function
    return ...

function_one()
```

**Ex 7.20** — Nested function:

````python
# Ex 7.20 — Nested function
def average(a, b):
    # inner function defined inside outer function
    def add():
        return a + b

    sum = add()       # call inner function
    avg = sum / 2
    print("Average =", avg)

average(7, 3)
````

### Recursive Functions

A **recursive function** is a function that **calls itself** from within its own body. Every recursive function must have a **base case** — a condition that stops the recursion.

```
┌─────────────────────────────────────────────────────┐
│           Recursive Factorial: 4!                   │
│                                                     │
│   factorial(4) = 4 × factorial(3)                   │
│   factorial(3) = 4 × 3 × factorial(2)               │
│   factorial(2) = 4 × 3 × 2 × factorial(1)           │
│   factorial(1) = 4 × 3 × 2 × 1 × factorial(0)       │
│   factorial(0) = 1  ← BASE CASE (stops recursion)   │
│                                                     │
│   General rule: factorial(n) = n × factorial(n-1)   │
└─────────────────────────────────────────────────────┘
```

**Advantages of recursive functions:**
- Reduces length of code and improves readability
- Solves complex problems in an elegant way

**Ex 7.22** — Factorial using recursive function:

````python
# Ex 7.22 — Factorial using recursive function
def factorial(n):
    if n == 0:                    # base case: stop here
        result = 1
    else:
        result = n * factorial(n - 1)  # recursive call
    return result

x = factorial(4)
print("Factorial of 4 is :", x)
````

## 🔭 Lesson 4 — Types of Variables (Scope)

Based on their **scope** (the region of code where they can be accessed), variables are classified into two types:

```
┌──────────────────────────────────────────────────────────────┐
│                    VARIABLE SCOPE                            │
├─────────────────────────┬────────────────────────────────────┤
│  Local Variables        │  Global Variables                  │
├─────────────────────────┼────────────────────────────────────┤
│  Declared INSIDE a      │  Declared OUTSIDE all functions    │
│  function               │                                    │
│  Accessible only WITHIN │  Accessible from ANY function      │
│  that function          │  in the program                    │
│  Destroyed when         │  Persist for the entire            │
│  function exits         │  program lifetime                  │
└─────────────────────────┴────────────────────────────────────┘
```

### 1. Local Variables

**Ex 7.23** — Local variable:

````python
# Ex 7.23 — Local variables (only accessible inside their function)
def add():
    a, b = 5, 2   # local variables
    print(a + b)

add()
````

````python
# Ex 7.24 — NameError: accessing local variables from another function
def add():
    a, b = 5, 2
    print(a + b)

def sub():
    print(a - b)  # NameError: 'a' is local to add(), not visible here!

# sub()  # uncomment to see: NameError: name 'a' is not defined
add()    # this works fine
````

### 2. Global Variables

**Ex 7.25** — Global variables (accessible from any function):

````python
# Ex 7.25 — Global variables
a, b = 5, 2   # declared outside — global variables

def add():
    print(a + b)  # can access global a and b

def sub():
    print(a - b)  # can also access global a and b

add()
sub()
````

````python
# Ex 7.26 — Same name: local variable shadows global variable
a, b = 5, 2  # global a = 5

def add():
    a = 9          # local a = 9 (shadows global a inside add)
    print(a + b)   # uses local a=9, global b=2 → 11

def sub():
    print(a - b)   # uses global a=5 (local a in add() is invisible here)

add()   # 11
sub()   # 3
````

### The `global` Keyword

Use the `global` keyword inside a function to make the function **refer to** and **modify** the global variable instead of creating a local copy.

<div style='border: 2px solid #0077cc; border-radius: 12px; padding: 15px; background: #e8f4fd; margin: 15px 0;'>
<strong style='color: #0077cc;'>💡 Two purposes of the <code>global</code> keyword:</strong>
<ul style='margin: 8px 0 0;'>
<li>To <strong>declare</strong> a global variable inside a function.</li>
<li>To make an existing <strong>global variable available</strong> for modification inside a function.</li>
</ul>
</div>

**Ex 7.27** — Using `global` keyword:

````python
# Ex 7.27 — global keyword: function modifies the global variable
a, b = 5, 2

def add():
    global a   # now 'a' inside this function refers to the global a
    a = 9      # changes the GLOBAL a to 9
    print(a + b)  # 9 + 2 = 11

def sub():
    print(a - b)  # global a has been changed to 9 by add()

add()   # 11
sub()   # 9-2 = 7
````

### The `globals()` Built-in Function

Using `global` makes the local variable unavailable. To access **both** local and global variables with the same name, use the `globals()` built-in function.

`globals()` returns a dictionary of all current global variables. Access a global variable by name: `globals()["variable_name"]`

**Ex 7.28** — `globals()` function:

````python
# Ex 7.28 — globals() built-in function
a, b = 5, 2

def add():
    a = 9                          # local a = 9
    print(a + b)                   # uses local a=9: 9+2 = 11
    print(globals()['a'] + b)      # uses global a=5: 5+2 = 7

add()
````

## ⚡ Lesson 5 — Lambda (Anonymous) Functions

The name of the function is mandatory in a normal `def` definition. But Python's `lambda` keyword lets you define a **simple one-line function without a name** — called a **lambda function** or **anonymous function**.

**Syntax:**
```python
lambda arguments_list : expression
```

```
┌─────────────────────────────────────────────────────────────┐
│          Normal Function vs Lambda Function                 │
```

## 🎭 Lesson 6 — Decorators and Generators

### What is a Decorator?

A **decorator** is a special function that **adds extra functionality** to an existing function — without changing the original function's code.

> A decorator is a function that **accepts** a function as a parameter and **returns** a function.

**Steps to create a decorator:**
```
Step 1: Decorator takes a function as an argument
Step 2: Decorator body has an inner function
Step 3: Decorator should return the inner function
Step 4: Extra functionality is added in the body of inner_function
```

**Syntax:**
```python
def decor(func):
    def inner_function():
        # extra functionality
        return func()
    return inner_function
```

**Ex 7.34** — Decorator that treats negative arguments as zero before adding:

````python
# Ex 7.34 — Decorator Function
def decor(func):
    # inner function adds extra logic
    def inner_function(x, y):
        if x < 0:
            x = 0
        if y < 0:
            y = 0
        return func(x, y)
    return inner_function

def add(a, b):
    res = a + b
    return res

add = decor(add)   # wrap add with the decorator

print(add(20, 30))   # 50
print(add(-10, 5))   # 0 + 5 = 5 (negative treated as 0)
````

<div style='border: 2px solid #0077cc; border-radius: 12px; padding: 15px; background: #e8f4fd; margin: 15px 0;'>
<strong style='color: #0077cc;'>💡 The @ Symbol — Cleaner Decorator Syntax:</strong>
<p style='margin: 8px 0 0;'>Instead of writing <code>add = decor(add)</code>, Python provides the <code>@decor</code> shortcut placed directly above the function definition. The decorator once created can also be used for other functions.</p>
</div>

**Ex 7.35** — Using `@decor` syntax:

````python
# Ex 7.35 — Using @ decorator syntax
def decor(func):
    def inner_function(x, y):
        if x < 0:
            x = 0
        if y < 0:
            y = 0
        return func(x, y)
    return inner_function

@decor   # equivalent to: sub = decor(sub)
def sub(a, b):
    res = a - b
    return res

print(sub(30, 20))   # 10
print(sub(10, -5))   # 10 - 0 = 10 (negative treated as 0)
````

### What are Generators?

**Generators** are special functions that produce a **sequence of values one at a time** using `yield` instead of `return`. They remember their state between calls.

```
┌──────────────────────────────────────────────────────────────┐
│          Normal Function vs Generator                        │
├────────────────────────────┬─────────────────────────────────┤
│  Uses return               │  Uses yield                     │
│  Returns ONE value         │  Produces MULTIPLE values       │
│  Exits after return        │  PAUSES at yield, resumes later │
│  Stores all data in memory │  Generates values ON-THE-FLY    │
└────────────────────────────┴─────────────────────────────────┘
```

**Advantages of Generators:**
- **Memory Efficiency**: generates values on-the-fly, not all at once
- **Lazy Evaluation**: values computed only when needed
- **Iteration Support**: naturally iterable — easy to loop over
- **Simplified Code**: break complex sequences into manageable pieces
- **Infinite Sequences**: handle sequences too large to fit in memory

The `next()` function retrieves the next value from a generator.

**Ex 7.36** — Generator with `yield` and `next()`:

````python
# Ex 7.36 — Generator with yield
def my_gen():
    n = 1
    print(n)
    yield n       # pause here, return n=1

    n += 1
    print(n)
    yield n       # pause here, return n=2

    n += 1
    print(n)
    yield n       # pause here, return n=3

a = my_gen()      # creates generator object — nothing runs yet!

print(type(a))
next(a)           # runs until first yield
next(a)           # resumes until second yield
next(a)           # resumes until third yield
````

````python
# Generator using a for loop — more common and Pythonic pattern
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for val in countdown(5):
    print(val)
````
