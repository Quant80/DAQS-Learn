## 📦 Lesson 1 — Introduction to Lists

### What is a Data Structure?

> A **data structure** is a particular way of organising data in a computer so that it can be used effectively.

When you write a program, you store lots of data. That data can become cluttered and difficult to manage. By grouping related data together and storing it in a single variable, you make it much easier to retrieve, access, and use.

### What is a List?

> A **list** is a data structure which can store a group of elements or objects in a **single variable**.

### Properties of a List

```
┌─────────────────────────────────────────────────────────────────┐
│                    LIST PROPERTIES                              │
├────────────────────────┬────────────────────────────────────────┤
│  Heterogeneous         │  Can store different data types        │
│  Dynamic               │  Size grows and shrinks automatically  │
│  Ordered               │  Insertion order is preserved          │
│  Duplicates Allowed    │  Same value can appear multiple times  │
│  Mutable               │  Content can be modified after creation│
│  Index-based           │  Elements accessed via index numbers   │
└────────────────────────┴────────────────────────────────────────┘
```

## 🛠️ Lesson 2 — Creating Lists

There are **three main ways** to create a list:

```
┌──────────────────────────┬────────────────────────────────────────┐
│  Method                  │  Syntax                                │
├──────────────────────────┼────────────────────────────────────────┤
│  1. Square brackets [ ]  │  list_name = [elem1, elem2, ...]       │
│  2. list() function      │  list_name = list()                    │
│  3. list() + range()     │  list_name = list(range(start,stop,step))│
└──────────────────────────┴────────────────────────────────────────┘
```

**Ex 9.1** — Creating a list using square brackets:

````python
# Ex 9.1 — Creating a list with elements
my_list = ['bag', 'pen', 'cap', 'pen']
print('my list:', my_list)
# Order preserved, duplicates allowed
````

````python
# Ex 9.2 — Empty list using list() function
l = list()
print(l)
print(type(l))
````

````python
# Ex 9.3 — List using list() and range() — multiples of 3 between 5 and 40
M3 = list(range(6, 40, 3))
print(M3)
print(type(M3))
````

````python
# Ex 9.4 — List can also store mixed data types
my_list = ['bag', 'pen', 'cap', 'pen']
mixed   = [1, 'hello', 3.14, True, None]
print('Shopping list:', my_list)
print('Mixed types:  ', mixed)
````

## 🔍 Lesson 3 — Accessing a List

There are three ways to access elements of a list:
1. **Indexing** — by position number
2. **Slicing** — extract a sub-list
3. **Loops** — iterate through all elements (covered in Chapter 10)

### 1. Indexing

```
  my_list = [ 'bag',  'pen',  'cap',  'pen' ]

  Positive:     0       1       2       3
  Negative:    -4      -3      -2      -1

  ┌────────┬────────┬────────┬────────┐
  │ 'bag'  │ 'pen'  │ 'cap'  │ 'pen'  │
  │   0    │   1    │   2    │   3    │
  │  -4    │  -3    │  -2    │  -1    │
  └────────┴────────┴────────┴────────┘
```

<div style='border: 2px solid #fd7e14; border-radius: 12px; padding: 15px; background: #fff8f0; margin: 15px 0;'>
<strong style='color: #fd7e14;'>⚠️ IndexError:</strong> Accessing beyond the range of list index raises <code>IndexError</code>.
</div>

**Ex 9.5** — Indexing:

````python
# Ex 9.5 — List indexing (positive and negative)
my_list = ['bag', 'pen', 'cap', 'pen']
print('my list:', my_list)
print('item at  3:', my_list[3])    # pen (positive)
print('item at -3:', my_list[-3])   # pen (negative)
print(type(my_list))
print(type(my_list[3]))
````

````python
# Ex 9.6 — IndexError: index out of range
my_list = ['bag', 'pen', 'cap', 'pen']
# print(my_list[10])  # IndexError: list index out of range
print('Valid last index is:', len(my_list) - 1)
````

### 2. Slicing

Extracts a **sub-list** from the main list.

**Syntax:** `list_name[start : stop : step]`

```
start  — index where slice starts (inclusive, default: 0)
stop   — index where slice ends   (exclusive, default: end)
step   — increment between indexes (default: 1)
```

**Ex 9.7** — Slicing:

````python
# Ex 9.7 — List slicing
my_list = ['bag', 'pen', 'cap', 'pen']
print('items from 1 to 3:', my_list[1:4])   # pen, cap, pen
print('odd index items:  ', my_list[1::2])  # pen, pen
````

### Modifying a Specific Element

Unlike strings, lists are **mutable** — you can change elements after creation.

**Syntax:** `list_name[i] = new_element`

**Ex 9.8** — List mutability:

````python
# Ex 9.8 — Modify a specific element (replace cap with hat)
my_list = ['bag', 'pen', 'cap', 'pen']
print('Old list:', my_list)
my_list[2] = 'hat'
print('Modified:', my_list)
````

### 3. `index()` and `enumerate()` Methods

**Ex 9.9** — `index()` returns the position of the first occurrence:

````python
# Ex 9.9 — index() method
my_list = ['bag', 'pen', 'cap', 'pen']
print('cap index:', my_list.index('cap'))   # 2
print('pen index:', my_list.index('pen'))   # 1 (first occurrence)
````

````python
# Ex 9.10 — enumerate() returns (index, value) pairs
my_list = ['bag', 'pen', 'cap', 'pen']
print(list(enumerate(my_list)))
# [(0, 'bag'), (1, 'pen'), (2, 'cap'), (3, 'pen')]
````

## ➕ Lesson 4 — Inserting Elements into a List

```
┌──────────────────────────┬──────────────────────────────────────────────┐
│  Method                  │  Description                                 │
├──────────────────────────┼──────────────────────────────────────────────┤
│  list.append(elem)       │  Add one element at the END                  │
│  list.insert(i, elem)    │  Add one element at specific index           │

└──────────────────────────┴──────────────────────────────────────────────┘
```

### 1. `append()` Method

Adds an element to the **end** of the list.

**Syntax:** `list_name.append(new_element)`

**Ex 9.11** — `append()`:

````python
# Ex 9.11 — append() method
my_list = ['bag', 'pen', 'cap', 'pen']
print('Old list:', my_list)
my_list.append('oil')
print('New list:', my_list)
````

### 2. `insert()` Method

Adds an element at a **specific index**.

**Syntax:** `list_name.insert(i, new_element)`

<div style='border: 2px solid #0077cc; border-radius: 12px; padding: 15px; background: #e8f4fd; margin: 15px 0;'>
<strong style='color: #0077cc;'>💡 Insert edge cases:</strong>
<ul style='margin: 8px 0 0;'>
<li>If index &gt; max index → element inserted at the <strong>last</strong> position.</li>
<li>If index &lt; min index → element inserted at the <strong>first</strong> position.</li>
</ul>
</div>

**Ex 9.12** — `insert()`:

````python
# Ex 9.12 — insert() method
my_list = ['bag', 'pen', 'cap', 'pen']
print('Old list:', my_list)
my_list.insert(0, 'oil')   # insert at beginning
print('New list:', my_list)
````

### 3. `extend()` Method

Adds **all elements** of another list into this list (modifies in place).

**Syntax:** `new_list.extend(old_list)`

**Ex 9.13** — `extend()`:

````python
# Ex 9.13 — extend() method
my_list  = ['bag', 'pen', 'cap', 'pen']
dad_list = ['tie', 'comb']
print('my list:    ', my_list)
print('my dad list:', dad_list)
my_list.extend(dad_list)
print('merged list:', my_list)
````

````python
# Ex 9.14 — extend() multiple times
my_list  = ['bag', 'pen', 'cap', 'pen']
mom_list = ['oil', 'dye']
dad_list = ['tie', 'comb']
my_list.extend(mom_list)
my_list.extend(dad_list)
print('my, mom and dad list:', my_list)
````

### 4. Concatenation Operator ( `+` )

Joins two lists and **returns a new list**. Both operands must be lists — mixing types causes `TypeError`.

**Syntax:** `list_3 = list_1 + list_2`

**Ex 9.15** — Concatenation:

````python
# Ex 9.15 — Concatenation operator + (creates new list)
my_list  = ['bag', 'pen', 'cap', 'pen']
mom_list = ['oil', 'dye']
dad_list = ['tie', 'comb']
my_list = my_list + mom_list + dad_list
print('Combined:', my_list)
````

````python
# Ex 9.16 — TypeError: can only concatenate list (not str) to list
a = [1, 2, 3]
b = 'pen'
# c = a + b   # TypeError!
c = a + [b]   # correct: wrap b in a list first
print(c)
````

### 1. `remove()` Method

Removes the **first occurrence** of the specified element.

<div style='border: 2px solid #fd7e14; border-radius: 12px; padding: 15px; background: #fff8f0; margin: 15px 0;'>
<strong style='color: #fd7e14;'>⚠️ ValueError:</strong> If the element does not exist in the list, <code>remove()</code> raises a <code>ValueError</code>. Always check with <code>in</code> before removing.
</div>

**Ex 9.17** — `remove()`:

````python
# Ex 9.17 — remove() removes FIRST occurrence
my_list = ['bag', 'pen', 'cap', 'pen']
print('Old list:', my_list)
my_list.remove('pen')   # removes first 'pen' at index 1
print('New list:', my_list)
````

````python
# Ex 9.18 — ValueError when element not found
my_list = ['bag', 'pen', 'cap', 'pen']
# my_list.remove('hat')  # ValueError: list.remove(x): x not in list

# Safe pattern: check before removing
if 'hat' in my_list:
    my_list.remove('hat')
else:
    print('hat not in list — nothing removed')
````

### 2. `pop()` Method

Removes and **returns** the element at the given index. If no index given, removes the **last** element.

**Syntax:** `list_name.pop(i)`

**Ex 9.19** — `pop()`:

````python
# Ex 9.19 — pop(3) removes element at index 3
my_list = ['bag', 'pen', 'cap', 'pen']
print('Old list:', my_list)
my_list.pop(3)
print('After pop(3):', my_list)

# Ex 9.20 — pop() with no index removes the LAST element
my_list = ['bag', 'pen', 'cap', 'pen']
my_list.pop()
print('After pop():', my_list)

# Ex 9.21 — IndexError when index out of range
# my_list.pop(10)  # IndexError: pop index out of range
````

### 3. `del` Statement

Deletes an element at a given index **or** a slice of elements.

**Syntax:** `del list_name[start : stop : step]`

**Ex 9.22** — `del` statement:

````python
# Ex 9.22 — del slice [1:3]
my_list = ['bag', 'pen', 'cap', 'pen']
print('Old list:', my_list)
del my_list[1:3]
print('After del [1:3]:', my_list)

# Ex 9.23 — del single element
my_list = ['bag', 'pen', 'cap', 'pen']
del my_list[1:2:]
print('After del [1:2:]:', my_list)

# Ex 9.24 — del all elements
my_list = ['bag', 'pen', 'cap', 'pen']
del my_list[:]
print('After del [:]:', my_list)
````

### 4. `clear()` and `*= 0`

Both remove all elements — leaving an empty list `[]`.

**Ex 9.25** — `clear()` and `*= 0`:

````python
# Ex 9.25 — clear() method
my_list = ['bag', 'pen', 'cap', 'pen']
print('Old list:', my_list)
my_list.clear()
print('After clear():', my_list)

# Ex 9.26 — *= 0 method
my_list = ['bag', 'pen', 'cap', 'pen']
my_list *= 0
print('After *= 0:   ', my_list)
````

## 📋 Lesson 7 — Aliasing vs Cloning

### What is Aliasing?

> **Aliasing** gives a new name to an existing list. Both names point to the **same memory address**. Modifying one changes both.

**Syntax:** `list_1 = list_2`

### What is Cloning?

> **Cloning** creates a **new independent copy** at a different memory address. Modifying one does NOT affect the other.

```
┌──────────────────────────────────────────────────────────────────┐
│  ALIASING (same memory)    │  CLONING (different memory)         │
│                            │                                     │
│  my_list ──┐               │  my_list → ['bag','pen'] addr:1000  │
│            ▼               │  twin    → ['bag','pen'] addr:2000  │
│      ['bag','pen','cap']   │                                     │
│            ▲               │  twin[0] = 'book':                  │
│  twin_list ┘               │  my_list still: ['bag','pen']       │
│                            │  twin now:      ['book','pen']      │
│  twin[0]='book':BOTH change│                                     │
└──────────────────────────────────────────────────────────────────┘
```

**Ex 9.31** — Aliasing:

````python
# Ex 9.31 — Aliasing (same object, same memory)
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = my_list   # alias — NOT a copy!
print('my list:  ', my_list)
print('twin list:', twin_list)
````

````python
# Ex 9.32 — Aliasing problem: modifying twin ALSO changes my_list!
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = my_list
twin_list[0] = 'book'
print('my list:  ', my_list)    # 'bag' changed to 'book' — both changed!
print('twin list:', twin_list)
print('Same memory address?', id(my_list) == id(twin_list))  # True
````

### 9 Methods to Clone a List

```
┌────┬───────────────────────────┬────────────────────────────────────────┐
│ No │  Method                   │  Syntax                                │
├────┼───────────────────────────┼────────────────────────────────────────┤
│  1 │  extend() method          │  twin = []; twin.extend(original)      │
│  2 │  + concatenation          │  twin = [] + original                  │
│  3 │  Slice operator           │  twin = original[start:stop:step]      │
│  4 │  list() function          │  twin = list(original)                 │
│  5 │  copy() method            │  twin = original.copy()                │
│  6 │  Shallow copy             │  twin = copy.copy(original)            │
│  7 │  Deep copy                │  twin = copy.deepcopy(original)        │
│  8 │  List Comprehension       │  twin = [i for i in original]          │
│  9 │  append() in loop         │  for i in original: twin.append(i)     │
└────┴───────────────────────────┴────────────────────────────────────────┘
```

**Ex 9.33** — Cloning using `extend()`:

````python
# Ex 9.33 — Method 1: Clone using extend()
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = []
twin_list.extend(my_list)
twin_list[0] = 'book'
print('my list:  ', my_list)    # unchanged
print('twin list:', twin_list)
print('Different memory?', id(my_list) != id(twin_list))  # True
````

````python
# Ex 9.34 — Method 2: Clone using concatenation (+)
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = [] + my_list
twin_list[0] = 'book'
print('Method 2 (+):  my_list:', my_list, '| twin:', twin_list)

# Ex 9.35 — Method 3: Clone using slice operator
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = my_list[:]
twin_list[0] = 'book'
print('Method 3 ([:]): my_list:', my_list, '| twin:', twin_list)

# Ex 9.36 — Method 4: Clone using list() function
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = list(my_list)
twin_list[0] = 'book'
print('Method 4 (list()): my_list:', my_list, '| twin:', twin_list)
````

````python
# Ex 9.37 — Method 5: Clone using copy() method
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = my_list.copy()
twin_list[0] = 'book'
print('Method 5 (.copy()): my_list:', my_list, '| twin:', twin_list)

# Ex 9.38 — Method 6: Clone using Shallow copy (copy module)
import copy
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = copy.copy(my_list)
twin_list[0] = 'book'
print('Method 6 (copy.copy()): my_list:', my_list, '| twin:', twin_list)
````

<div style='border: 2px solid #fd7e14; border-radius: 12px; padding: 15px; background: #fff8f0; margin: 15px 0;'>
<strong style='color: #fd7e14;'>⚠️ Shallow Copy Problem with Nested Lists:</strong>
Shallow copy does NOT copy nested lists — it only copies the <strong>reference</strong> to them. So modifying a nested element in the clone ALSO changes the original. Use <strong>deepcopy</strong> for nested lists.
</div>

**Ex 9.39** — Shallow copy problem, **Ex 9.40** — Deep copy solution:

````python
import copy

# Ex 9.39 — Shallow copy bug: nested lists are still shared!
my_list = [['oat', 'egg'], ['bag', 'pen']]
twin_list = copy.copy(my_list)
twin_list[1][0] = 'book'
print('Shallow copy — my_list:', my_list)    # ALSO changed!
print('Shallow copy — twin:   ', twin_list)

print()

# Ex 9.40 — Deep copy: creates fully independent nested copy
my_list = [['oat', 'egg'], ['bag', 'pen']]
twin_list = copy.deepcopy(my_list)
twin_list[1][0] = 'book'
print('Deep copy — my_list:', my_list)      # unchanged!
print('Deep copy — twin:   ', twin_list)
````

```
┌──────────────────────────┬───────────────────────────────────────────┐
│  Method / Function       │  Description                              │
├──────────────────────────┼───────────────────────────────────────────┤
│  len(list)               │  Count of elements                        │

│  max(list)               │  Largest element                          │

│  sum(list)               │  Sum of all numeric elements              │
```

**Ex 9.41–9.46** — Mathematical methods:

````python
import random

my_list  = ['bag', 'pen', 'cap', 'pen']
twin_list = [['oat', 'egg'], ['bag', 'pen']]
n = [7, 6, 9, 2, 3]

# Ex 9.41 — len()
print('my list length:  ', len(my_list))     # 4
print('twin list length:', len(twin_list))   # 2 (counts inner lists)

# Ex 9.42 — count()
print('\npen in my_list:  ', my_list.count('pen'))     # 2
print('pen in twin_list:', twin_list.count('pen'))    # 0 (nested)

# Ex 9.43 — max()
print('\nmax(n):', max(n))             # 9
print('max(my_list):', max(my_list))  # pen (alphabetical)

# Ex 9.44 — min()
print('\nmin(n):', min(n))             # 2
print('min(my_list):', min(my_list))  # bag

# Ex 9.45 — random.choice()
players = {'Messi', 'Ronaldo', 'Salah'}
print('\nThe Winner is:', random.choice(list(players)))

# Ex 9.46 — sum()
print('\nsum(n):', sum(n))   # 27
````

### Operations Using List Operators

**Ex 9.47** — Multiplication (`*`), **Ex 9.48–9.49** — Relational, **Ex 9.50** — Membership:

````python
# Ex 9.47 — Multiplication (*) operator — repeats elements
mom_list = ['oil', 'dye']
print('Old list:', mom_list)
print('2x list: ', 2 * mom_list)   # ['oil', 'dye', 'oil', 'dye']
````

````python
# Ex 9.48 — Relational operators (compare element by element, left to right)
print([1, 2, 3] < [2, 2, 3])    # True  (1 < 2)
print([1, 2, 3] <= [1, 2, 3])   # True  (equal)
print([1, 2, 3, 4] > [1, 6, 3]) # False (2 < 6)
print([1, 2, 1] == [4])          # False

# Ex 9.49 — String list comparison (case sensitive)
x = ['abc', 'def', 'ghi']
y = ['abc', 'def', 'ghi']
z = ['ABC', 'DEF', 'GHI']
print('\nx == y:', x == y)   # True
print('x == z:', x == z)   # False (case sensitive)
````

````python
# Ex 9.50 — Membership operators: in and not in
my_list   = ['bag', 'pen', 'cap', 'pen']
twin_list = [['oat', 'egg'], ['bag', 'pen']]

print('bag in my_list:     ', 'bag' in my_list)         # True
print('Bag not in my_list: ', 'Bag' not in my_list)     # True (case sensitive)
print('book in my_list:    ', 'book' in my_list)        # False
print('oat in twin_list:   ', 'oat' in twin_list)       # False (checks inner lists, not items)
````

## 📌 Lesson 9 — Introduction to Tuples

### What is a Tuple?

> A **tuple** is a collection of objects that are **ordered** and **immutable** (cannot be changed).

### Characteristics of Tuples

```
┌──────────────────────────────────────────────────────────────────┐
│                    TUPLE CHARACTERISTICS                         │
├────────────────────────┬─────────────────────────────────────────┤
│  Insertion order       │  Preserved — same order as created      │
│  Immutable             │  CANNOT modify elements after creation  │
│  Duplicates allowed    │  Same value can appear multiple times   │
│  Heterogeneous         │  Can store different data types         │
└────────────────────────┴─────────────────────────────────────────┘
```

<div style='border: 2px solid #0077cc; border-radius: 12px; padding: 15px; background: #e8f4fd; margin: 15px 0;'>
<strong style='color: #0077cc;'>💡 When to use a Tuple:</strong> When your data should <strong>not change</strong> over time — e.g., days of the week, calendar months, GPS coordinates, RGB colour values.
</div>

### Creating Tuples

**Ex 9.51** — Empty tuple:

````python
# Ex 9.51 — Empty tuple
emp = ()
print(emp)
print(type(emp))
````

````python
# Ex 9.52 — Single value WITHOUT comma → NOT a tuple!
name = ("John")   # str, not tuple
a    = (11)        # int, not tuple
print(name, type(name))
print(a,    type(a))

print()

# Ex 9.53 — Single value WITH trailing comma → tuple!
name = ("John",)
a    = (11,)
print(name, type(name))
print(a,    type(a))
````

````python
# Ex 9.54 — Tuple with group of values (parentheses)
weekdays = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday")
print('Weekdays:', weekdays)
print(type(weekdays))

# Ex 9.55 — Parentheses are OPTIONAL (comma creates the tuple)
weekdays2 = "Monday", "Tuesday", "Wednesday", "Thursday", "Friday"
print('Same result:', weekdays2)

# Ex 9.56 — Create tuple from list using tuple() function
my_list  = [11, 22, 33]
my_tuple = tuple(my_list)
print('From list:', my_tuple)

# Ex 9.57 — Create tuple using range()
my_tuple = tuple(range(1, 10, 2))
print('From range:', my_tuple)   # (1, 3, 5, 7, 9)
````

### Tuple Immutability

Once created, tuple elements **cannot be modified**. This is the key difference from lists.

**Ex 9.58** — Tuple immutability:

````python
# Ex 9.58 — Tuple Immutability → TypeError
my_tuple = (10, 20, 30, 40)
try:
    my_tuple[1] = 70   # TypeError!
except TypeError as e:
    print('Error:', e)
    print('Tuples are immutable — elements cannot be changed!')
````

````python
# Ex 9.59 — sorted() on a tuple returns a LIST
my_tuple = (40, 10, 30, 20)
sorted_asc  = sorted(my_tuple)               # ascending
sorted_desc = sorted(my_tuple, reverse=True) # descending
print('Original tuple:', my_tuple)
print('Sorted asc:    ', sorted_asc)    # list!
print('Sorted desc:   ', sorted_desc)
````

### Tuple Packing and Unpacking

**Packing** — group multiple variables into a tuple.
**Unpacking** — extract tuple values into separate variables.

**Ex 9.61** — Packing, **Ex 9.62** — Unpacking:

````python
# Ex 9.61 — Tuple Packing
a, b, c, d = 10, 20, 30, 40
my_tuple = a, b, c, d
print('Packed tuple:', my_tuple)

# Ex 9.62 — Tuple Unpacking
my_tuple = (10, 20, 30, 40)
a, b, c, d = my_tuple
print('a =', a, ' b =', b, ' c =', c, ' d =', d)

# Practical: swap two variables in one line!
x, y = 5, 10
print('\nBefore swap: x =', x, 'y =', y)
x, y = y, x
print('After swap:  x =', x, 'y =', y)
````

### Difference Between List and Tuple

```
┌──────────────────────┬────────────────────────┬────────────────────────┐
│  Feature             │  List                  │  Tuple                 │
├──────────────────────┼────────────────────────┼────────────────────────┤
│  Syntax              │  [1, 2, 3]             │  (1, 2, 3)             │
│  Mutability          │  Mutable ✎             │  Immutable 🔒          │
│  Performance         │  Slower                │  Faster                │
│  Memory              │  Uses more             │  Uses less             │
│  Available methods   │  Many                  │  Only count(), index() │
│  Use case            │  Data may change       │  Data won't change     │
│  Example             │  Shopping list         │  Days of the week      │
└──────────────────────┴────────────────────────┴────────────────────────┘
```

````python
# Tuple methods — only 2: count() and index()
my_tuple = (10, 20, 30, 20, 40, 20)
print('count of 20:', my_tuple.count(20))   # 3
print('index of 30:', my_tuple.index(30))   # 2

# Standard functions work on tuples too
print('max:  ', max(my_tuple))   # 40
print('min:  ', min(my_tuple))   # 10
print('len:  ', len(my_tuple))   # 6
print('sum:  ', sum(my_tuple))   # 140
print('20 in tuple:', 20 in my_tuple)  # True
````
