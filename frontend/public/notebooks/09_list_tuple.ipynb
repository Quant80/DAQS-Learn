## 📦 Lesson 1 — Introduction to Data Structures and Lists

### What is a Data Structure?

A **data structure** is a particular way of organizing data in a computer so that it can be used efficiently.

Consider this scenario: You are going to the market and you want to organize the items you want to buy. What would you use? A **shopping list**!

Python uses **lists** for exactly this purpose — to store a group of items in a single variable.

### What is a List?

A **list** is a data structure which can store a group of elements or objects in a **single variable**. It is one of the most versatile and commonly used data types in Python.

### Characteristics and Features of Lists


## 🔍 Lesson 3 — Accessing List Elements

### Positive and Negative Indexing

Every element in a list has an **index** (position number). Python supports both positive (left-to-right) and negative (right-to-left) indexing.

```
  my_list = [ 'bag',  'pen',  'cap',  'pen' ]

  Positive:     0       1       2       3
  Negative:    -4      -3      -2      -1

  ┌───────┬───────┬───────┬───────┐
  │ 'bag' │ 'pen' │ 'cap' │ 'pen' │
  │   0   │   1   │   2   │   3   │
  │  -4   │  -3   │  -2   │  -1   │
  └───────┴───────┴───────┴───────┘
```

### Slicing Lists

Slicing extracts a **portion** of a list using `[start:stop:step]`:
- `start` — index to begin (inclusive, default 0)
- `stop` — index to end (exclusive, default end)
- `step` — how many to jump (default 1)

````python
# Ex 9.5 - Positive indexing
my_list = ['bag', 'pen', 'cap', 'pen']

print('First item:', my_list[0])      # bag
print('Second item:', my_list[1])     # pen
print('Third item:', my_list[2])      # cap
print('Last item:', my_list[3])       # pen
````

````python
# Ex 9.6 - Negative indexing
my_list = ['bag', 'pen', 'cap', 'pen']

print('Last item:', my_list[-1])      # pen
print('Second to last:', my_list[-2]) # cap
print('Third to last:', my_list[-3])  # pen
print('First item:', my_list[-4])     # bag
````

````python
# Ex 9.7 - Slicing lists
my_list = ['bag', 'pen', 'cap', 'pen']

print('All items:', my_list[:])         # all
print('First two:', my_list[0:2])       # bag, pen
print('Last two:', my_list[2:])         # cap, pen
print('Every 2nd:', my_list[0:4:2])     # bag, cap
print('Reversed:', my_list[::-1])       # pen, cap, pen, bag
````

````python
# Ex 9.8 - Modifying elements by index
my_list = ['bag', 'pen', 'cap', 'pen']
print('Before:', my_list)

my_list[0] = 'book'
print('After changing index 0:', my_list)
````

````python
# Ex 9.9 - index() method — find position of an element
# Syntax: list_name.index(element, start, end)
my_list = ['bag', 'pen', 'cap', 'pen']

print('Index of cap:', my_list.index('cap'))         # 2
print('Index of pen:', my_list.index('pen'))         # 1 (first occurrence)
print('Second pen at:', my_list.index('pen', 2))     # 3 (search from index 2)
````

````python
# Ex 9.10 - enumerate() — get both index and value
my_list = ['bag', 'pen', 'cap', 'pen']

for index, value in enumerate(my_list):
    print(f'Index {index}: {value}')
````

```
┌─────────────────────────────────────────────────────────────────────┐
│                    LIST INSERTION METHODS                           │
├──────────────────────────┬──────────────────────────────────────────┤
│  Method                  │  Description                             │
├──────────────────────────┼──────────────────────────────────────────┤
│  list.append(elem)       │  Add one element at the END              │
```

````python
# Ex 9.11 - append() — adds element at the END of the list
my_list = ['bag', 'pen', 'cap']
print('Before:', my_list)

my_list.append('book')
print('After append:', my_list)
````

````python
# Ex 9.12 - insert(i, elem) — inserts element at specific INDEX
my_list = ['bag', 'pen', 'cap']
print('Before:', my_list)

my_list.insert(1, 'book')  # insert 'book' at index 1
print('After insert at 1:', my_list)

my_list.insert(0, 'ruler')  # insert at beginning
print('After insert at 0:', my_list)
````

````python
# Ex 9.13 - extend() — merges another list into this list
my_list = ['bag', 'pen', 'cap']
extra_items = ['book', 'ruler']
print('Before:', my_list)

my_list.extend(extra_items)
print('After extend:', my_list)
````

## 🗑️ Lesson 5 — Deleting Elements from a List

```
┌───────────────────────────────────────────────────────────────────────────┐
│                     LIST DELETION METHODS                                 │
├──────────────────────────────┬────────────────────────────────────────────┤
│  Method / Statement          │  Description                               │
├──────────────────────────────┼────────────────────────────────────────────┤
│  list.remove(elem)           │  Remove FIRST occurrence of element        │
│  list.pop(index)             │  Remove element by INDEX (default: last)   │
│  del list[start:stop]        │  Delete element(s) by index/slice          │
│  list.clear()                │  Remove ALL elements (empty list remains)  │
│  list *= 0                   │  Clear list (alternative to clear())       │
└──────────────────────────────┴────────────────────────────────────────────┘
```

````python
# Ex 9.15 - remove() — removes FIRST occurrence of element
my_list = ['bag', 'pen', 'cap', 'pen']
print('Before:', my_list)

my_list.remove('pen')   # removes first 'pen' at index 1
print('After remove(pen):', my_list)
````

````python
# Ex 9.16 - pop() — removes element by INDEX, returns the removed element
my_list = ['bag', 'pen', 'cap', 'pen']
print('Before:', my_list)

removed = my_list.pop()     # removes and returns last element
print('Removed:', removed)
print('After pop():', my_list)

removed2 = my_list.pop(0)   # removes and returns element at index 0
print('Removed:', removed2)
print('After pop(0):', my_list)
````

````python
# Ex 9.17 - del statement — delete by index or slice
my_list = ['bag', 'pen', 'cap', 'pen']
print('Before:', my_list)

del my_list[1]    # delete index 1
print('After del [1]:', my_list)

my_list2 = ['bag', 'pen', 'cap', 'pen']
del my_list2[1:3]  # delete slice from index 1 to 3 (exclusive)
print('After del [1:3]:', my_list2)
````

````python
# Ex 9.18 - clear() — removes ALL elements
my_list = ['bag', 'pen', 'cap', 'pen']
print('Before:', my_list)

my_list.clear()
print('After clear():', my_list)    # empty list []

# Alternative: *= 0
my_list2 = ['bag', 'pen', 'cap']
my_list2 *= 0
print('After *= 0:', my_list2)
````

## 🔤 Lesson 6 — Arranging List Elements & Nested Lists

### Sorting and Reversing

````python
# Ex 9.19 - sort() — arranges in ascending order (default)
numbers = [7, 2, 9, 1, 5, 3]
names = ['Zara', 'Alice', 'Mike', 'Bob']

print('Before sort:', numbers)
numbers.sort()
print('After sort():', numbers)

print('Before sort:', names)
names.sort()
print('After sort():', names)

# Sort in descending order
numbers.sort(reverse=True)
print('Descending:', numbers)
````

````python
# Ex 9.20 - reverse() — reverses the ORDER of elements (not sort!)
my_list = ['bag', 'pen', 'cap', 'pen']
print('Before:', my_list)

my_list.reverse()
print('After reverse():', my_list)
````

### Nested Lists

A **nested list** is a list within another list. Think of it like a table with rows and columns.

```
twin_list = [['oat', 'egg'], ['bag', 'pen']]

┌─────────────────────────────────────┐
│  twin_list[0] → ['oat', 'egg']      │
│    twin_list[0][0] → 'oat'          │
│    twin_list[0][1] → 'egg'          │
│  twin_list[1] → ['bag', 'pen']      │
│    twin_list[1][0] → 'bag'          │
│    twin_list[1][1] → 'pen'          │
└─────────────────────────────────────┘
```

````python
# Ex 9.21 - Nested lists
twin_list = [['oat', 'egg'], ['bag', 'pen']]

print('Full nested list:', twin_list)
print('First inner list:', twin_list[0])       # ['oat', 'egg']
print('Second inner list:', twin_list[1])      # ['bag', 'pen']
print('First element of first list:', twin_list[0][0])   # oat
print('Second element of second list:', twin_list[1][1]) # pen
````

````python
# Ex 9.22 - Class grades using nested list (like a 2D table)
grades = [
    ['Alice', 85, 90, 78],
    ['Bob',   70, 65, 80],
    ['Carol', 95, 88, 92]
]

for student in grades:
    name = student[0]
    avg = sum(student[1:]) / len(student[1:])
    print(f'{name}: average = {avg:.1f}')
````

## 📋 Lesson 7 — Aliasing vs Cloning

### What is Aliasing?

**Aliasing** is giving a new name to an existing list. Both variables **point to the same memory address** — changing one changes both!

### What is Cloning?

**Cloning** creates a **new, independent copy** of a list. Changing one does NOT affect the other.

```
ALIASING (same memory):            CLONING (different memory):

my_list ─────┐                     my_list ──► [bag, pen, cap]  addr: 1000
             ▼                     twin    ──► [bag, pen, cap]  addr: 2000
         [bag, pen, cap]           
             ▲                     Modify twin[0] = 'book':
twin_list ───┘                     my_list still: [bag, pen, cap]
                                   twin    now:   [book, pen, cap]
Modify twin[0] = 'book':           
BOTH change to [book, pen, cap]!
```

````python
# Ex 9.23 - Aliasing — same memory address
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = my_list   # aliasing — NOT a copy!

twin_list[0] = 'book'

print('my list:', my_list)    # also changed!
print('twin list:', twin_list)
print('Memory Address of my_list:', id(my_list))
print('Memory Address of twin_list:', id(twin_list))
print('Same memory?', id(my_list) == id(twin_list))
````

### 7 Methods to Clone a List

```
┌────┬──────────────────────────────┬───────────────────────────────────────┐
│ No │  Method                      │  Syntax                               │
├────┼──────────────────────────────┼───────────────────────────────────────┤
│  1 │  += (extend operator)        │  twin = []; twin += original          │
│  2 │  + concatenation             │  twin = [] + original                 │
│  3 │  Slice operator              │  twin = original[start:stop:step]     │
│  4 │  list() function             │  twin = list(original)                │
│  5 │  copy() method               │  twin = original.copy()               │
│  6 │  Shallow copy (copy module)  │  twin = copy.copy(original)           │
│  7 │  Deep copy (copy module)     │  twin = copy.deepcopy(original)       │
│  8 │  List Comprehension          │  twin = [i for i in original]         │
│  9 │  append() in a loop          │  for i in original: twin.append(i)    │
└────┴──────────────────────────────┴───────────────────────────────────────┘
```

````python
# Ex 9.34 - Method 1: Cloning using += (extend operator)
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = []

twin_list += my_list     # clone
twin_list[0] = 'book'   # modify clone

print('my list:', my_list)    # unchanged
print('twin list:', twin_list)
print('Memory Address of my_list:', id(my_list))
print('Memory Address of twin_list:', id(twin_list))
````

````python
# Ex 9.35 - Method 3: Cloning using slice operator
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = my_list[0:4:2]   # clone every 2nd element

twin_list[0] = 'book'
print('my list:', my_list)
print('twin list:', twin_list)
print('Memory Address of my_list:', id(my_list))
print('Memory Address of twin_list:', id(twin_list))

# Clone ALL elements with [:]
full_clone = my_list[:]
full_clone[0] = 'notebook'
print('\nFull clone:', full_clone)
print('Original still:', my_list)
````

````python
# Ex 9.36 - Method 4: Cloning using list() function
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = list(my_list)

twin_list[0] = 'book'
print('my list:', my_list)
print('twin list:', twin_list)
print('Memory Address of my_list:', id(my_list))
print('Memory Address of twin_list:', id(twin_list))
````

````python
# Ex 9.37 - Method 5: Cloning using copy() method
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = my_list.copy()

twin_list[0] = 'book'
print('my list:', my_list)
print('twin list:', twin_list)
print('Memory Address of my_list:', id(my_list))
print('Memory Address of twin_list:', id(twin_list))
````

````python
# Ex 9.38 - Method 6: Cloning using Shallow Copy (copy module)
import copy
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = copy.copy(my_list)

twin_list[0] = 'book'
print('my list:', my_list)
print('twin list:', twin_list)
print('Memory Address of my_list:', id(my_list))
print('Memory Address of twin_list:', id(twin_list))
````

````python
# Ex 9.39 - The Problem with Shallow Copy on NESTED lists
# Shallow copy copies references, not the nested list itself!
import copy
my_list = [['oat', 'egg'], ['bag', 'pen']]
twin_list = copy.copy(my_list)

twin_list[1][0] = 'book'
print('my list:', my_list)    # ALSO changed! Bug!
print('twin list:', twin_list)
print('Both lists changed because nested lists share references!')
````

````python
# Ex 9.40 - Method 7: Cloning using Deep Copy (solves nested list problem)
import copy
my_list = [['oat', 'egg'], ['bag', 'pen']]
twin_list = copy.deepcopy(my_list)

twin_list[1][0] = 'book'
print('my list:', my_list)    # unchanged! Deep copy works
print('twin list:', twin_list)
print('Only twin_list changed — deep copy creates fully independent copy')
````

## 🔢 Lesson 8 — Mathematical Operations on Lists

### Operations Using List Methods

```
┌────────────────────────────────────────────────────────────────────┐
│                  LIST MATHEMATICAL METHODS                         │
├──────────────────────────┬─────────────────────────────────────────┤
│  Method / Function       │  Description                            │
├──────────────────────────┼─────────────────────────────────────────┤
│  len(list)               │  Count of elements                      │
```

````python
# Ex 9.41 - len() — number of elements in list
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = [['oat', 'egg'], ['bag', 'pen']]

print('my list length:', len(my_list))    # 4
print('twin list length:', len(twin_list)) # 2 (2 inner lists)
````

````python
# Ex 9.42 - count() — count occurrences of a specific element
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = [['oat', 'egg'], ['bag', 'pen']]

print('Pen occurs in my_list:', my_list.count('pen'))       # 2
print('Pen occurs in twin_list:', twin_list.count('pen'))   # 0 (nested!)
````

````python
# Ex 9.43 - max() — largest element
n = [7, 6, 9, 2, 3]
my_list = ['bag', 'pen', 'cap', 'pen']

print('max of numbers list:', max(n))        # 9
print('max of my list (alphabetically):', max(my_list)) # pen (p > c > b)
````

````python
# Ex 9.44 - min() — smallest element
n = [7, 6, 9, 2, 3]
my_list = ['bag', 'pen', 'cap', 'pen']

print('min of numbers list:', min(n))        # 2
print('min of my list (alphabetically):', min(my_list)) # bag
````

````python
# Ex 9.45 - random.choice() — pick a random element
import random
players = ['Messi', 'Ronaldo', 'Salah']

print('The Winner is', random.choice(players))
print('Run again to see different results!')
````

````python
# Ex 9.46 - sum() — adds all numeric elements
n = [7, 6, 9, 2, 3]
print('Sum of numbers list:', sum(n))   # 27

# Note: max(), min(), sum() only work with homogeneous (same type) lists
# mixed = [1, 'hello']  → max(mixed) would give TypeError
````

### Operations Using List Operators

````python
# Ex 9.47 - Multiplication (*) operator — repeat elements
mom_list = ['oil', 'dye']

print('Old list:', mom_list)
print('2x list:', 2 * mom_list)
print('3x list:', 3 * mom_list)
````

````python
# Ex 9.48 - Relational (comparison) operators on lists
# Compares element by element, left to right
print([1, 2, 3] < [2, 2, 3])    # True  (1 < 2)
print([1, 2, 3] <= [1, 2, 3])   # True  (equal)
print([1, 2, 3, 4] > [1, 6, 3]) # False (2 < 6 at index 1)
print([1, 2, 3, 4] >= [1, 6, 3])# False
print([1, 2, 1] == [4])          # False
````

````python
# Ex 9.49 - String list comparison
# Considers: number of elements, order, content (case-sensitive)
x = ['abc', 'def', 'ghi']
y = ['abc', 'def', 'ghi']
z = ['ABC', 'DEF', 'GHI']   # uppercase
a = ['abc', 'def', 'ghi', 'jkl']  # more elements

print(x < z)    # False ('a' > 'A' in ASCII)
print(a < x)    # False (a has more elements)
print(x == y)   # True  (identical)
print(x == z)   # False (case different)
print(x == a)   # False (different length)
````

````python
# Ex 9.50 - Membership operators: in and not in
my_list = ['bag', 'pen', 'cap', 'pen']
twin_list = [['oat', 'egg'], ['bag', 'pen']]

print('my list:', my_list)
print('twin list:', twin_list)
print('bag in my_list:', 'bag' in my_list)         # True
print('Bag not in my_list:', 'Bag' not in my_list) # True (case sensitive!)
print('book in my_list:', 'book' in my_list)       # False
print('oat in twin_list:', 'oat' in twin_list)     # False (nested — checks inner lists, not their items)
````

## 📌 Lesson 9 — Introduction to Tuples

### What is a Tuple?

A **tuple** is a collection of objects that are **ordered** and **cannot be changed (immutable)**. Tuples use **parentheses** `( )` instead of square brackets.

### Characteristics of Tuples

```
┌──────────────────────────────────────────────────────────────────┐
│                    TUPLE CHARACTERISTICS                         │
├─────────────────────────┬────────────────────────────────────────┤
│  Feature                │  Description                           │
├─────────────────────────┼────────────────────────────────────────┤
│  Insertion order        │  Preserved (same order as created)     │
│  Immutable              │  CANNOT modify elements after creation │
│  Duplicates allowed     │  Same value can appear multiple times  │
│  Heterogeneous          │  Can store different types             │
└─────────────────────────┴────────────────────────────────────────┘
```

**When should you use a tuple?**  
When your data should **NOT change** — e.g., days of the week, months, coordinates, RGB colours.

````python
# Ex 9.51 - Empty Tuple
emp = ()
print(emp)
print(type(emp))
````

````python
# Ex 9.52 - Single Value Tuple — PROBLEM
# Without a trailing comma, Python treats it as just parentheses!
name = ("John")   # NOT a tuple — it's a string!
a = (11)          # NOT a tuple — it's an int!

print(name, type(name))
print(a, type(a))
````

````python
# Ex 9.53 - Single Value Tuple — SOLUTION: add trailing comma!
name = ("John",)   # tuple with one string
a = (11,)          # tuple with one int
b = 42,            # also a tuple (no parentheses, comma makes it a tuple)

print(name, type(name))
print(a, type(a))
print(b, type(b))
````

````python
# Ex 9.54 - Tuple with group of values
weekdays = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday")
print("Weekdays:", weekdays)
print(type(weekdays))
````

````python
# Ex 9.55 - Parentheses are OPTIONAL for tuples (comma creates tuple)
weekdays = "Monday", "Tuesday", "Wednesday", "Thursday", "Friday"
print("Weekdays:", weekdays)
print(type(weekdays))
````

````python
# Ex 9.56 - Create tuple using tuple() function from a list
my_list = [11, 22, 33]
my_tuple = tuple(my_list)
print(my_tuple)
print(type(my_tuple))
````

````python
# Ex 9.57 - Create tuple using range() function
my_tuple = tuple(range(1, 10, 2))
print(my_tuple)   # (1, 3, 5, 7, 9)
````

### Tuple Immutability

Once a tuple is created, you **cannot change its elements**. This is the key difference from lists.

````python
# Ex 9.58 - Tuple Immutability — TypeError when trying to modify
my_tuple = (10, 20, 30, 40)

try:
    my_tuple[1] = 70   # This will raise TypeError!
except TypeError as e:
    print('Error:', e)
    print('Tuples are immutable — you cannot change their elements!')
````

````python
# Ex 9.59 - sorted() function on Tuples
# Note: sorted() returns a LIST, not a tuple!
my_tuple = (40, 10, 30, 20)
sorted_result = sorted(my_tuple)        # ascending
print('Original tuple:', my_tuple)
print('Sorted result:', sorted_result)  # [10, 20, 30, 40] — a list!

# Sort descending
sorted_desc = sorted(my_tuple, reverse=True)
print('Sorted descending:', sorted_desc)
````

````python
# Ex 9.61 - Tuple Packing — group variables into a tuple
a, b, c, d = 10, 20, 30, 40
my_tuple = a, b, c, d
print('Packed tuple:', my_tuple)
````

````python
# Ex 9.62 - Tuple Unpacking — extract tuple values into separate variables
my_tuple = (10, 20, 30, 40)
a, b, c, d = my_tuple
print('a =', a, ' b =', b, ' c =', c, ' d =', d)

# Practical use: swap two values in one line!
x, y = 5, 10
print('\nBefore swap: x =', x, 'y =', y)
x, y = y, x    # tuple packing/unpacking swap
print('After swap:  x =', x, 'y =', y)
````

````python
# Bonus: Extended unpacking with *
my_tuple = (1, 2, 3, 4, 5)
first, *middle, last = my_tuple
print('First:', first)
print('Middle:', middle)
print('Last:', last)
````

### Difference Between List and Tuple

```
┌────────────────────────────────────────────────────────────────────────┐
│                    LIST vs TUPLE                                        │
├────────────────────────┬───────────────────┬───────────────────────────┤
│  Feature               │  List             │  Tuple                    │
├────────────────────────┼───────────────────┼───────────────────────────┤
│  Syntax                │  [1, 2, 3]        │  (1, 2, 3)                │
│  Mutability            │  Mutable ✎        │  Immutable 🔒             │
│  Performance           │  Slower           │  Faster                   │
│  Memory                │  More             │  Less                     │
│  Methods               │  Many             │  Few (count, index)       │
│  Use case              │  Data may change  │  Data won't change        │
│  Example               │  Shopping list    │  Days of the week         │
└────────────────────────┴───────────────────┴───────────────────────────┘
```

````python
# Tuple methods — only 2: count() and index()
my_tuple = (10, 20, 30, 20, 40, 20)

print('count of 20:', my_tuple.count(20))   # 3
print('index of 30:', my_tuple.index(30))   # 2

# Membership and other operators work too
print('20 in tuple:', 20 in my_tuple)       # True
print('max:', max(my_tuple))                # 40
print('min:', min(my_tuple))                # 10
print('len:', len(my_tuple))                # 6
print('sum:', sum(my_tuple))                # 140
````
