## 📦 Lesson 1 — Introduction to Variables

### What is a Variable?

A **variable** is the name we give to a memory location that holds some data. With a variable we can **store, access, and manipulate** data.

> *PS. A variable is a container where you can store a value.*

To summarise, a variable is:
- A **name** (a label we create)
- That **refers to a value**
- That **holds some data**
- That is the **name of a memory location**

```
  THE LIBRARY ANALOGY:

  Think about a librarian who stores and accesses thousands of books.
  They make a catalogue of books based on genre with a reference location.
  That is EXACTLY how variables work!

  ┌─────────────────────────────────────────────────────────────────────┐
  │                      COMPUTER MEMORY (RAM)                          │
  │                                                                     │
  │  Address: 0x1A2B           Address: 0x3C4D                         │
  │  ┌─────────────────────┐   ┌─────────────────────┐                 │
  │  │  Variable: age      │   │  Variable: name     │                 │
  │  │  Value:   21        │   │  Value:  "John"     │                 │
  │  └─────────────────────┘   └─────────────────────┘                 │
  │                                                                     │
  │  We use the variable NAME to find and use the data — just like     │
  │  a librarian uses a catalogue label to find the right shelf.       │
  └─────────────────────────────────────────────────────────────────────┘
```

### How to Create a Variable in Python

In Python, to create a variable we just need to specify the **name** and **assign a value** using `=`:

```
  Syntax:
  name_of_the_variable = value

  Example:
  age = 21
  ↑         ↑
  name      value

  The = sign is the ASSIGNMENT operator (not "equals" like in maths).
  It means: "store the value on the right into the name on the left".
```

### Variable Properties — Real Examples

Each variable has 5 properties. Here is how every property applies to a real Python variable:

```
  variable:  learner_name = "Sipho"

  ┌──────────────┬──────────────────────┬─────────────────────────────────────────────────────┐
  │  Property    │  Value for example   │  Meaning                                            │
  ├──────────────┼──────────────────────┼─────────────────────────────────────────────────────┤
  │  1. Type     │  str (string / text) │  The kind of data stored — text, number, True/False │
  │  2. Scope    │  Global              │  Where in the program the variable can be used       │
  │  3. Value    │  "Sipho"             │  The actual piece of data inside the container       │
  │  4. Location │  e.g. 0x7f3a2b1c    │  The memory address Python assigned automatically    │
  │  5. Lifetime │  Until del / end     │  How long the variable stays alive in memory         │
  └──────────────┴──────────────────────┴─────────────────────────────────────────────────────┘

Here are more everyday examples showing all 5 properties at a glance:

```
  age         = 17       → Type: int (integer)    │ Value: 17      │ Scope: wherever it is defined
  pass_rate   = 73.5     → Type: float (decimal)  │ Value: 73.5    │ Scope: wherever it is defined
  school_name = "DAQS"   → Type: str  (text)      │ Value: "DAQS"  │ Scope: wherever it is defined

````python
# Variable Properties Demo — all 5 properties in one block
learner_name = "Sipho"   # Type: str, Value: "Sipho"
age          = 17        # Type: int, Value: 17
pass_rate    = 73.5      # Type: float, Value: 73.5
has_passed   = True      # Type: bool, Value: True

# Check the TYPE of each variable using type()
print("learner_name →", learner_name,  "| Type:", type(learner_name))
print("age          →", age,           "| Type:", type(age))
print("pass_rate    →", pass_rate,     "| Type:", type(pass_rate))
print("has_passed   →", has_passed,    "| Type:", type(has_passed))

# Check the LOCATION (memory address) of each variable using id()
print("\nMemory address of learner_name:", id(learner_name))
print("Memory address of age         :", id(age))
````

### 3 Ways to Print a Variable with Text

Python gives you three styles for combining text and variables in a `print()` statement. All three produce the same output — choose whichever feels natural at your level:

```
  Variable:  name = "Sipho"   age = 17

  ┌──────────────────────────────────────────────────────────────────────────────────────────┐
  │  Method 1 — Comma  (simplest, adds a space automatically)                               │
  │  print("My name is", name, "and I am", age, "years old.")                               │
  │  Output: My name is Sipho and I am 17 years old.                                        │
  ├──────────────────────────────────────────────────────────────────────────────────────────┤
  │  Method 2 — f-string  (most flexible, reads like a sentence)                            │
  │  print(f"My name is {name} and I am {age} years old.")                                  │
  │  Note: Place the letter f before the opening quote.                                     │
  │        Wrap variable names in curly braces { } inside the string.                       │
  ├──────────────────────────────────────────────────────────────────────────────────────────┤
  │  Method 3 — Concatenation  (joining strings with +)                                     │
  │  print("My name is " + name + " and I am " + str(age) + " years old.")                 │
  │  Warning: all parts must be strings — use str() to convert numbers.                     │
  └──────────────────────────────────────────────────────────────────────────────────────────┘

  Recommended for beginners: Method 1 (comma) or Method 2 (f-string)
```

````python
# 3 ways to print — all produce the same output
name = "Sipho"
age  = 17

# Method 1: comma (Python adds a space between each item automatically)
print("Method 1 (comma)      :", "My name is", name, "and I am", age, "years old.")

# Method 2: f-string (put f before the quote, wrap variables in {})
print("Method 2 (f-string)   :", f"My name is {name} and I am {age} years old.")

# Method 3: concatenation (join with +, must convert numbers with str())
print("Method 3 (concat)     :", "My name is " + name + " and I am " + str(age) + " years old.")
````

### The `input()` Function

What if we want to submit a value from outside our code — like a questionnaire? We use `input()`.

**`input()`** is a function that can accept values from a user.

```
  Syntax:
  name_of_the_variable = input("Ask user to enter a value")

  Example:
  age = input("Enter your age?\n")

  The \n inside the string moves the cursor to a new line before the user types.

  Flow:
  Program runs → hits input() → PAUSES → waits for user to type → user presses Enter → continues
```

````python
# Ex 3.5 — Take input from a user
age = input("Enter your age?\n")
print("Your age is", age)
````

### What does `\n` mean inside a string?

`\n` is an **escape character** — a special code that tells Python to move to a **new line**.

```
  input("Enter your age?\n")
                         ↑
                         \n  →  moves cursor to a new line before user types

  Without \n:   Enter your age? █   (cursor on same line as the question)
  With    \n:   Enter your age?
               █               (cursor jumps to the next line — easier to read)
```

Other common escape characters you will meet soon:
```
  \n   →  new line          (like pressing Enter)
  \t   →  tab space         (like pressing Tab)
  \\   →  literal backslash
  \"   →  double quote inside a string
```

### 6 Rules for Naming Variables

```
┌────┬──────────────────────────────────────────────────┬────────────┬───────────────────────┐
│ #  │  Rule                                            │  Valid ✅  │  Invalid ❌            │
├────┼──────────────────────────────────────────────────┼────────────┼───────────────────────┤
│ 1  │  MUST start with a letter or underscore _        │  age  _x   │  1age  9score         │
│ 2  │  CANNOT start with a number                      │  score1    │  1score  2fast        │
│ 3  │  CANNOT use special symbols ($ @ # ! %)          │  my_var    │  $dollar  my@name     │
│ 4  │  CANNOT use spaces (use underscore instead)      │  my_age    │  my age  first name   │
│ 5  │  ARE case sensitive (age ≠ Age ≠ AGE)            │  age       │  Age  (if 'age' set)  │
│ 6  │  CANNOT be a Python keyword                      │  my_class  │  class  for  return   │
└────┴──────────────────────────────────────────────────┴────────────┴───────────────────────┘

  PS. Case sensitive means you CANNOT use lowercase letters in place of uppercase and vice versa.
```

### Naming Convention — Use `snake_case`

While you *can* name a variable anything that follows the 6 rules, Python programmers follow an agreed naming style called **snake_case**:

```
  snake_case:  all letters are LOWERCASE, words separated by underscores _

  ✅  Good (snake_case):          ❌  Avoid (other styles):
  first_name = "Sipho"           firstName = "Sipho"    ← camelCase (used in Java/JS)
  total_marks = 285              TotalMarks = 285       ← PascalCase (used for Classes)
  is_registered = True           TOTAL_MARKS = 285      ← ALL_CAPS (reserved for constants)
  school_name = "DAQS Academy"   totalmarks = 285       ← no separators (hard to read)
```

> **Why does it matter?** Python will accept any valid name — but snake_case makes your code easier to read, and all Python programmers worldwide recognise it as the standard variable style.

### Multiple Variables in a Single Line

We can assign multiple variables to multiple values in one single line. The number of variable names on the left must **equal** the number of values on the right — otherwise Python raises a `ValueError`.

```
  Syntax:
  var1, var2, var3 = value1, value2, value3

  ✅  Sid, age, grade = 4327, 24, 9    (3 names = 3 values)
  ❌  Sid, age, grade = 4327, 24       (3 names ≠ 2 values → ValueError)
  ❌  Sid, age = 4327, 24, 9           (2 names ≠ 3 values → ValueError)
```

### Multiple Assignment — Real-World Uses

Multiple assignment is useful when related data belongs together, such as a learner's profile or test results:

```
  # Learner profile — all on one line
  name, grade, age = "Nomsa Dlamini", 11, 16

  # Swapping two variables — no temporary variable needed!
  a, b = 10, 20
  a, b = b, a         ← after this: a = 20, b = 10  (values swapped!)

  # RGB colour values
  red, green, blue = 255, 128, 0

  # Coordinates
  latitude, longitude = -26.2041, 28.0473   ← Johannesburg, SA
```

> The **swap trick** (`a, b = b, a`) is one of the most elegant uses of multiple assignment. In most other languages you need a third variable to swap — Python does it in one line!

````python
# Multiple assignment real-world demo

# Learner profile in one line
name, grade, age = "Nomsa Dlamini", 11, 16
print(f"Learner: {name} | Grade: {grade} | Age: {age}")

# The swap trick — swap two values without a temp variable
score_a = 85
score_b = 72
print(f"\nBefore swap: score_a = {score_a}, score_b = {score_b}")
score_a, score_b = score_b, score_a
print(f"After swap : score_a = {score_a}, score_b = {score_b}")

# Same-value assignment — useful for initialising counters or scores
maths_mark = english_mark = science_mark = 0
print(f"\nAll marks initialised to: {maths_mark}, {english_mark}, {science_mark}")
````

## ⚙️ Lesson 3 — Variable Operations

```
  THREE OPERATIONS YOU CAN PERFORM ON VARIABLES:

  ┌─────────────────────────┬────────────────────────────────────────────────────────┐
  │  Operation              │  What it does                                          │
  ├─────────────────────────┼────────────────────────────────────────────────────────┤
  │  1. Memory Location     │  Find the memory address where a variable is stored    │
  │  2. Reassigning         │  Update / change the value stored in a variable        │
  │  3. Deleting            │  Remove the variable from memory completely            │
  └─────────────────────────┴────────────────────────────────────────────────────────┘
```

### Operation 1 — Memory Location using `id()`

You can get the memory address (location) of a variable using the `id()` function.

```
  Syntax:
  id(variable_name)

  PS. For simplicity, MA = Memory Address of a variable.
```

### How Python Stores Variables in Memory

When you create a variable, Python does three things automatically:

```
  Step 1:  Allocates a space in RAM (gives it an address)
  Step 2:  Stores your value at that address
  Step 3:  Links your variable name to that address

  Example:  subject = "Mathematics"

  Your code         Python's RAM
  ──────────        ────────────────────────────────────────
  subject  ────────► Address: 0x7f3a2b1c
                     Value:   "Mathematics"

  When you write:  print(subject)
  Python follows:  subject → 0x7f3a2b1c → "Mathematics" → prints it
```

When you **reassign** a variable, Python creates a **new memory location**:

```
  subject = "Mathematics"   → stored at address 0x7f3a2b1c
  subject = "Physics"       → stored at NEW address 0x9d4e5f2a
                              (old value "Mathematics" at 0x7f3a2b1c is released)

  Python's garbage collector automatically frees the old memory — you don't have to.
```

> This is why `id(variable)` **changes** every time you reassign a variable. The name points to a brand-new location in memory.

````python
# Ex 3.17 — Get the memory location of a variable
dollar = 69
print("1$ = ", dollar)
print("MA of dollar:", id(dollar))
````

### Operation 2 — Reassigning a Variable

You can **update or reassign** the value stored in a variable using the `=` operator again.

```
  Syntax:
  variable_name = new_value

  PS. When a variable is reassigned, its memory location ALSO changes!
```

````python
# Ex 3.18 — Updating the value of a variable (notice the memory address changes!)
dollar = 69
print("In 2019, $1 =", dollar)
print("MA of dollar in 2019:", id(dollar))

dollar = 72    # Reassign — give dollar a new value
print("In 2020, $1 =", dollar)
print("MA of dollar in 2020:", id(dollar))

# Notice: the memory address changed when we reassigned the value!
````

### Operation 3 — Deleting a Variable using `del`

You can **remove a variable from memory** entirely using the `del` keyword.

```
  Syntax:
  del variable_name

  After deletion — any attempt to use the variable raises a NameError.
```

````python
# Ex 3.19 — Deleting a variable
book = 5
print("Before delete — book:", book)

del book    # Variable is completely removed from memory

# Uncomment to see: NameError: name 'book' is not defined
# print(book)

print("After del book — accessing 'book' now would raise a NameError")
````

<div style='border: 2px solid #0077cc; border-radius: 12px; padding: 15px; background: #e8f4fd; margin: 15px 0;'>
  <strong style='color: #0077cc;'>💡 When should you use <code>del</code>?</strong>
  <ul style='margin: 8px 0 0;'>
    <li><strong>Free up memory</strong> — if you stored a very large amount of data and no longer need it</li>
    <li><strong>Security</strong> — remove sensitive data (like a password) from memory after use</li>
    <li><strong>Prevent accidental reuse</strong> — make sure a variable is not mistakenly used later in the program</li>
  </ul>
  <p style='margin: 10px 0 0;'>In most beginner programs you will not need <code>del</code> often — Python automatically cleans up unused variables. But it is important to know it exists.</p>
</div>

### Summary — All 3 Variable Operations at a Glance

```python
  price = 99          # create (assign)

  id(price)           # Operation 1: get memory address  → e.g. 140234567

  price = 120         # Operation 2: reassign (update)
  id(price)           # memory address CHANGED              → e.g. 140299811

  del price           # Operation 3: delete from memory
  # print(price)      # NameError — price no longer exists
```

### Constants

A **constant** is a type of variable whose value **cannot be changed**. It is helpful to think of constants as containers that hold information which should not be modified later.

```
  Example: GRAVITY = 9.81

  Convention: Write constant names in ALL_CAPITAL_LETTERS
  This signals to other programmers: "do not change this value"

  Examples of constants:
  GRAVITY   = 9.81
  PI        = 3.14159
  MAX_SCORE = 100
  SCHOOL    = "DAQSLearn Academy"

  PS. Python does not enforce constants — this is a programmer convention.
      We will discuss more about constants in a later chapter.
```

````python
# Constants — written in ALL_CAPS by convention
GRAVITY   = 9.81
PI        = 3.14159
MAX_SCORE = 100
SCHOOL    = "DAQSLearn Academy"

print("Gravity:", GRAVITY, "m/s²")
print("Pi:", PI)
print("Max possible score:", MAX_SCORE)
print("School:", SCHOOL)
````
