## 📦 Lesson 1 — Introduction to Variables

### What is a Variable?

A **variable** is the name we give to a memory location that holds some data. With a variable we can **store, access, and manipulate** data.

> *PS. A variable is a container where you can store a value.*

To summarise, a variable is:
- A **name** (a label we create)
- That **refers to a value**
- That **holds some data**
- That is the **name of a memory location**

```
  THE LIBRARY ANALOGY:

  Think about a librarian who stores and accesses thousands of books.
  They make a catalogue of books based on genre with a reference location.
  That is EXACTLY how variables work!

  ┌─────────────────────────────────────────────────────────────────────┐
  │                      COMPUTER MEMORY (RAM)                          │
  │                                                                     │
  │  Address: 0x1A2B           Address: 0x3C4D                         │
  │  ┌─────────────────────┐   ┌─────────────────────┐                 │
  │  │  Variable: age      │   │  Variable: name     │                 │
  │  │  Value:   21        │   │  Value:  "John"     │                 │
  │  └─────────────────────┘   └─────────────────────┘                 │
  │                                                                     │
  │  We use the variable NAME to find and use the data — just like     │
  │  a librarian uses a catalogue label to find the right shelf.       │
  └─────────────────────────────────────────────────────────────────────┘
```

### 5 Properties of a Variable

```
┌────────────────┬────────────────────────┬───────────────────────────────────────────────────┐
│  Property      │  Question it answers   │  Detail                                           │
├────────────────┼────────────────────────┼───────────────────────────────────────────────────┤
│  1. Type       │  What TYPE of data?    │  Each variable belongs to a data type             │
│                │                        │  (number, text, boolean, list...)                 │
├────────────────┼────────────────────────┼───────────────────────────────────────────────────┤
│  2. Scope      │  WHO can access it?    │  The part of the program within which the         │
│                │                        │  variable is accessible (local or global)          │
├────────────────┼────────────────────────┼───────────────────────────────────────────────────┤
│  3. Value      │  WHAT is stored?       │  The actual data — can be accessed or modified    │
│                │                        │  during the flow of the program                   │
├────────────────┼────────────────────────┼───────────────────────────────────────────────────┤
│  4. Location   │  WHERE is it stored?   │  The memory address — Python assigns this for you │
│                │                        │  Each memory location has an address               │
├────────────────┼────────────────────────┼───────────────────────────────────────────────────┤
│  5. Lifetime   │  TILL WHEN available?  │  The time period for which the memory location    │
│                │                        │  associated with the variable stores the data      │
└────────────────┴────────────────────────┴───────────────────────────────────────────────────┘
```

### How to Create a Variable in Python

In Python, to create a variable we just need to specify the **name** and **assign a value** using `=`:

```
  Syntax:
  name_of_the_variable = value

  Example:
  age = 21
  ↑         ↑
  name      value

  The = sign is the ASSIGNMENT operator (not "equals" like in maths).
  It means: "store the value on the right into the name on the left".
```

<div style='border: 2px solid #fd7e14; border-radius: 12px; padding: 15px; background: #fff8f0; margin: 15px 0;'>
  <strong style='color: #fd7e14;'>⚠️ Important:</strong> The variable name must ALWAYS be on the <strong>LEFT</strong> of <code>=</code> and the value on the <strong>RIGHT</strong>. Swapping them causes a <code>SyntaxError</code>.
</div>

````python
# Ex 3.1 — Create a variable
age = 21
````

````python
# Ex 3.2 — Print a variable
# Syntax: print(name_of_the_variable)
print(age)
````

````python
# Ex 3.4 — Print text along with a variable
# Syntax: print("write your message here", name_of_the_variable)
print("My age is", age)
````

### The `input()` Function

What if we want to submit a value from outside our code — like a questionnaire? We use `input()`.

**`input()`** is a function that can accept values from a user.

```
  Syntax:
  name_of_the_variable = input("Ask user to enter a value")

  Example:
  age = input("Enter your age?\n")

  The \n inside the string moves the cursor to a new line before the user types.

  Flow:
  Program runs → hits input() → PAUSES → waits for user to type → user presses Enter → continues
```

````python
# Ex 3.5 — Take input from a user
age = input("Enter your age?\n")
print("Your age is", age)
````

### Invalid Cases for Variables

While defining variables in Python:
- Variable **names** must be on the **LEFT side** of `=`
- **Values** must be on the **RIGHT side** of `=`

Violating this rule results in a **SyntaxError**.

````python
# Ex 3.6 — Creating a variable in wrong direction (value on left, name on right)
# Uncomment to see the error:
# 17 = age
# SyntaxError: cannot assign to literal here. Maybe you meant '==' instead of '='?

print("Variable name must ALWAYS be on the LEFT side of '='")
````

### 6 Rules for Naming Variables

```
┌────┬──────────────────────────────────────────────────┬────────────┬───────────────────────┐
│ #  │  Rule                                            │  Valid ✅  │  Invalid ❌            │
├────┼──────────────────────────────────────────────────┼────────────┼───────────────────────┤
│ 1  │  MUST start with a letter or underscore _        │  age  _x   │  1age  9score         │
│ 2  │  CANNOT start with a number                      │  score1    │  1score  2fast        │
│ 3  │  CANNOT use special symbols ($ @ # ! %)          │  my_var    │  $dollar  my@name     │
│ 4  │  CANNOT use spaces (use underscore instead)      │  my_age    │  my age  first name   │
│ 5  │  ARE case sensitive (age ≠ Age ≠ AGE)            │  age       │  Age  (if 'age' set)  │
│ 6  │  CANNOT be a Python keyword                      │  my_class  │  class  for  return   │
└────┴──────────────────────────────────────────────────┴────────────┴───────────────────────┘

  PS. Case sensitive means you CANNOT use lowercase letters in place of uppercase and vice versa.
```

````python
# Ex 3.7 — Rule 1: CAN start with an underscore
_age = 23
print(_age)
````

````python
# Ex 3.8 — Rule 2: CANNOT start with a number
# Uncomment to see: SyntaxError: invalid decimal literal
# 1dollar = 76

dollar1 = 76    # Correct version — number at the END is fine
print(dollar1)
````

````python
# Ex 3.9 — Rule 3: CANNOT use special symbols
# Uncomment to see: SyntaxError: invalid syntax
# $dollar = 74

dollar_amount = 74    # Correct version
print(dollar_amount)
````

````python
# Ex 3.10 — Rule 4: CANNOT use spaces (use underscore instead)
# Uncomment to see: SyntaxError: invalid syntax
# My Age = 24

my_age = 24    # Correct version — underscore instead of space
print(my_age)
````

````python
# Ex 3.11 — Rule 5: Variables ARE case sensitive
age = 21
# Uncomment to see: NameError: name 'Age' is not defined
# print(Age)

print(age)    # Correct — exact same case as when we defined it
````

All Python keywords contain only alphabet symbols. All are in lowercase **except** `True`, `False`, and `None`.

├───────────┼───────────┼───────────┼───────────┼───────────┼───────────┼───────────────┤
│  await    │  break    │  class    │  continue │  def      │  del      │  elif         │

│  import   │  in       │  is       │  lambda   │  nonlocal │  not      │  or           │

### Multiple Variables in a Single Line

We can assign multiple variables to multiple values in one single line. The number of variable names on the left must **equal** the number of values on the right — otherwise Python raises a `ValueError`.

```
  Syntax:
  var1, var2, var3 = value1, value2, value3

  ✅  Sid, age, grade = 4327, 24, 9    (3 names = 3 values)
  ❌  Sid, age, grade = 4327, 24       (3 names ≠ 2 values → ValueError)
  ❌  Sid, age = 4327, 24, 9           (2 names ≠ 3 values → ValueError)
```

````python
# Ex 3.13 — Assigning multiple variables in a single line
Sid, age, grade = 4327, 24, 9
print(age, grade, Sid)
````

````python
# Ex 3.14 — Unequal: too FEW values on the right (3 variables, only 2 values)
# Uncomment to see: ValueError: not enough values to unpack (expected 3, got 2)
# Sid, age, grade = 4327, 24

print("Number of variable names must equal number of values!")
````

````python
# Ex 3.15 — Unequal: too MANY values on the right (2 variables, 3 values)
# Uncomment to see: ValueError: too many values to unpack (expected 2)
# Sid, age = 4327, 24, 9

print("Always match the number of names with the number of values!")
````

### Single Value for Multiple Variables

We can also assign the **same single value** to **multiple variables** at the same time.

```
  Syntax:
  var1 = var2 = var3 = same_value

  Example:
  Sid = age = grade = 10
```

````python
# Ex 3.16 — Assign a single value to multiple variables simultaneously
Sid = age = grade = 10
print(Sid, age, grade)
````

## ⚙️ Lesson 3 — Variable Operations

```
  THREE OPERATIONS YOU CAN PERFORM ON VARIABLES:

  ┌─────────────────────────┬────────────────────────────────────────────────────────┐
  │  Operation              │  What it does                                          │
  ├─────────────────────────┼────────────────────────────────────────────────────────┤
  │  1. Memory Location     │  Find the memory address where a variable is stored    │
  │  2. Reassigning         │  Update / change the value stored in a variable        │
  │  3. Deleting            │  Remove the variable from memory completely            │
  └─────────────────────────┴────────────────────────────────────────────────────────┘
```

### Operation 1 — Memory Location using `id()`

You can get the memory address (location) of a variable using the `id()` function.

```
  Syntax:
  id(variable_name)

  PS. For simplicity, MA = Memory Address of a variable.
```

````python
# Ex 3.17 — Get the memory location of a variable
dollar = 69
print("1$ = ", dollar)
print("MA of dollar:", id(dollar))
````

### Operation 2 — Reassigning a Variable

You can **update or reassign** the value stored in a variable using the `=` operator again.

```
  Syntax:
  variable_name = new_value

  PS. When a variable is reassigned, its memory location ALSO changes!
```

````python
# Ex 3.18 — Updating the value of a variable (notice the memory address changes!)
dollar = 69
print("In 2019, $1 =", dollar)
print("MA of dollar in 2019:", id(dollar))

dollar = 72    # Reassign — give dollar a new value
print("In 2020, $1 =", dollar)
print("MA of dollar in 2020:", id(dollar))

# Notice: the memory address changed when we reassigned the value!
````

### Operation 3 — Deleting a Variable using `del`

You can **remove a variable from memory** entirely using the `del` keyword.

```
  Syntax:
  del variable_name

  After deletion — any attempt to use the variable raises a NameError.
```

````python
# Ex 3.19 — Deleting a variable
book = 5
print("Before delete — book:", book)

del book    # Variable is completely removed from memory

# Uncomment to see: NameError: name 'book' is not defined
# print(book)

print("After del book — accessing 'book' now would raise a NameError")
````

### Constants

A **constant** is a type of variable whose value **cannot be changed**. It is helpful to think of constants as containers that hold information which should not be modified later.

```
  Example: GRAVITY = 9.81

  Convention: Write constant names in ALL_CAPITAL_LETTERS
  This signals to other programmers: "do not change this value"

  Examples of constants:
  GRAVITY   = 9.81
  PI        = 3.14159
  MAX_SCORE = 100
  SCHOOL    = "DAQSLearn Academy"

  PS. Python does not enforce constants — this is a programmer convention.
      We will discuss more about constants in a later chapter.
```

````python
# Constants — written in ALL_CAPS by convention
GRAVITY   = 9.81
PI        = 3.14159
MAX_SCORE = 100
SCHOOL    = "DAQSLearn Academy"

print("Gravity:", GRAVITY, "m/s²")
print("Pi:", PI)
print("Max possible score:", MAX_SCORE)
print("School:", SCHOOL)
````
