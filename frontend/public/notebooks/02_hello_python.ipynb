## 🐍 Lesson 1 — Introduction to Python

To better communicate with a computer you need to learn a programming language — even though you have translators like compilers and interpreters. But **which** programming language?

The answer is simple: one that is used for **different and multiple purposes** — and that is **Python**.

### What is Python?

**Python** is a simple, general purpose, dynamic, object oriented, interpreted as well as compiled, high-level programming language.

```
┌─────────────────────────┬──────────────────────────────────────────────────────────────┐
│  Property               │  What it means                                               │
├─────────────────────────┼──────────────────────────────────────────────────────────────┤
│  Simple                 │  Easy to learn and code — reads almost like English           │
│  General Purpose        │  Can be used to build just about anything                    │
│  High-Level             │  Human-readable language — easily understandable             │
│  Easy to Debug          │  Easy to find and fix errors in your code                   │
│  Dynamic                │  No need to declare variable types — Python handles it       │
│  Object-Oriented        │  Code can be structured and grouped based on its behavior    │
│  Interpreted+Compiled   │  Source code → Bytecode → PVM → Output                      │
└─────────────────────────┴──────────────────────────────────────────────────────────────┘
```

### How Python Runs Your Code

In Python, the source code is first compiled into **bytecode**, then the bytecode is sent to the **PVM (Python Virtual Machine)** for execution. The PVM is an interpreter that runs the bytecode.

```
  Python Execution Flow:

  ┌──────────────┐    ┌──────────────────────┐    ┌──────────────┐    ┌───────────┐
  │  Your Code   │───►│  Python Compiler      │───►│  Bytecode    │───►│    PVM    │───► Output
  │  (source.py) │    │  (internal step)      │    │  (.pyc file) │    │(interpreter│
  └──────────────┘    └──────────────────────┘    └──────────────┘    └───────────┘

  PVM = Python Virtual Machine
  • The PVM reads bytecode and executes it on YOUR computer
  • Whether you are on Windows, Mac, or Linux — same bytecode runs everywhere
  • This is what makes Python PLATFORM INDEPENDENT
```

<div style='border: 2px solid #fd7e14; border-radius: 12px; padding: 15px; background: #fff8f0; margin: 15px 0;'>
  <strong style='color: #fd7e14;'>⚠️ Note:</strong> Python is basically called an <strong>interpreted language</strong>, but it can also be called a <strong>compiled-interpreted language</strong> because of this two-step process: compile to bytecode first, then interpret.
</div>

### 🎭 Why the Name "Python"?

There was a TV show called **"Monty Python's Flying Circus"** — a very popular fun show in the 1970s. While creating Python, Guido also used to read this show's published scripts. Guido needed a **short and unique name** for the language, so he named it **"Python"** — after the show, **not the snake!**

## 🌍 Lesson 2 — Applications of Python

Python is a programming language that **does it all** — from web applications to video games, data science, machine learning, real-time applications, and embedded systems. In this lesson we take a deeper dive into **10 major application areas**.

### 1. Web Development

Python is used to develop fully functional websites using web development frameworks like **Flask** and **Django**. Using Python for web development offers several benefits including security, easy scalability, and convenience in the development process.

### 2. Game Development

Just like web development, Python comes equipped with tools and libraries for **2D, 3D, and video game development** like **PyGame**.

### 3. Artificial Intelligence and Machine Learning

AI and Machine Learning are among the hottest topics of this decade. These are the brains behind smart technology we rely on to help make optimised decisions. Python has seen a steep increase in use for AI and ML-powered solutions like YouTube video recommendations. Libraries include **TensorFlow, PyTorch, Pandas, Scikit-Learn, NumPy, SciPy**, and more.

### 4. Desktop and Mobile Applications

Python offers options to desktop and mobile developers to build super-fast, responsive, and fully functional GUIs using tools like **Tkinter** and **Kivy**.

### 5. Image Processing

Due to the ever-increasing use of machine learning, image processing tools have skyrocketed. Python offers libraries for face recognition and image detection. Popular libraries include **OpenCV** and **Python Imaging Library (PIL)**.

### 6. Text Processing

Text processing is among the most common uses of Python. It handles enormous volumes of text flexibly. Ever wondered how Facebook detects hate speech posts and comments? That is done with Python's text processing capabilities using **NLTK** and **SpaCy**.

### 7. Audio and Video Applications

Python is fully equipped for audio work: basic signal processing, creative audio manipulation, audio recognition using libraries like **Pyo, pyAudioAnalysis, Dejavu**. For video: **Scikit-video, OpenCV, SciPy**. Popular streaming apps like **Spotify, Netflix**, and **YouTube** are built with Python.

### 8. Web Scraping Applications

Web scrapers crawl websites and collect information from web pages for researchers, analysts, and organisations. Tools include **PythonRequest, BeautifulSoup**, and **Selenium**.

### 9. Data Science and Data Visualization

Data science involves identifying problems, data collection, processing, exploration, analysis, and visualisation. Python's ecosystem offers libraries to tackle all stages. Most widely used data visualisation tools: **Matplotlib** and **Seaborn**.

### 10. Scientific and Numeric Applications

AI, ML, and data science projects require intensive computations — linear algebra, high-level mathematical functions. Python has helped scientists conclude countless number-crunching problems. Real-world examples: **FreeCAD** and **Abaqus** (engineering simulations).

## 📓 Lesson 3 — Python Installation

Python IDEs are software applications that provide tools and features to facilitate coding in Python. They offer features like code writing, execution, and debugging — making the development process more efficient.

### Jupyter Notebook

**Jupyter Notebook** is an open source web application that you can use to create and share live code. It is a popular IDE for data science and machine learning tasks — and it is the IDE we use in this course.

### How to Install Jupyter Notebook

```
  ANACONDA INSTALLATION STEPS:

  Step 1  → Download Anaconda
  Step 2  → Visit https://www.anaconda.com/download
  Step 3  → Select Windows / Mac / Linux
  Step 4  → Download the current Python version
  Step 5  → Open the downloaded file
  Step 6  → Click Continue
  Step 7  → Click 'Install for me only'
  Step 8  → Wait for: 'The installation was completed successfully'
  Step 9  → Click Close
```

### How to Write and Run Code on Jupyter Notebook

```
  Step 1  → Open Anaconda Prompt
  Step 2  → Select your desired folder from the list of directories
  Step 3  → Click New
  Step 4  → Click Python 3
  Step 5  → Write your code inside the cells
  Step 6  → Example: print("Hello World")
  Step 7  → Rename your file → Click File → Click Save As → Click Save
  Step 8  → Press 'Shift + Enter' to run your code
  Step 9  → You will get the output 'Hello World' on the next line below your code
             If not — you made a mistake, so recheck it once again!
```
