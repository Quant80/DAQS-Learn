## ➕ Lesson 1 — Introduction to Operators

A computer accepts input data, **processes** it, and produces output. Processing means doing some sort of calculation on data. Programming is more than storing data in variables — **operators** allow us to do exciting things with that data.

> **Operator:** A symbol applied on one or more **operands** (usually variables) to perform a certain action or operation.

```
  THE OPERATOR CONCEPT:

  a  =  5
  b  =  3
  c  =  a  +  b
          │     │
       operand  operand
           └──┬──┘
              +  ← operator
              │
              ▼
  c  =  8   (result stored in c)

  Note: + and = are operators
        a, b, c are operands
```

### Basic Classification of Operators

```
┌────────────────┬────────────────┬──────────────────────────────────────┐
│  Unary         │  1 operand     │  -5   (makes 5 negative)             │
│  Operator      │                │  -a   (negates variable a)           │
├────────────────┼────────────────┼──────────────────────────────────────┤
│  Binary        │  2 operands    │  a + b   (add two values)            │
│  Operator      │                │  a == b  (compare two values)        │
├────────────────┼────────────────┼──────────────────────────────────────┤
│  Ternary       │  3 operands    │  x if condition else y               │
│  Operator      │                │  (covered in Chapter 6 — if…else)   │
└────────────────┴────────────────┴──────────────────────────────────────┘
```

### Python's 9 Types of Operators

```
  1. Unary Minus       →  -  (negate)           THIS CHAPTER
  2. Assignment        →  =  (store value)      THIS CHAPTER
  3. Arithmetic        →  + - * / // % **       THIS CHAPTER
  4. Compound          →  += -= *= /= %= //=    THIS CHAPTER
  5. Relational        →  == != > < >= <=       THIS CHAPTER
  6. Identity          →  is  /  is not         THIS CHAPTER
  7. Logical           →  and  or  not          CHAPTER 6
  8. Bitwise           →  & | ^ ~ << >>         CHAPTER 6
  9. Membership        →  in  /  not in         CHAPTER 6
```

### Unary Minus Operator ( `-` )

The `-` operator is called the **Unary Minus** because it operates on a **single operand**. It is used to change a positive number to negative and vice-versa.

```
  a = 10

  print(a)   → 10   (positive)
  print(-a)  → -10  (flipped to negative)

  Rule: -( positive ) → negative
         -( negative ) → positive
```

````python
# Ex 5.1 — Unary Minus operator
a = 10
print(a)    # 10
print(-a)   # -10
````

## 📌 Lesson 2 — Assignment Operators

> **Assignment Operators** are operators used for **assigning values to variables**. `=` is the assignment operator.

You have been using `=` since Chapter 3. You can also use it to **copy** one variable into another.

```
  age = 21
      │
      = ← assignment operator
      │
  Memory:  [ age ] ─────────────► [ 21 ]

  Copying a variable:
  new_variable = old_variable

  v = 10
  c = v     ← c gets the value of v

  PS. Memory Address of both new (copied) and old variables is unique.
```

````python
# Ex 5.2 — Copy a variable using Assignment operator
v = 10
c = v
print("Old variable is", v)
print("Copied variable is", c)
print("Memory Address of Old variable:", id(v))
print("Memory Address of Copy variable", id(c))
````

## 🔢 Lesson 3 — Arithmetic Operators

> **Arithmetic Operators** perform basic mathematical calculations like addition, subtraction, multiplication, division etc. They are: `+`, `-`, `*`, `/`, `%`, `**`, `//`

```
┌────────────┬──────────────────┬──────────────────┬───────────────────┐
│  Operator  │  Operation       │  Example (a=20,  │  Result           │
│            │                  │   b=12)          │                   │
├────────────┼──────────────────┼──────────────────┼───────────────────┤
│     +      │  Addition        │  a + b           │  32               │
│     -      │  Subtraction     │  a - b           │  8                │
│     *      │  Multiplication  │  a * b           │  240              │
│     /      │  Division        │  a / b           │  1.6667 (float!)  │
│     //     │  Floor Division  │  a // b          │  1 (quotient)     │
│     %      │  Modulus         │  a % b           │  8 (remainder)    │
│     **     │  Exponentiation  │  a ** b          │  4096000000000000 │
└────────────┴──────────────────┴──────────────────┴───────────────────┘
```

<div style='border: 2px solid #0077cc; border-radius: 12px; padding: 15px; background: #e8f4fd; margin: 15px 0;'>
  <strong style='color: #0077cc;'>💡 Note on Division:</strong>
  <ul style='margin: 8px 0 0;'>
    <li>Division operator <code>/</code> always performs floating point arithmetic, so it returns a float value</li>
    <li>Floor division <code>//</code> can perform both: if values are <code>int</code> type → result is <code>int</code>; if at least one is <code>float</code> → result is <code>float</code></li>
  </ul>
</div>

````python
# Ex 5.3 — Arithmetic operators
a = 20
b = 12
print(a + b)          # Addition        → 32
print(a - b)          # Subtraction     → 8
print(a / b)          # Division        → 1.666... (always float!)
print(a // b)         # Quotient        → 1
print(a % b)          # Remainder       → 8
print(a * b)          # Multiplication  → 240
print(a ** b)         # Square          → 4096000000000000
print(a - b + 3*4/3)  # BODMAS          → 12.0
````

## 🔄 Lesson 4 — Compound Operators

> **Compound Operators** are a combination of some arithmetic and assignment operators (`+=`, `-=`, `*=`, `/=`, `%=`, `**=`, `//=`).

They update a variable **in place** — no need to write the variable name twice.

```
  WITHOUT compound (verbose):
  balance = balance + deposit

  WITH compound (clean):
  balance += deposit

  They do EXACTLY the same thing!
```

```
┌────────────────┬──────────────────────┬─────────────────────────────┐
│  Compound      │  Long Form Equivalent │  Operation                  │
├────────────────┼──────────────────────┼─────────────────────────────┤
│  x  += y      │  x = x + y           │  Add and assign             │
│  x  -= y      │  x = x - y           │  Subtract and assign        │
│  x  *= y      │  x = x * y           │  Multiply and assign        │
│  x  /= y      │  x = x / y           │  Divide and assign (float)  │
│  x  //= y     │  x = x // y          │  Floor divide and assign    │
│  x  %= y      │  x = x % y           │  Modulus and assign         │
│  x  **= y     │  x = x ** y          │  Exponentiate and assign    │
└────────────────┴──────────────────────┴─────────────────────────────┘

  Syntax:
  result_variable operator= operand_variable
```

````python
# Ex 5.5 — When you deposit on your bank account (WITHOUT compound)
balance = 12345
print("Balance before deposit:", balance)
deposit = 2655
balance = balance + deposit     # writing 'balance' twice!
print("Balance after deposit:", balance)
````

````python
# Ex 5.6 — Addition using Compound Operator (same result, cleaner code)
balance = 12345
print("Balance before deposit:", balance)
deposit = 2655
balance += deposit              # same as: balance = balance + deposit
print("Balance after deposit:", balance)
````

````python
# Ex 5.7 — All compound operators in sequence
a, b = 2, 3

a += b;   print("After += :", a)   # 2 + 3 = 5
a -= b;   print("After -= :", a)   # 5 - 3 = 2
a *= b;   print("After *= :", a)   # 2 * 3 = 6
a //= b;  print("After //=:", a)   # 6 // 3 = 2
a %= b;   print("After %= :", a)   # 2 % 3 = 2
a **= b;  print("After **=:", a)   # 2 ** 3 = 8
````

## ⚖️ Lesson 5 — Relational Operators

> **Relational Operators** are operators which are used to check for some relation like greater than or less than etc. between operands. They always return **`True`** or **`False`**.

```
  operand_A  [operator]  operand_B  →  True or False

  Example:  1  ==  2   →  False  (1 is not equal to 2)
            1  !=  2   →  True   (1 is indeed not equal to 2)
            1  <   2   →  True   (1 is less than 2)
```

```
┌────────────┬──────────────────────────┬──────────────┬─────────────┐
│  Operator  │  Meaning                 │  Example     │  Result     │
├────────────┼──────────────────────────┼──────────────┼─────────────┤
│    ==      │  Equal to                │  1 == 2      │  False      │
│    !=      │  Not equal to            │  1 != 2      │  True       │
│    >       │  Greater than            │  1 > 2       │  False      │
│    <       │  Less than               │  1 < 2       │  True       │
│    >=      │  Greater than or equal   │  1 >= 2      │  False      │
│    <=      │  Less than or equal      │  1 <= 2      │  True       │
└────────────┴──────────────────────────┴──────────────┴─────────────┘
```

<div style='border: 2px solid #fd7e14; border-radius: 12px; padding: 15px; background: #fff8f0; margin: 15px 0;'>
  <strong style='color: #fd7e14;'>⚠️ Common Mistake: = vs ==</strong>
  <ul style='margin: 8px 0 0;'>
    <li><code>=</code> is <strong>assignment</strong> — it STORES a value in a variable: <code>age = 21</code></li>
    <li><code>==</code> is <strong>comparison</strong> — it COMPARES two values and returns True/False: <code>age == 21</code></li>
    <li>Never confuse them — using <code>=</code> inside a condition causes a <code>SyntaxError</code></li>
  </ul>
</div>

````python
# Ex 5.8 — Relational operators
print(1 == 2)   # Equal              → False
print(1 != 2)   # Not Equal          → True
print(1 > 2)    # Less Than          → False
print(1 < 2)    # Greater Than       → True
print(1 <= 2)   # Less Than Or Equal → True
print(1 >= 2)   # Greater Than Or Equal → False
````

## 🔍 Lesson 6 — Identity Operators

> **Identity Operators** are operators which are used to check for identity by comparing the **memory location (address)** of two variables. With these operators, we can know whether two variables are pointing to the same location or not.

Recall: The memory location of an object can be seen using the `id()` function.

```
  There are two identity operators: is  and  is not

  1. is:
     A is B  →  True   if both A and B point to the SAME address
     A is B  →  False  if A and B point to DIFFERENT addresses

  2. is not:
     A is not B  →  True   if A and B point to DIFFERENT addresses
     A is not B  →  False  if A and B point to the SAME address

  Memory illustration:

  a = 25    b = 25    c = 30
  │          │          │
  ▼          ▼          ▼
  [addr:100] [addr:100] [addr:200]
      25         25         30

  a is b  →  True   (both point to addr 100)
  a is c  →  False  (a=addr100, c=addr200)

  PS. Python caches small integers (-5 to 256),
      so same small values may share an address.
```

````python
# Ex 5.9 — Identity operators
a, b, c = 25, 25, 30
print(a is b)     # True  — same cached integer object
print(a is c)     # False — different value, different object
print(id(a))
print(id(b))
print(id(c))
````

<div style='border: 2px solid #fd7e14; border-radius: 12px; padding: 15px; background: #fff8f0; margin: 15px 0;'>
  <strong style='color: #fd7e14;'>⚠️ Important: Identity operators are NOT comparing values of objects.</strong>
  <p style='margin: 8px 0 0;'>They compare <strong>memory addresses</strong> of objects. If we want to compare the <strong>value</strong> of objects, we should use the relational operator <code>==</code>.</p>
</div>

````python
# Ex 5.10 — Identity Operators vs Relational Operator
a = []
b = []
c = a       # c points to THE SAME list object as a

print(id(a))
print(id(b))
print(id(c))
print(a == b)   # True  — same values (both empty)
print(a is b)   # False — DIFFERENT objects in memory!
print(a is c)   # True  — c literally IS a
print(a == c)   # True
````

    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>Assignment</strong></td><td style='padding:8px;'>The <code>=</code> operator assigns values to variables; can copy a variable with <code>new = old</code></td></tr>
    <tr><td style='padding:8px;'><strong>Arithmetic</strong></td><td style='padding:8px;'><code>+  -  *  /  //  %  **</code> — <code>/</code> always returns float; Python follows BODMAS order</td></tr>
