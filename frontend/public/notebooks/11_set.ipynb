## 🔷 Lesson 1 — Introduction to Set

### What is a Set?

A **set** is a data type used to represent a group of **unique elements**. Sets are one of Python's four built-in data structures, alongside lists, tuples, and dictionaries.

> **Key Rule:** A set **cannot store duplicate elements**.

### Characteristics of Sets

```
┌───────────────────────────────────────────────────────────────────────┐
│                      SET CHARACTERISTICS                              │
├──────────────────────────────┬────────────────────────────────────────┤
│  Feature                     │  Description                           │
├──────────────────────────────┼────────────────────────────────────────┤
│  Insertion order NOT preserved│ Elements may appear in any order      │
│  No duplicates               │ Each element is unique                 │
│  Indexing/Slicing NOT allowed│ Cannot do set[0] or set[1:3]          │
│  Nested sets NOT allowed     │ Sets cannot contain other sets         │
│  Mutable                     │ Can add/remove elements after creation │
│  Heterogeneous               │ Can store different data types         │
└──────────────────────────────┴────────────────────────────────────────┘
```

### Bracket Reminder

```
  [ ] ──► List      (ordered, mutable, duplicates OK)
  ( ) ──► Tuple     (ordered, immutable, duplicates OK)
  { } ──► Set       (unordered, mutable, NO duplicates)
  { } ──► Dictionary (key:value pairs)   ← same brackets as set!

  ⚠️  {} creates an empty DICTIONARY, not an empty set.
      Use set() to create an empty set.
```

### How to Create a Set?

```
┌────────────────────────────────────────────────────────────────────┐
│                    3 WAYS TO CREATE A SET                          │
├──────────────────────────┬─────────────────────────────────────────┤
│  Method                  │  Syntax                                 │
├──────────────────────────┼─────────────────────────────────────────┤
│  Curly braces {}         │  my_set = {elem1, elem2, ...}           │
│  range() function        │  my_set = set(range(start, stop, step)) │
│  set() function          │  my_set = set(list_or_tuple)            │
│  Empty set               │  my_set = set()     ← NOT {}!          │
└──────────────────────────┴─────────────────────────────────────────┘
```

````python
# Ex 11.1 - Creating a set using curly braces {}
forward_players = {'Messi', 'Fati', 'Pedri'}

print('FCB Forwards:', forward_players)
print(type(forward_players))
print('Notice: order may differ from how you typed them!')
````

````python
# Ex 11.2 - Creating a set using range() function
Multiples_of_5 = set(range(5, 51, 5))

print('Multiples of 5 up to 50:', Multiples_of_5)
print(type(Multiples_of_5))
print('Insertion order is NOT preserved!')
````

````python
# Ex 11.3 - Duplicates are automatically removed
forward_players = {'Messi', 'Fati', 'Pedri', 'Fati'}  # 'Fati' appears twice

print('FCB Forwards:', forward_players)  # Fati appears only once
print('Duplicates are NOT allowed in sets!')
````

````python
# Ex 11.4 - The empty set trap — {} creates a dict, not a set!
d = {}
print('Type of {}:', type(d))   # dict, NOT set!

s = set()
print('Type of set():', type(s))  # set — correct!
print('Empty set:', s)
````

````python
# Creating a set from a list (removes duplicates automatically!)
grades = [85, 90, 78, 90, 85, 95, 78]
unique_grades = set(grades)
print('Original list (with duplicates):', grades)
print('Set (unique only):', unique_grades)
print('Useful trick: use set to remove duplicates from a list!')
````

## ➕ Lesson 2 — Insertion of Set Elements

```
┌─────────────────────────────────────────────────────────────────────────┐
│                    SET INSERTION METHODS                                │
├──────────────────────────┬──────────────────────────────────────────────┤
│  Method                  │  Description                                 │
├──────────────────────────┼──────────────────────────────────────────────┤
│  set.add(element)        │  Add ONE element                             │
│  set.update(set2)        │  Add MULTIPLE elements from another sequence │
│  new = set.copy()        │  Create a shallow copy of the set            │
└──────────────────────────┴──────────────────────────────────────────────┘

  Key Difference:
  ┌──────────────────────┬────────────────────────────────────────────┐
  │  add()               │  Takes ONE argument (a single element)     │
  │  update()            │  Takes sequence types (list, set, tuple)   │
  └──────────────────────┴────────────────────────────────────────────┘
```

````python
# Ex 11.5 - add() method — add ONE element
forward_players = {'Messi', 'Fati', 'Pedri'}
print('Before add:', forward_players)

forward_players.add('Ronaldo')
print('After add(Ronaldo):', forward_players)
````

````python
# Ex 11.6 - update() method — add MULTIPLE elements from a sequence
forward_players = {'Messi', 'Fati', 'Pedri'}
new_forwards = {'Ronaldo', 'Neymar'}

print('Before update:', forward_players)
forward_players.update(new_forwards)
print('After update:', forward_players)

# update() also works with lists and tuples
forward_players.update(['Salah', 'Mbappe'])
print('After update with list:', forward_players)
````

````python
# Ex 11.7 - copy() method — shallow copy of the set
forward_players = {'Messi', 'Fati', 'Pedri'}
copy_forwards = forward_players.copy()

# Modify copy — original not affected
copy_forwards.add('Ronaldo')
print('Original:', forward_players)   # unchanged
print('Copied (with Ronaldo):', copy_forwards)
````

## 🗑️ Lesson 3 — Deletion of Set Elements

```
┌──────────────────────────────────────────────────────────────────────────┐
│                     SET DELETION METHODS                                 │
├──────────────────────────┬──────────────────────┬────────────────────────┤
│  Method                  │  What is removed?    │  If not found?         │
├──────────────────────────┼──────────────────────┼────────────────────────┤
│  set.pop()               │  A RANDOM element    │  KeyError (if empty)   │
│  set.remove(element)     │  Specific element    │  KeyError raised!      │
│  set.discard(element)    │  Specific element    │  No error (safe)       │
│  set.clear()             │  ALL elements        │  N/A — always works    │
└──────────────────────────┴──────────────────────┴────────────────────────┘

  Key Rule: Use discard() when you're unsure if element exists.
             Use remove() when element must exist (stricter check).
```

````python
# Ex 11.8 - pop() — removes and returns a RANDOM element
forward_players = {'Messi', 'Fati', 'Pedri'}

removed = forward_players.pop()
print(removed, 'is sold.')
print('FCB Forwards remaining:', forward_players)
print('Note: which player gets sold is random!')
````

````python
# Ex 11.9 - remove() — removes a SPECIFIC element
forward_players = {'Messi', 'Fati', 'Pedri'}
forward_players.remove('Pedri')
print('FCB Forwards after remove(Pedri):', forward_players)
````

````python
# Ex 11.10 - remove() raises KeyError if element not found
forward_players = {'Messi', 'Fati', 'Pedri'}
try:
    forward_players.remove('Ronaldo')  # Ronaldo is not in the set!
except KeyError as e:
    print('KeyError:', e)
    print('remove() raises KeyError if element not found!')
````

````python
# Ex 11.11 - discard() — removes specific element, NO error if not found
forward_players = {'Messi', 'Fati', 'Pedri'}
forward_players.discard('Pedri')
print('After discard(Pedri):', forward_players)
````

````python
# Ex 11.12 - discard() does NOT raise error for missing elements
forward_players = {'Messi', 'Fati', 'Pedri'}
forward_players.discard('Ronaldo')  # Ronaldo not in set — no error!
print('After discard(Ronaldo) — set unchanged:', forward_players)
print('discard() is the SAFE version of remove()')
````

````python
# Ex 11.13 - clear() — removes ALL elements
forward_players = {'Messi', 'Fati', 'Pedri'}
print('Before clear:', forward_players)
forward_players.clear()
print('After clear():', forward_players)  # empty set
````

## ➗ Lesson 4 — Mathematical Operations on Sets

Sets in Python mirror the mathematical set theory you may know from Maths class. The operations are best visualised using **Venn diagrams**.

We'll use these two sets throughout this lesson:
- `barcelona_forwards = {'Messi', 'Fati', 'Pedri'}`
- `spain_forwards = {'Fati', 'Pedri', 'Torres'}`

```
  Venn Diagram Representation:

  ┌───────────────────────────────────────────────────┐
  │   barcelona_forwards        spain_forwards         │
  │   ┌─────────────────────────────────────────┐     │
  │   │   ┌──────────────┐   ┌──────────────┐   │     │
  │   │   │    Messi     │   │    Torres    │   │     │
  │   │   │          ┌───┴───┴───┐          │   │     │
  │   │   │          │ Fati      │          │   │     │
  │   │   └──────────│ Pedri     │──────────┘   │     │
  │   │              └───────────┘              │     │
  │   └─────────────────────────────────────────┘     │
  │                                                   │
  │  Left only: Messi    Both: Fati, Pedri  Right only: Torres │
  └───────────────────────────────────────────────────┘
```

### Complete Set Operations Reference

```
┌──────────────────────────────────────────────────────────────────────────────┐
│                     SET OPERATIONS SUMMARY                                   │
├────────────────────────────┬──────────────┬──────────────────────────────────┤
│  Operation                 │  Operator    │  Result                          │
├────────────────────────────┼──────────────┼──────────────────────────────────┤
│  union()                   │  A | B       │  All elements from A AND B       │
│  intersection()            │  A & B       │  Only elements in BOTH A and B   │
│  difference()              │  A - B       │  In A but NOT in B               │
│  symmetric_difference()    │  A ^ B       │  In A or B but NOT both          │
│  issubset()                │  A <= B      │  True if A is inside B           │
│  issuperset()              │  A >= B      │  True if A contains B            │
│  isdisjoint()              │  (none)      │  True if A and B share nothing   │
└────────────────────────────┴──────────────┴──────────────────────────────────┘

  Commutative rules:
  A | B == B | A   (union is commutative)
  A & B == B & A   (intersection is commutative)
  A ^ B == B ^ A   (symmetric difference is commutative)
  A - B != B - A   (difference is NOT commutative)
```

````python
# Ex 11.14 - union() — ALL elements from BOTH sets (no duplicates)
#
#  Messi  ┌──────────────┐  Torres
#         │ Fati  Pedri  │
#  Result: ███████████████  ← everything shaded

barcelona_forwards = {'Messi', 'Fati', 'Pedri'}
spain_forwards = {'Fati', 'Pedri', 'Torres'}

# Two ways to write union:
print('union() method:', barcelona_forwards.union(spain_forwards))
print('| operator:    ', barcelona_forwards | spain_forwards)
print('Spanish & FCB Players:', barcelona_forwards | spain_forwards)
````

````python
# Ex 11.15 - intersection() — only elements in BOTH sets
#
#  Messi  ┌──────────────┐  Torres
#         │█Fati  Pedri█ │
#  Result: only middle shaded

barcelona_forwards = {'Messi', 'Fati', 'Pedri'}
spain_forwards = {'Fati', 'Pedri', 'Torres'}

print('intersection() method:', barcelona_forwards.intersection(spain_forwards))
print('& operator:           ', spain_forwards & barcelona_forwards)
print('Spanish FCB Players:', spain_forwards & barcelona_forwards)
````

````python
# Ex 11.16 - intersection_update() — updates the set IN-PLACE with the intersection
barcelona_forwards = {'Messi', 'Fati', 'Pedri'}
spain_forwards = {'Fati', 'Pedri', 'Torres'}

barcelona_forwards.intersection_update(spain_forwards)
print('Barcelona (updated to intersection):', barcelona_forwards)
````

````python
# Ex 11.17 - difference() — in A but NOT in B  (A - B)
#
#  ██████ ┌──────────────┐  Torres
#  Messi  │ Fati  Pedri  │
#  Result: only LEFT side shaded (elements only in barcelona)

barcelona_forwards = {'Messi', 'Fati', 'Pedri'}
spain_forwards = {'Fati', 'Pedri', 'Torres'}

print('A - B (Barcelona NOT Spain):', barcelona_forwards - spain_forwards)   # Messi
print('B - A (Spain NOT Barcelona):', spain_forwards - barcelona_forwards)   # Torres
print('Note: A-B ≠ B-A  (NOT commutative!)')
````

````python
# Ex 11.18 & 11.19 - difference_update() — updates set IN-PLACE
barcelona_forwards = {'Messi', 'Fati', 'Pedri'}
spain_forwards = {'Fati', 'Pedri', 'Torres'}

barcelona_forwards.difference_update(spain_forwards)
print('Non-Spanish FCB Player (updated):', barcelona_forwards)  # Messi

barcelona_forwards2 = {'Messi', 'Fati', 'Pedri'}
spain_forwards2 = {'Fati', 'Pedri', 'Torres'}

spain_forwards2.difference_update(barcelona_forwards2)
print('Spanish, not FCB Player (updated):', spain_forwards2)   # Torres
````

````python
# Ex 11.20 - symmetric_difference() — in A or B but NOT both  (A ^ B)
#
#  ██████ ┌──────────────┐  ███████
#  Messi  │ Fati  Pedri  │  Torres
#  Result: LEFT and RIGHT shaded — middle excluded

barcelona_forwards = {'Messi', 'Fati', 'Pedri'}
spain_forwards = {'Fati', 'Pedri', 'Torres'}

print('symmetric_difference():', barcelona_forwards.symmetric_difference(spain_forwards))
print('^ operator:            ', barcelona_forwards ^ spain_forwards)
print('Players NOT playing for BOTH FCB & Spain:', barcelona_forwards ^ spain_forwards)
````

````python
# Ex 11.21 - symmetric_difference_update() — updates set IN-PLACE
barcelona_forwards = {'Messi', 'Fati', 'Pedri'}
spain_forwards = {'Fati', 'Pedri', 'Torres'}

barcelona_forwards.symmetric_difference_update(spain_forwards)
print('Players NOT in both (updated):', barcelona_forwards)
````

### issubset(), issuperset(), isdisjoint()

```
  issubset(B)  — Is every element of A also in B?
  ┌──────────────────────────────────────────┐
  │   B = { Fati, Pedri, Messi, Torres }     │
  │   ┌──────────────────────┐               │
  │   │ A = { Fati, Pedri } ◄── subset of B │
  │   └──────────────────────┘               │
  │   A.issubset(B) → True                   │
  └──────────────────────────────────────────┘

  issuperset(B) — Does A contain all elements of B?
  isdisjoint(B) — Do A and B share NO elements? (No overlap)
```

````python
# Ex 11.22 - issubset()
barcelona_forwards = {'Messi', 'Fati', 'Pedri'}
spain_forwards = {'Fati', 'Pedri'}

print('Is barcelona a subset of spain?', barcelona_forwards.issubset(spain_forwards))  # False
print('Is spain a subset of barcelona?', spain_forwards.issubset(barcelona_forwards))  # True
````

````python
# Ex 11.23 - issuperset()
barcelona_forwards = {'Messi', 'Fati', 'Pedri'}
spain_forwards = {'Fati', 'Pedri'}

print('Is barcelona a superset of spain?', barcelona_forwards.issuperset(spain_forwards))  # True
print('Is spain a superset of barcelona?', spain_forwards.issuperset(barcelona_forwards))  # False
````

````python
# Ex 11.24 - isdisjoint() — no elements in common → True
barcelona_forwards = {'Messi', 'Fati', 'Pedri'}
spain_forwards = {'Fati', 'Pedri'}
argentinian_forwards = {'Messi', 'Aguero'}

print('Barcelona ∩ Spain = {}?', barcelona_forwards.isdisjoint(spain_forwards))        # False (common: Fati, Pedri)
print('Spain ∩ Argentina = {}?', spain_forwards.isdisjoint(argentinian_forwards))      # True (no overlap)
print('Argentina ∩ Spain = {}?', argentinian_forwards.isdisjoint(spain_forwards))      # True
````

## 🔒 Frozen Set

A **frozen set** is exactly like a set, but it is **immutable** — once created, its elements cannot be changed.

```
  Regular Set  ──►  Mutable   (add/remove elements)
  Frozen Set   ──►  Immutable (read-only — cannot change after creation)

  Use frozen sets when:
  • You need a set as a dictionary key (sets can't be keys, frozen sets can)
  • You want to ensure the set cannot be accidentally changed

  Syntax: frozen_set_name = frozenset(set_name)
```

````python
# Ex 11.25 - Creating a frozen set
barcelona_forwards = {'Messi', 'Fati', 'Pedri'}
frozen_barcelona = frozenset(barcelona_forwards)

print(frozen_barcelona)
print(type(frozen_barcelona))

# Frozen sets support all read operations (membership, iteration, etc.)
print('Is Messi in frozen set?', 'Messi' in frozen_barcelona)
print('Length:', len(frozen_barcelona))

# But NOT write operations:
try:
    frozen_barcelona.add('Ronaldo')  # This will fail!
except AttributeError as e:
    print('Error trying to add:', e)
````

## 📊 Other Mathematical Operations on Sets

The following operations work on sets exactly as they do on lists:

```
┌────────────────────────────────────────────────────────────────────┐
│              SET MATHEMATICAL FUNCTIONS                            │
├───────────────────────┬────────────────────────────────────────────┤
│  Function/Operator    │  Description                               │
├───────────────────────┼────────────────────────────────────────────┤
│  len(set)             │  Number of elements                        │
│  max(set)             │  Largest element (homogeneous only)        │
│  min(set)             │  Smallest element (homogeneous only)       │
│  sum(set)             │  Sum of all numeric elements               │
│  elem in set          │  Membership test → True/False              │
│  elem not in set      │  Non-membership test → True/False          │
│  set1 == set2         │  Equality check (order doesn't matter)     │
└───────────────────────┴────────────────────────────────────────────┘
```

````python
# Set mathematical operations
scores = {85, 90, 78, 95, 72}

print('Set:', scores)
print('Length:', len(scores))
print('Max:', max(scores))
print('Min:', min(scores))
print('Sum:', sum(scores))
print('Average:', sum(scores) / len(scores))
print('Is 90 in scores?', 90 in scores)
print('Is 100 in scores?', 100 not in scores)

# Sets are equal if they contain the same elements (order irrelevant)
a = {1, 2, 3}
b = {3, 1, 2}
print('\n{1,2,3} == {3,1,2}?', a == b)  # True
````

## 📋 Key Takeaways

<div style='border: 2px solid #6c757d; border-radius: 12px; padding: 20px; background: #f8f9fa; margin: 20px 0;'>
  <h3 style='color: #6c757d; margin-top: 0;'>📋 Key Takeaways — Chapter 11</h3>
  <table style='width:100%; border-collapse: collapse;'>
    <tr style='background:#e9ecef;'><th style='padding:8px; text-align:left;'>Concept</th><th style='padding:8px; text-align:left;'>Key Point</th></tr>
    <tr><td style='padding:8px;'><strong>Set</strong></td><td style='padding:8px;'>A collection of unique, unordered elements — no duplicates allowed</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>Set Characteristics</strong></td><td style='padding:8px;'>Unordered, no duplicates, no indexing/slicing, no nested sets, mutable</td></tr>
    <tr><td style='padding:8px;'><strong>Creating a Set</strong></td><td style='padding:8px;'>Use <code>{elem1, elem2}</code> or <code>set(range(...))</code> or <code>set(list)</code>. Use <code>set()</code> for empty — not <code>{}</code>!</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>add() vs update()</strong></td><td style='padding:8px;'><code>add()</code> adds one element; <code>update()</code> adds multiple from a sequence</td></tr>
    <tr><td style='padding:8px;'><strong>copy()</strong></td><td style='padding:8px;'>Returns a shallow copy of the set</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>remove() vs discard()</strong></td><td style='padding:8px;'><code>remove()</code> raises KeyError if element missing; <code>discard()</code> does not</td></tr>
    <tr><td style='padding:8px;'><strong>pop()</strong></td><td style='padding:8px;'>Removes and returns a random element; raises KeyError if set is empty</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>union() — A | B</strong></td><td style='padding:8px;'>All elements from both sets; commutative (A|B = B|A)</td></tr>
    <tr><td style='padding:8px;'><strong>intersection() — A & B</strong></td><td style='padding:8px;'>Only elements present in BOTH sets; commutative</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>difference() — A - B</strong></td><td style='padding:8px;'>Elements in A but NOT in B; NOT commutative (A-B ≠ B-A)</td></tr>
    <tr><td style='padding:8px;'><strong>symmetric_difference() — A ^ B</strong></td><td style='padding:8px;'>Elements in A or B but NOT both; commutative</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>issubset() / issuperset()</strong></td><td style='padding:8px;'>Test whether one set is contained in (or contains) another</td></tr>
    <tr><td style='padding:8px;'><strong>isdisjoint()</strong></td><td style='padding:8px;'>True if two sets share NO elements at all</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>Frozen Set</strong></td><td style='padding:8px;'>Immutable version of a set — cannot be changed after creation; created with <code>frozenset()</code></td></tr>
