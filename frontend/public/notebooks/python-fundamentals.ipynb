````python
# ── Auto-Grader Setup ─────────────────────────────────────────────────────────
# Run this cell first! It sets up the scoring system.

class Grader:
    def __init__(self):
        self.results = []
        self.total = 0
        self.passed = 0

    def check(self, question, condition, points=1, hint=""):
        self.total += points
        if condition:
            self.passed += points
            self.results.append(f"  ✅  Q{len(self.results)+1}: {question} (+{points} pt{'s' if points>1 else ''})")
        else:
            msg = f"  ❌  Q{len(self.results)+1}: {question}"
            if hint:
                msg += f"\n       💡 Hint: {hint}"
            self.results.append(msg)

    def score(self):
        print("=" * 55)
        print("  📊  DAQS LEARN — ASSIGNMENT RESULTS")
        print("=" * 55)
        for r in self.results:
            print(r)
        pct = round(self.passed / self.total * 100) if self.total else 0
        grade = ("🏆 Distinction" if pct >= 80 else
                 "🥇 Merit"      if pct >= 65 else
                 "🥈 Pass"       if pct >= 50 else
                 "📖 Keep Studying")
        print("=" * 55)
        print(f"  SCORE: {self.passed}/{self.total}  ({pct}%)  {grade}")
        print("=" * 55)

grader = Grader()
print("✅ Grader ready! Complete the assignment cells below, then run grader.score()")
````

# Chapter 1 — Variables & Data Types

## 📖 Theory

A **variable** is a named container that stores a value. In Python you create one with `=`:

```python
age = 17          # integer
height = 1.75     # float
name = "Themba"   # string
passed = True     # boolean
```

Python has **4 core data types**:

| Type | Example | `type()` returns |
|------|---------|------------------|
| Integer | `42` | `<class 'int'>` |
| Float | `3.14` | `<class 'float'>` |
| String | `"hello"` | `<class 'str'>` |
| Boolean | `True` / `False` | `<class 'bool'>` |

### 🔑 Key Rules
- Variable names are **case-sensitive**: `age` ≠ `Age`
- Use **snake_case** for names: `student_name` not `StudentName`
- Cannot start with a number: ❌ `1name` → ✅ `name1`
- Cannot use Python keywords: ❌ `for = 5`

````python
# ── Example: Variables in action ──────────────────────────────────────────────
student_name = "Themba Moyo"
student_age  = 17
average_mark = 78.5
has_passed   = True

print("Name:", student_name)
print("Age:", student_age)
print("Average:", average_mark)
print("Passed:", has_passed)
print()

# Check types
print("Types:")
print(f"  student_name → {type(student_name)}")
print(f"  student_age  → {type(student_age)}")
print(f"  average_mark → {type(average_mark)}")
print(f"  has_passed   → {type(has_passed)}")
````

````python
# ── Type Conversion ────────────────────────────────────────────────────────────
# Convert between types using int(), float(), str(), bool()

x = "42"          # this is a STRING
y = int(x)        # convert to integer
z = float(x)      # convert to float

print(f"'{x}' as int   → {y}  (type: {type(y).__name__})")
print(f"'{x}' as float → {z}  (type: {type(z).__name__})")
print()

# Useful: input() always returns a string, so convert!
age_str = "17"           # simulating user input
age_int = int(age_str)
print(f"Next year you will be {age_int + 1}")
````

# Chapter 2 — Operators & Expressions

## 📖 Theory

### Arithmetic Operators
| Operator | Name | Example | Result |
|----------|------|---------|--------|
| `+` | Addition | `7 + 3` | `10` |
| `-` | Subtraction | `7 - 3` | `4` |
| `*` | Multiplication | `7 * 3` | `21` |
| `/` | Division | `7 / 3` | `2.333...` |
| `//` | Floor Division | `7 // 3` | `2` |
| `%` | Modulus (remainder) | `7 % 3` | `1` |
| `**` | Exponent | `2 ** 8` | `256` |

### Comparison Operators (return True/False)
| Operator | Meaning |
|----------|---------|
| `==` | Equal to |
| `!=` | Not equal to |
| `>` / `<` | Greater / Less than |
| `>=` / `<=` | Greater or equal / Less or equal |

### Logical Operators
| Operator | Meaning | Example |
|----------|---------|--------|
| `and` | Both must be True | `age > 16 and has_id == True` |
| `or` | At least one True | `is_admin or is_teacher` |
| `not` | Inverts True/False | `not is_banned` |

````python
# ── Example: School marks calculator ─────────────────────────────────────────
maths   = 78
science = 85
english = 62
history = 71

total    = maths + science + english + history
average  = total / 4
passed   = average >= 50

print(f"Total:   {total}")
print(f"Average: {average:.1f}%")
print(f"Passed:  {passed}")
print()

# Modulus example — useful for checking even/odd
for n in [10, 17, 24, 33]:
    parity = "even" if n % 2 == 0 else "odd"
    print(f"  {n} is {parity}")
````

````python
# ── Example: Augmented assignment (shorthand) ─────────────────────────────────
score = 0
score += 10   # same as: score = score + 10
score += 5
score *= 2    # same as: score = score * 2
print(f"Final score: {score}")   # expect 30

# Operator precedence — PEMDAS / BODMAS
result = 2 + 3 * 4       # multiplication first → 14 (not 20)
result2 = (2 + 3) * 4    # brackets first → 20
print(f"2 + 3 * 4   = {result}")
print(f"(2 + 3) * 4 = {result2}")
````

# Chapter 3 — Strings & Formatting

## 📖 Theory

Strings are sequences of characters. Python gives you powerful tools to work with them:

### Creating Strings
```python
s1 = 'single quotes'
s2 = "double quotes"
s3 = """triple quotes
span multiple lines"""
```

### Key String Methods
| Method | What it does | Example |
|--------|--------------|---------|
| `.upper()` | All uppercase | `"hello".upper()` → `"HELLO"` |
| `.lower()` | All lowercase | `"HELLO".lower()` → `"hello"` |
| `.strip()` | Remove whitespace | `" hi ".strip()` → `"hi"` |
| `.replace(a,b)` | Replace text | `"cat".replace("c","b")` → `"bat"` |
| `.split(sep)` | Split into list | `"a,b,c".split(",")` → `["a","b","c"]` |
| `.startswith(s)` | Starts with? | `"Python".startswith("Py")` → `True` |
| `.count(s)` | Count occurrences | `"banana".count("a")` → `3` |
| `len(s)` | Length | `len("hello")` → `5` |

### f-strings (Formatted Strings) ← Most important!

# Chapter 4 — Control Flow

## 📖 Theory

Control flow lets your program **make decisions** and **repeat actions**.

### if / elif / else
```python
mark = 74

if mark >= 80:
    grade = "A"
elif mark >= 70:
    grade = "B"
elif mark >= 60:
    grade = "C"
elif mark >= 50:
    grade = "D"
else:
    grade = "F"

print(f"Mark: {mark} → Grade: {grade}")   # Mark: 74 → Grade: B
```

### for loops — iterate over a sequence
```python
for i in range(5):        # 0 1 2 3 4
    print(i)

for item in ["a", "b", "c"]:
    print(item)

for i, item in enumerate(["x", "y", "z"]):
    print(i, item)        # 0 x / 1 y / 2 z
```

### while loops — repeat while condition is True
```python
count = 0
while count < 5:
    print(count)
    count += 1    # ⚠️ always change the condition or you get infinite loop!
```

### Loop control
- `break` → exit the loop immediately
- `continue` → skip to next iteration
- `pass` → do nothing (placeholder)

````python
# ── Example: Grade calculator with loop ──────────────────────────────────────
marks = [45, 78, 92, 56, 83, 61, 34, 77]

def get_grade(mark):
    if mark >= 80:   return "A"
    elif mark >= 70: return "B"
    elif mark >= 60: return "C"
    elif mark >= 50: return "D"
    else:            return "F"

print(f"{'Mark':>6}  {'Grade':>5}  {'Status':>10}")
print("-" * 28)
for i, mark in enumerate(marks, 1):
    grade  = get_grade(mark)
    status = "✅ Pass" if mark >= 50 else "❌ Fail"
    print(f"  {i}.  {mark:>3}     {grade}     {status}")

passing = [m for m in marks if m >= 50]
print(f"\nPass rate: {len(passing)}/{len(marks)} ({len(passing)/len(marks)*100:.0f}%)")
````

````python
# ── Example: FizzBuzz (classic programming challenge) ─────────────────────────
# Rules: 1-30, print Fizz if divisible by 3, Buzz if by 5, FizzBuzz if both

results = []
for n in range(1, 31):
    if n % 3 == 0 and n % 5 == 0:
        results.append("FizzBuzz")
    elif n % 3 == 0:
        results.append("Fizz")
    elif n % 5 == 0:
        results.append("Buzz")
    else:
        results.append(str(n))

print(" ".join(results))
````

# Chapter 5 — Functions

## 📖 Theory

Functions are **reusable blocks of code**. Define once, use many times.

```python
#  keyword  name    parameters
def greet(name, greeting="Hello"):
    """Greet a person. greeting defaults to 'Hello'."""
    return f"{greeting}, {name}!"

greet("Themba")               # "Hello, Themba!"
greet("Siyanda", "Sawubona")  # "Sawubona, Siyanda!"
```

### Key Concepts
| Concept | Explanation |
|---------|-------------|
| **Parameters** | Variables in the function definition |
| **Arguments** | Values passed when calling the function |
| **Default values** | `def f(x=10)` — used if argument not provided |
| **return** | Sends a value back to the caller |
| **Docstring** | `"""..."""` — documents what the function does |

### Lambda (anonymous) functions
```python
square = lambda x: x ** 2
square(5)    # 25

# Useful with sorted(), map(), filter()
names = ["Zanele", "Bonga", "Sipho"]
sorted(names, key=lambda n: len(n))  # sort by name length
```

````python
# ── Example: Student report card generator ────────────────────────────────────
def calculate_average(marks):
    """Return the average of a list of marks."""
    return sum(marks) / len(marks) if marks else 0

def get_symbol(mark):
    """Return CAPS symbol for a mark (South African grading)."""
    if mark >= 80:   return "7 (Outstanding)"
    elif mark >= 70: return "6 (Meritorious)"
    elif mark >= 60: return "5 (Substantial)"
    elif mark >= 50: return "4 (Adequate)"
    elif mark >= 40: return "3 (Moderate)"
    elif mark >= 30: return "2 (Elementary)"
    else:            return "1 (Not achieved)"

def print_report(student_name, subject_marks):
    """Print a formatted report card."""
    avg = calculate_average(list(subject_marks.values()))
    print(f"\n{'='*45}")
    print(f"  DAQS LEARN — STUDENT REPORT")
    print(f"  {student_name}")
    print(f"{'='*45}")
    for subject, mark in subject_marks.items():
        symbol = get_symbol(mark)
        print(f"  {subject:<18} {mark:>3}%   Level {symbol[0]}")
    print(f"{'─'*45}")
    print(f"  {'AVERAGE':<18} {avg:>5.1f}%  Level {get_symbol(avg)[0]}")
    print(f"{'='*45}")

print_report("Themba Moyo", {
    "Mathematics": 78,
    "Physical Science": 85,
    "English HL": 61,
    "Life Sciences": 74,
    "History": 69,
})
````

# Chapter 6 — Lists & Dictionaries

## 📖 Theory

### Lists — ordered, mutable sequences
```python
marks = [78, 85, 61, 74, 69]
marks.append(90)        # add to end
marks.insert(0, 55)     # insert at index 0
marks.remove(61)        # remove first occurrence of 61
marks.sort()            # sort ascending
marks.sort(reverse=True)  # sort descending
marks[0]                # first element
marks[-1]               # last element
marks[1:4]              # slice: elements 1, 2, 3
len(marks)              # number of elements
sum(marks)              # sum of all elements
max(marks) / min(marks) # highest / lowest
```

### List Comprehensions — elegant one-liners
```python
# Traditional:
passing = []
for m in marks:
    if m >= 50:
        passing.append(m)

# Comprehension (same result, one line!):
passing = [m for m in marks if m >= 50]
doubled = [m * 2 for m in marks]
```

### Dictionaries — key:value pairs
```python
student = {
    "name": "Themba",
    "age": 17,
    "marks": [78, 85, 61]
}
student["name"]            # access value
student["email"] = "..."   # add new key
student.get("grade", "N/A") # safe access with default
student.keys()             # all keys
student.values()           # all values
student.items()            # key-value pairs (for loops!)
"name" in student          # check if key exists → True
```

````python
# ── Example: Class marks analysis ─────────────────────────────────────────────
class_data = {
    "Themba":  [78, 85, 61, 74],
    "Zanele":  [92, 88, 95, 91],
    "Bonga":   [45, 52, 48, 55],
    "Siyanda": [70, 65, 72, 68],
    "Nomvula": [83, 79, 88, 76],
}

# Calculate averages
averages = {name: sum(marks)/len(marks) for name, marks in class_data.items()}

# Sort by average (descending)
ranked = sorted(averages.items(), key=lambda x: x[1], reverse=True)

print("Class Rankings:")
print(f"{'Rank':<6} {'Name':<12} {'Average':>8}  {'Grade'}")
print("-" * 35)
for rank, (name, avg) in enumerate(ranked, 1):
    grade = "A" if avg >= 80 else "B" if avg >= 70 else "C" if avg >= 60 else "D" if avg >= 50 else "F"
    print(f"  {rank}    {name:<12}  {avg:>5.1f}%    {grade}")

all_avgs = list(averages.values())
print(f"\nClass Average: {sum(all_avgs)/len(all_avgs):.1f}%")
print(f"Top student:   {ranked[0][0]} ({ranked[0][1]:.1f}%)")
print(f"Students ≥ 70: {sum(1 for a in all_avgs if a >= 70)}")
````
