## 📖 Lesson 1 — Introduction to Dictionary

### What is a Dictionary?

A **dictionary** is an unordered collection of items where each item is stored as a **key:value pair**. While lists use integer indices to locate data, dictionaries use **meaningful keys**.

Think of it like a real dictionary: you look up a **word** (key) to find its **definition** (value).

### Dictionary Structure

```
  A real dictionary:              A Python dictionary:
  ┌──────────────────────┐        goal_scorers = {
  │  Word  │ Definition  │           'Salah': 13,
  ├──────────────────────┤           'Kane' : 12,
  │  Apple │ A fruit...  │           'Son'  : 12
  │  Bank  │ A place...  │        }
  │  Code  │ Instructions│
  └──────────────────────┘        key ──► 'Salah'
                                  value ──► 13
                                  item ──► 'Salah': 13
                                  separator ──► colon  :
```

### Characteristics of Dictionaries

```
┌────────────────────────────────────────────────────────────────────────────┐
│                     DICTIONARY CHARACTERISTICS                             │
├──────────────────────────────────┬─────────────────────────────────────────┤
│  Feature                         │  Description                            │
├──────────────────────────────────┼─────────────────────────────────────────┤
│  Key:Value pairs                 │  Data stored as key → value             │
│  Colon : separator               │  key:value uses colon                   │
│  One key-value pair = one item   │  {'name': 'Alice'} has 1 item           │
│  Duplicate keys NOT allowed      │  Keys must be unique                    │
│  Duplicate values ARE allowed    │  Multiple keys can share same value     │
│  Heterogeneous keys & values     │  Keys and values can be any type        │
│  Insertion order not preserved   │  Output order may differ from input     │
│  Mutable & dynamic               │  Can add/remove items after creation    │
│  Indexing/slicing NOT applicable │  Access by key, not by position         │
└──────────────────────────────────┴─────────────────────────────────────────┘
```

### 5 Ways to Create a Dictionary

```
┌────┬─────────────────────────┬──────────────────────────────────────────────┐
│ No │  Method                 │  Syntax                                      │
├────┼─────────────────────────┼──────────────────────────────────────────────┤
│  1 │  Empty {}               │  d = {}                                      │
│  2 │  dict() function        │  d = dict()                                  │
│  3 │  {} with colon          │  d = {key1:val1, key2:val2, ...}             │
│  4 │  fromkeys() method      │  d = dict.fromkeys(sequence, default_value)  │
│  5 │  zip() function         │  d = dict(zip(keys_seq, values_seq))         │
└────┴─────────────────────────┴──────────────────────────────────────────────┘
```

````python
# Ex 12.1 - Method 1: Empty dictionary using {}
p = {}
print(p)
print(type(p))    # <class 'dict'>
````

````python
# Ex 12.2 - Method 2: Empty dictionary using dict() function
p = dict()
print(p)
print(type(p))    # <class 'dict'>
````

````python
# Ex 12.3 - Method 3: Dictionary with {} and colon : (most common)
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}

print(goal_scorers)
print(type(goal_scorers))
````

````python
# Ex 12.4 - Method 4: fromkeys() — all keys get the same value
players = {'Salah', 'Kane', 'Son'}
goals = 13

goal_scorers = dict.fromkeys(players, goals)
print(goal_scorers)
print(type(goal_scorers))
````

````python
# Ex 12.5 - fromkeys() without value — defaults to None
players = {'Salah', 'Kane', 'Son'}
goal_scorers = dict.fromkeys(players)   # no value → None
print(goal_scorers)
````

````python
# Ex 12.6 - Method 5: zip() — pair two sequences into a dict
#
#  keys:   ('Salah',  'Kane',  'Son')
#           ↕         ↕        ↕
#  values: [  13,      12,     12 ]
#  result: {'Salah':13, 'Kane':12, 'Son':12}

players = ('Salah', 'Kane', 'Son')
goals   = [13, 12, 12]

goal_scorers = dict(zip(players, goals))
print(goal_scorers)
print(type(goal_scorers))
````

## ✏️ Lesson 2 — Inserting and Updating Dictionary Elements

```
  ┌──────────────────────────────────────────────────────────────────────┐
  │  SAME syntax for INSERTING (new key) and UPDATING (existing key):   │
  │                                                                      │
  │      dictionary_name[key] = value                                    │
  │                                                                      │
  │  • If key does NOT exist → INSERTS new key:value pair               │
  │  • If key ALREADY exists → UPDATES the value for that key           │
  └──────────────────────────────────────────────────────────────────────┘
```

````python
# Ex 12.7 - Inserting a new key:value pair
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}
print('Before:', goal_scorers)

goal_scorers['Vardy'] = 11    # NEW key 'Vardy' — insert
print('After insert Vardy:', goal_scorers)
````

````python
# Ex 12.8 - Updating an existing key's value
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}
print('Before:', goal_scorers)

goal_scorers['Son'] = 14      # EXISTING key 'Son' — update value
print('After Son scores 2 more goals:', goal_scorers)
````

````python
# Ex 12.9 - update() method — insert/update multiple key=value pairs at once
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}

goal_scorers.update(Haaland=11, Vardy=12, Son=14)  # Son updated, Haaland & Vardy added
print(goal_scorers)
````

````python
# Ex 12.10 - update() with another dictionary
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}
new_scorers  = {'Haaland': 11, 'Vardy': 12, 'Son': 14}

goal_scorers.update(new_scorers)  # merges new_scorers into goal_scorers
print(goal_scorers)
````

## 🔍 Lesson 3 — Accessing Dictionary Elements

```
┌──────────────────────────────────────────────────────────────────────────────┐
│                   5 WAYS TO ACCESS DICTIONARY DATA                           │
├──────────────────────────────┬───────────────────────────────────────────────┤
│  Method                      │  Behaviour if key missing                     │
├──────────────────────────────┼───────────────────────────────────────────────┤
│  dict[key]                   │  ❌ Raises KeyError                            │
│  dict.get(key)               │  ✅ Returns None (no error)                   │
│  dict.get(key, default)      │  ✅ Returns default value (no error)          │
│  dict.setdefault(key, val)   │  ✅ Returns value; INSERTS key if missing     │
│  random.choice(list(keys))   │  ✅ Returns a random key                      │
│  dict.items()                │  All key-value pairs as tuples                │
│  dict.keys()                 │  All keys                                     │
│  dict.values()               │  All values                                   │
│  dict.copy()                 │  Shallow copy of the dictionary               │
└──────────────────────────────┴───────────────────────────────────────────────┘
```

````python
# Ex 12.11 - Accessing using key directly
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}

print("Son's Goals:", goal_scorers['Son'])    # 12
print("Salah's Goals:", goal_scorers['Salah'])  # 13
````

````python
# Ex 12.12 - KeyError when key doesn't exist
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}

try:
    print(goal_scorers['Messi'])   # Messi not in dict!
except KeyError as e:
    print('KeyError:', e)
    print('The key', e, 'does not exist in the dictionary!')
````

````python
# Ex 12.13 - get() method — safe access, returns None if key missing
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}

print("Son's Goal:", goal_scorers.get('Son'))      # 12
print("Messi's Goal:", goal_scorers.get('Messi'))  # None — no error!
````

````python
# Ex 12.14 - get() with a custom default value
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}

print("Son's Goal:", goal_scorers.get('Son', 'Not found'))
print(goal_scorers.get('Messi', "Messi isn't a Premier League player."))
````

````python
# Ex 12.15 - Handling KeyError with 'in' operator
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}

if 'Messi' in goal_scorers:
    print("Messi's Goals:", goal_scorers['Messi'])
else:
    print("Messi isn't found here!")
````

````python
# Ex 12.16 - setdefault() — returns value if key exists; INSERTS key with default if missing
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}

print('Son Goals:',   goal_scorers.setdefault('Son'))          # 12 (exists)
print('Alison Goals:', goal_scorers.setdefault('Alison'))      # None + inserts 'Alison':None
print('Dict after setdefault:', goal_scorers)                  # Alison now in dict!
````

````python
# Ex 12.17 - setdefault() with a custom default value
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}

print('Son Goals:',   goal_scorers.setdefault('Son', 0))   # 12 (key exists, default ignored)
print('Can Goals:',   goal_scorers.setdefault('Can', 0))   # 0 + inserts 'Can':0
print('Dict after:', goal_scorers)
````

````python
# Ex 12.18 - random.choice() — pick a random KEY from dictionary
import random
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}

winner = random.choice(list(goal_scorers.keys()))
print('The Winner is', winner)
print('Goals scored:', goal_scorers[winner])
````

````python
# Ex 12.19 - items(), keys(), values() — the three view methods
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}

print('Full dict:', goal_scorers)
print()
print('items()  → key-value pairs:', goal_scorers.items())
print('keys()   → all keys:       ', goal_scorers.keys())
print('values() → all values:     ', goal_scorers.values())
````

````python
# Iterating with items(), keys(), values()
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}

print('--- Iterating with items() ---')
for player, goals in goal_scorers.items():
    print(f'  {player} scored {goals} goals')

print('\n--- Iterating with keys() ---')
for player in goal_scorers.keys():
    print(f'  Player: {player}')

print('\n--- Iterating with values() ---')
for goals in goal_scorers.values():
    print(f'  Goals: {goals}')
````

````python
# Ex 12.20 - copy() — create a copy of the dictionary
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}

fa_cup_scorers = goal_scorers.copy()
fa_cup_scorers['Vardy'] = 5  # modify copy — original unchanged

print('Premier League Scorers:', goal_scorers)
print('FA Cup Scorers (copy):', fa_cup_scorers)
````

## 🗑️ Lesson 4 — Deletion of Dictionary Elements

```
┌──────────────────────────────────────────────────────────────────────────┐
│                  DICTIONARY DELETION METHODS                             │
├──────────────────────┬──────────────────────────┬────────────────────────┤
│  Method              │  What is removed?        │  If key not found?     │
├──────────────────────┼──────────────────────────┼────────────────────────┤
│  del dict[key]       │  Item with that key      │  ❌ KeyError           │
│  dict.pop(key)       │  Item; returns value     │  ❌ KeyError           │
│  dict.pop(key, def)  │  Item; returns default   │  ✅ Returns default    │
│  dict.popitem()      │  Last inserted item      │  ❌ KeyError if empty  │
│  dict.clear()        │  ALL items               │  N/A                   │
└──────────────────────┴──────────────────────────┴────────────────────────┘
```

````python
# Ex 12.21 - del keyword — delete item by key
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}
print('Old Top Goal Scorers:', goal_scorers)

del goal_scorers['Kane']   # Kane is disqualified
print('New Top Goal Scorers:', goal_scorers)
````

````python
# Ex 12.22 - del raises KeyError if key doesn't exist
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}
try:
    del goal_scorers['Messi']   # Messi not in dict!
except KeyError as e:
    print('KeyError:', e)
````

````python
# Ex 12.23 - pop() — removes item by key and RETURNS the value
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}
print('Before:', goal_scorers)

removed_goals = goal_scorers.pop('Kane')
print('Kane scored', removed_goals, 'goals (removed)')
print('After pop(Kane):', goal_scorers)

# pop() with a default to avoid KeyError
default_result = goal_scorers.pop('Messi', 0)
print('Messi goals (default 0):', default_result)
````

````python
# Ex 12.25 - popitem() — removes and returns the LAST inserted item as a tuple
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}
print('Before:', goal_scorers)

last_item = goal_scorers.popitem()  # removes last item
print('Removed item (key, value):', last_item)
print('After popitem():', goal_scorers)
````

````python
# Ex 12.26 - popitem() raises KeyError on empty dictionary
p = {}
try:
    p.popitem()
except KeyError as e:
    print('KeyError:', e)
````

````python
# Ex 12.27 - clear() — removes ALL items, keeps empty dict
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}
print('Before clear():', goal_scorers)

goal_scorers.clear()
print('After clear():', goal_scorers)  # {}
````

````python
# Ex 12.28 - len() — number of key:value pairs in dictionary
goal_scorers = {'Salah': 13, 'Kane': 12, 'Son': 12}
print('There are', len(goal_scorers), 'top goal scorers.')
````

## ⚡ Lesson 5 — Dictionary Comprehension

Just as list comprehension creates lists concisely, **dictionary comprehension** creates dictionaries in a single line.

```
  Syntax:
  {key_expr: value_expr for element in iterable if condition}

  Mapping to a regular loop:
  ┌──────────────────────────────────────────────────────────────────┐
  │  Regular:                          Comprehension:                │
  │  d = {}                            d = {i: i**2                 │
  │  for i in numbers:                      for i in numbers}        │
  │      d[i] = i**2                                                 │
  └──────────────────────────────────────────────────────────────────┘
```

````python
# Ex 12.29 - Dictionary comprehension — square each number
numbers = [1, 2, 3, 4, 5]

squared = {i: i**2 for i in numbers}
print(squared)
print(type(squared))
````

````python
# Dictionary comprehension with condition — only even numbers
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

even_squares = {i: i**2 for i in numbers if i % 2 == 0}
print('Even number squares:', even_squares)
````

````python
# Dictionary comprehension — word lengths
words = ['hello', 'python', 'world', 'code']
word_lengths = {word: len(word) for word in words}
print('Word lengths:', word_lengths)
````

````python
# Dictionary comprehension — invert a dictionary (swap keys and values)
original = {'Salah': 13, 'Kane': 12, 'Son': 11}
inverted = {v: k for k, v in original.items()}
print('Original:', original)
print('Inverted:', inverted)
````

## 💡 Practical Examples — Real-World Dictionary Uses

### Example 1: Student Grade Book

````python
# Student grade book using nested dictionary
grade_book = {
    'Alice': {'Maths': 90, 'English': 85, 'Science': 92},
    'Bob':   {'Maths': 72, 'English': 68, 'Science': 75},
    'Carol': {'Maths': 95, 'English': 88, 'Science': 91},
}

print('=== Grade Report ===')
for student, marks in grade_book.items():
    avg = sum(marks.values()) / len(marks)
    print(f'{student}: Maths={marks["Maths"]}, English={marks["English"]}, Science={marks["Science"]}, Average={avg:.1f}')

# Find top student
top = max(grade_book, key=lambda s: sum(grade_book[s].values()))
print(f'\n🏆 Top student: {top}')
````

### Example 2: Word Frequency Counter

````python
# Count word frequency in a sentence
sentence = 'the cat sat on the mat and the cat sat'
words = sentence.split()

freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1  # get returns 0 if key missing

print('Word frequencies:')
for word, count in sorted(freq.items(), key=lambda x: x[1], reverse=True):
    print(f'  "{word}" → {count} times')
````

## 📊 Dictionary vs List vs Set vs Tuple

```
┌──────────────┬──────────────┬──────────────┬──────────────┬──────────────────────┐
│  Feature     │  List        │  Tuple       │  Set         │  Dictionary          │
├──────────────┼──────────────┼──────────────┼──────────────┼──────────────────────┤
│  Syntax      │  [1, 2, 3]   │  (1, 2, 3)   │  {1, 2, 3}   │  {'a':1, 'b':2}      │
│  Ordered?    │  Yes ✅       │  Yes ✅       │  No ❌       │  No ❌               │
│  Mutable?    │  Yes ✅       │  No ❌        │  Yes ✅      │  Yes ✅              │
│  Duplicates? │  Yes ✅       │  Yes ✅       │  No ❌       │  Keys: No / Values: Yes │
│  Index?      │  Yes ✅       │  Yes ✅       │  No ❌       │  By key only         │
│  Use when    │  Ordered     │  Fixed data  │  Unique only │  Key-value pairs     │
│              │  mutable data│              │              │                      │
└──────────────┴──────────────┴──────────────┴──────────────┴──────────────────────┘
```

  <h3 style='color: #6c757d; margin-top: 0;'>📋 Key Takeaways — Chapter 12: Dictionary</h3>
  <table style='width:100%; border-collapse: collapse;'>
    <tr style='background:#e9ecef;'><th style='padding:8px; text-align:left;'>Concept</th><th style='padding:8px; text-align:left;'>Key Point</th></tr>
    <tr><td style='padding:8px;'><strong>What is a dictionary?</strong></td><td style='padding:8px;'>An unordered, mutable collection of <strong>key:value pairs</strong> — access data by meaningful keys, not index positions</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>Key rules</strong></td><td style='padding:8px;'>Keys must be <strong>unique</strong> and immutable (strings, numbers, tuples); duplicate keys silently overwrite the earlier value</td></tr>
    <tr><td style='padding:8px;'><strong>Creating dicts</strong></td><td style='padding:8px;'>5 ways: <code>{}</code>, <code>dict()</code>, <code>{k:v, ...}</code>, <code>dict.fromkeys(seq, val)</code>, <code>dict(zip(keys, vals))</code></td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>Inserting / Updating</strong></td><td style='padding:8px;'><code>d[key] = value</code> inserts if key is new, updates if it already exists. <code>d.update(...)</code> handles multiple pairs at once</td></tr>
    <tr><td style='padding:8px;'><strong>Safe access</strong></td><td style='padding:8px;'><code>d[key]</code> raises <code>KeyError</code> if missing. Prefer <code>d.get(key, default)</code> for safe reads that never crash</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>setdefault()</strong></td><td style='padding:8px;'>Returns the value for a key AND <strong>inserts</strong> the key with the default value if it does not exist — useful for counting patterns</td></tr>
    <tr><td style='padding:8px;'><strong>random.choice()</strong></td><td style='padding:8px;'>Pick a random key with <code>random.choice(list(d.keys()))</code>; then retrieve its value with <code>d[key]</code></td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>Iteration methods</strong></td><td style='padding:8px;'><code>d.items()</code> → (key, val) pairs; <code>d.keys()</code> → all keys; <code>d.values()</code> → all values — use in for-loops to traverse the dict</td></tr>
    <tr><td style='padding:8px;'><strong>Deletion</strong></td><td style='padding:8px;'><code>del d[key]</code> or <code>d.pop(key)</code> remove one item; <code>d.popitem()</code> removes the last inserted item; <code>d.clear()</code> empties the whole dict</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>Dict comprehension</strong></td><td style='padding:8px;'><code>{k_expr: v_expr for elem in iterable if condition}</code> — builds a dict in one line, just like list comprehension builds a list</td></tr>
    <tr><td style='padding:8px;'><strong>Nested dicts</strong></td><td style='padding:8px;'>Values can themselves be dicts — useful for grade books, JSON-style records, and hierarchical data</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>Dict vs List</strong></td><td style='padding:8px;'>Use a <strong>list</strong> when order and position matter; use a <strong>dict</strong> when you need to look up data by a meaningful label/name</td></tr>
    <tr><td style='padding:8px;'><strong>copy()</strong></td><td style='padding:8px;'><code>d.copy()</code> creates a shallow copy — modifying the copy does not affect the original (for nested dicts use <code>copy.deepcopy()</code>)</td></tr>
