## 💻 Lesson 1 — Introduction to Programming

> *"Computers are dumb. Computers are only smart because we program them to be."*

### What is Programming?

**Programming** is the process of preparing an instructional program for a device to do some task without making mistakes.

A computer on its own **cannot think, decide, or act**. It only does exactly what it is told, precisely and literally. Programming is the way we communicate those instructions.

```
  WITHOUT PROGRAMMING          WITH PROGRAMMING
  ┌─────────────┐              ┌─────────────┐       ┌──────────────┐
  │  Computer   │              │ Programmer  │──────►│   Computer   │
  │   (does     │              │  (writes    │       │  (executes   │
  │   nothing)  │              │   code)     │       │   program)   │
  └─────────────┘              └─────────────┘       └──────────────┘
```

There are **two types of languages** for communication between a computer and a programmer:

```
┌───────────────────────────────────────────────────────────────────────────┐
│  TYPE 1 — Machine Language                                                │
│  • Binary code: a series of 0s and 1s                                    │
│  • Computers understand it directly ✅                                     │
│  • Humans find it VERY difficult to read ❌                               │
│  • Example: 01001000 01100101 01101100 01101100 01101111                  │
├───────────────────────────────────────────────────────────────────────────┤
│  TYPE 2 — Programming Language                                            │
│  • Source code: human-readable text                                       │
│  • Humans understand it ✅                                                 │
│  • Computers CANNOT execute it directly ❌                                │
│  • Example: print("Hello World")                                         │
└───────────────────────────────────────────────────────────────────────────┘

  ▸ PROBLEM: We write in programming language,
             but computers only understand machine language.
  ▸ SOLUTION: We need a TRANSLATOR — this is what compilers and interpreters do.
```

### What is a Programming Language?

A **programming language** is a formal language which includes a set of instructions that helps a programmer to communicate or interact with a computer.

```
┌──────────────────────┬──────────────────────────┬────────────────────────────┐
│  Language            │  Who understands it       │  Example                  │
├──────────────────────┼──────────────────────────┼────────────────────────────┤
│  Machine Language    │  Computer ✅              │  01001000 01100101...      │
│  Programming Language│  Human ✅                 │  print("Hello World")     │
└──────────────────────┴──────────────────────────┴────────────────────────────┘
```

Programming language is classified into **two categories** based on the level (degree) of abstraction and gap between source code and machine code:

1. **Source code** — Any collection of code written using a human-readable programming language, usually as plain text.
2. **Machine code** — A computer program written in machine language instructions that can be executed directly by a computer's control unit.

### High-Level vs Low-Level Languages

```
┌──────────────────────────────────────────────────────────────────────────────┐
│                  PROGRAMMING LANGUAGE LEVELS                                 │
├──────────────────────────────┬───────────────────────────────────────────────┤
│  HIGH-LEVEL LANGUAGES        │  LOW-LEVEL LANGUAGES                         │
├──────────────────────────────┼───────────────────────────────────────────────┤
│  Programmer-friendly         │  Machine-friendly                             │
│  Humans can easily read it   │  Humans can't understand it (not readable)    │
│  Easy to maintain and debug  │  Very difficult to write and debug            │
│  Further from hardware       │  Closer to hardware                           │
│  Examples: Python, Java, C   │  Examples: Assembly, Machine language         │
└──────────────────────────────┴───────────────────────────────────────────────┘

  PS. Computers only understand machine code (0s and 1s).
      Python is a HIGH-LEVEL language — it reads almost like plain English.
```

## ⚙️ Lesson 2 — Compiler and Interpreter

All programming languages must be converted to machine-understandable language for execution. This conversion is handled by **compilers** and **interpreters**.

### What is a Compiler?

A **compiler** is a computer program that translates source code written in one programming language (the source language) into another language (the target language) to create an executable program.

A compiler takes the **entire source code at once**, converts it into machine-executable code, and the processor then runs that executable file.

```
  COMPILER WORKFLOW:

  ┌──────────────┐       ┌──────────────┐       ┌────────────────┐       ┌──────────┐
  │ Source Code  │──────►│   COMPILER   │──────►│ Executable File│──────►│ Run once │
  │  (all code)  │       │ (ALL at once)│       │  (machine code)│       │  (fast!) │
  └──────────────┘       └──────────────┘       └────────────────┘       └──────────┘

  ✅ Advantage:  Faster runtime — code is already translated before running
  ❌ Disadvantage: If ONE error exists, NOTHING runs until ALL errors are fixed
  📌 Examples: Java, C++
```

### What is an Interpreter?

An **interpreter** is a computer program that directly executes instructions written in a programming language, **without** requiring them to be previously compiled into machine language. An interpreter runs the code **line by line**, translating and executing one instruction at a time.

```
  INTERPRETER WORKFLOW:

  ┌──────────────┐
  │ Source Code  │
  └──────┬───────┘
         │
         ▼  (one line at a time)
  ┌──────────────┐
  │ Line 1       │──── translate ──── run
  ├──────────────┤
  │ Line 2       │──── translate ──── run
  ├──────────────┤
  │ Line 3 ❌    │──── ERROR found → STOP and show error message
  ├──────────────┤
  │ Line 4       │  (never reached — interpreter stopped at line 3)
  └──────────────┘

  ✅ Advantage:  Errors appear immediately on the exact line — easy to debug
  ❌ Disadvantage: Slower at runtime (translates as it runs)
  📌 Examples: Python, PHP, Ruby
```

### Compiler vs Interpreter — Comparison

```
┌──────────────────────────┬──────────────────────────────┬──────────────────────────────┐
│  Feature                 │  Compiler                    │  Interpreter                 │
├──────────────────────────┼──────────────────────────────┼──────────────────────────────┤
│  Translation method      │  Translates ALL code at once │  Translates line by line      │
│  Speed at runtime        │  Fast (pre-translated)       │  Slower (translates as it runs│
│  Error handling          │  Shows all errors before run │  Stops at first error found   │
│  Output                  │  Creates an executable file  │  No separate file created     │
│  Debugging               │  Harder (batch errors)       │  Easier (error shown inline)  │
│  Examples                │  Java, C++                   │  Python, PHP, Ruby            │
└──────────────────────────┴──────────────────────────────┴──────────────────────────────┘
```

### 🎬 The Italian Movie Analogy

Imagine watching an Italian movie but you only understand English. You have two options:

```
  ┌──────────────────────────────────────────────────────────────────────────┐
  │  OPTION 1: English DUBBED version                                        │
  │  → All Italian has been converted to English BEFORE you watch            │
  │  → You watch the whole thing in English — all translated upfront         │
  │  → This is like a COMPILER (translates everything first)                 │
  ├──────────────────────────────────────────────────────────────────────────┤
  │  OPTION 2: Italian movie with English SUBTITLES                          │
  │  → Italian plays SCENE BY SCENE, subtitles appear in real time           │
  │  → Each scene is translated as it plays                                  │
  │  → This is like an INTERPRETER (translates one line at a time)           │
  └──────────────────────────────────────────────────────────────────────────┘

  Python works like SUBTITLES — it reads and runs your code one line at a time.
```

## 🖥️ Lesson 3 — How to Write and Run Code

### How to Write Code?

To write and execute a Python program on our computer we need to install two basic application softwares:
1. **Editor (IDE)** — used to write our source code
2. **Compiler / Interpreter** — used to translate and execute our source code

There are many editors and compilers available for Python. We use **Jupyter Notebook** as our IDE.

### What is an IDE?

**IDE (Integrated Development Environment)** is a place (graphical interface) to **write, run, debug (check errors), and convert** your source code to machine code.

```
┌─────────────────────────────────────────────────────────────────────────────┐

```

### How to Run (Execute) Code?

You can't simply type anything into an IDE, convert it, and expect the computer to understand it. Each programming language has its own set of rules (grammar) you must follow, called **Syntax**.

### What is Syntax?

**Syntax** is the set of rules you must follow to run your program correctly. Syntax for each programming language is **unique**.

<div style='border: 2px solid #fd7e14; border-radius: 12px; padding: 15px; background: #fff8f0; margin: 15px 0;'>
  <strong style='color: #fd7e14;'>⚠️ Important:</strong> Breaking programming rules will result in an <strong>error</strong>. Your program will NOT run until you fix the syntax error.
</div>

```
  ENGLISH EXAMPLE — grammar rules:
  ❌  Mens clothes are cheap.      ← missing apostrophe — grammar error
  ✅  Men's clothes are cheap.     ← correct

  PYTHON EXAMPLE — syntax rules:
  ❌  print("Hello World"          ← missing closing bracket — SyntaxError
  ✅  print("Hello World")         ← correct

  ❌  Print("Hello World")         ← capital P — NameError (Python is case-sensitive)
  ✅  print("Hello World")         ← correct (lowercase p)
```

```
  HOW CODE FLOWS THROUGH AN IDE:

  You type code ──► IDE runs Interpreter/Compiler ──► Machine Code ──► Output

  If there's a Syntax Error:
  You type code ──► IDE detects error ──► Error Message ──► NO output
                                           (fix and try again)
```
