## 🍰 Lesson 1 — Introduction to Exception Handling

### The Baking Analogy

Imagine you are following a recipe to bake a cake. As you go step by step, you may encounter unexpected situations like:
- Running out of a specific ingredient
- Accidentally burning the cake
- Your oven not working

**Exception handling** is like having a backup plan or alternative actions to take when these unexpected situations happen.

In programming, similar unexpected situations can occur:
- Trying to open a file that doesn't exist
- Dividing a number by zero
- Receiving invalid input from a user

### What is an Error?

An **error** is an unexpected or exceptional condition that occurs during the execution of a program. Errors can arise due to various reasons:

```
  Causes of Errors
  ┌────────────────────────────────────────────────────────────┐
  │  • Invalid input from a user                               │
  │  • Resource unavailability (file not found, no network)    │
  │  • Programming mistakes (bugs)                             │
  │  • External factors beyond the program's control           │
  └────────────────────────────────────────────────────────────┘

  When an error occurs → the normal flow is disrupted
                       → undesired or incorrect behaviour follows
```

### Types of Errors

```
┌──────────────────────────────────────────────────────────────────────────────┐
│                        TYPES OF ERRORS                                       │
├──────────────────┬──────────────────────────────────────────────────────────┤
│  Type            │  Description                                              │
├──────────────────┼──────────────────────────────────────────────────────────┤
│  Syntax Error    │  Occurs BEFORE execution — invalid syntax/grammar         │
│                  │  e.g. missing colon, wrong indentation, misspelled keyword │
│                  │  The PROGRAMMER must fix this — no exception handling      │
├──────────────────┼──────────────────────────────────────────────────────────┤
│  Runtime Error   │  Occurs DURING execution — program crashes mid-run        │
│                  │  e.g. dividing by zero, accessing an invalid index         │
│                  │  These are called EXCEPTIONS — can be handled             │
└──────────────────┴──────────────────────────────────────────────────────────┘

  ⚠️  Exception Handling applies ONLY to Runtime Errors, NOT Syntax Errors.
```

````python
# Ex 14.1 - Syntax Error (missing colon in if statement)
# This code CANNOT be run — it will show SyntaxError immediately
# Uncomment it to see the error:

# speed = 113
# if speed > 113          # <-- missing colon!
#     print("Driving beyond the speed limit!")

# Error message:
#   File "<ipython-input>", line 2
#       if speed > 113
#                    ^
#   SyntaxError: expected ':'

print("Fix the syntax first — only then can the program execute.")
````

````python
# Ex 14.2 - Runtime Error (ZeroDivisionError during execution)
distance = 120
time = 0              # <-- dividing by zero causes RuntimeError
fuel_consumed = 10

try:
    speed = distance / time
except ZeroDivisionError as e:
    print(f"ZeroDivisionError caught: {e}")
    print("The program was executing fine until it reached division by zero.")
````

### Types of Program Execution Flow

```
  PROGRAM EXECUTION FLOW
  ═══════════════════════════════════════════════════════════════════════

  NORMAL FLOW:
  ┌──────────┐     ┌──────────┐     ┌──────────┐     ┌──────────┐
  │Statement │ --> │Statement │ --> │Statement │ --> │  END ✓   │
  │    1     │     │    2     │     │    3     │     │          │
  └──────────┘     └──────────┘     └──────────┘     └──────────┘
  All statements execute successfully → expected output produced.

  ABNORMAL FLOW:
  ┌──────────┐     ┌──────────┐     ┌──────────┐
  │Statement │ --> │Statement │ --> │ CRASH! ✗ │
  │    1     │     │  2 ERROR │     │ Remaining│
  └──────────┘     └──────────┘     │ statements│
                                    │ skipped  │
                                    └──────────┘
  A RuntimeError at statement 2 → program terminates immediately
  → all statements after the error are NEVER executed.
  ═══════════════════════════════════════════════════════════════════════
```

````python
# Ex 14.3 - NORMAL FLOW: all statements execute successfully
distance = 120
time = 2
fuel_consumed = 10

speed   = distance / time
print(f"Speed   = {speed} mph")       # executes
mileage = distance / fuel_consumed
print(f"Mileage = {mileage} MPG")    # executes
````

````python
# Ex 14.4 - ABNORMAL FLOW: error at line 5 kills the rest of the program
# To demonstrate, we catch the error so the notebook doesn't stop:
distance = 120
time = 0              # PROBLEM: this will cause ZeroDivisionError
fuel_consumed = 10

try:
    speed = distance / time     # ERROR here → jumps out
    print(f"Speed   = {speed} mph")    # NEVER reached
    mileage = distance / fuel_consumed # NEVER reached
    print(f"Mileage = {mileage} MPG") # NEVER reached
except ZeroDivisionError as e:
    print(f"Program stopped abnormally. Error: {e}")
    print("Lines after the error were NOT executed.")
````

### What is an Exception?

> An **exception** is an unwanted or unexpected event which disturbs the normal flow of the program.

### Error vs Exception

```
┌──────────────────────────────────┬────────────────────────────────────────┐
│  Error                           │  Exception                             │
├──────────────────────────────────┼────────────────────────────────────────┤
│  Major problems — program        │  Specific problems that CAN be fixed   │
│  CANNOT continue at all          │  Program CAN catch and handle them     │
│  Caused by bugs in code or       │  Interrupt normal program flow         │
│  system-level issues             │  but execution can resume after        │
│  Cannot be handled at runtime    │  handling                              │
└──────────────────────────────────┴────────────────────────────────────────┘

  Remember: Exception Handling ≠ Fixing the exception.
  It means defining an alternative path so the program doesn't terminate abnormally.
```

### How Python Handles Exceptions Internally

```
  Python Virtual Machine (PVM) Process:

  Exception occurs
       │
       ▼
  PVM creates an Exception object of the matching class
       │
       ▼
  Is handling code available?  ─── YES ──► except block runs (program survives)
       │
       NO
       │
       ▼
  Python terminates program abnormally
  Prints exception name + message to console
  Remaining program code is NOT executed
```

### Python Exception Hierarchy

```
  BaseException  ← root of ALL exceptions
  │
  ├── SystemExit
  ├── KeyboardInterrupt
  ├── GeneratorExit
  └── Exception  ← focus here — parent of all catchable exceptions
       │
       ├── ArithmeticError
       │    └── ZeroDivisionError
       │    └── OverflowError
       ├── LookupError
       │    └── IndexError
       │    └── KeyError
       ├── ValueError
       ├── TypeError
       ├── NameError
       ├── AttributeError
       ├── FileNotFoundError
       ├── IOError
       ├── ImportError
       ├── MemoryError
       ├── RecursionError
       └── StopIteration

  Note: Every exception in Python is a CLASS.
        All exception classes are child classes of BaseException.
```

### When to CATCH vs When to Let it CRASH

```
  CATCH the exception when:                    LET IT CRASH when:
  ─────────────────────────────────────────    ──────────────────────────────────────────
  ✓ User input is the cause — they can retry   ✗ It's a programming bug (fix the code)
  ✓ External resource (file, network, API)     ✗ Catching it would hide the real problem
    may not always be available                ✗ You have no useful recovery action
  ✓ You have a clear alternative path          ✗ It would leave data in a broken state
  ✓ You can give the user a helpful message    ✗ It's a SyntaxError (can't be caught)

  GOLDEN RULE:
  ┌──────────────────────────────────────────────────────────────────────────────────┐
  │  Only catch exceptions you know HOW to handle.                                  │
  │  A bare  except: pass  that silently swallows errors is WORSE than a crash —    │
  │  it hides bugs and makes them nearly impossible to debug later.                 │
  └──────────────────────────────────────────────────────────────────────────────────┘
```

## 🛡️ Lesson 2 — Exception Handling Using try-except

### Syntax

```python
try:
    # code that may raise an exception
except ExceptionName:
    # code to handle the exception
```

### How it Works

```
  ┌─────────────────────────────────────────────────────────────────────────┐
  │                     try-except FLOW                                     │
  │                                                                         │
  │   enter try block                                                       │
  │        │                                                                │
  │        ▼                                                                │
  │   exception raised?  ─── NO ──► execute remaining try code             │
  │        │                        skip except block(s)                   │
  │       YES                       continue after try-except               │
  │        │                                                                │
  │        ▼                                                                │
  │   jump immediately to matching except block                             │
  │   (remaining try code is SKIPPED)                                       │
  │        │                                                                │
  │        ▼                                                                │
  │   run except block code                                                 │
  │        │                                                                │
  │        ▼                                                                │
  │   continue after try-except (program survives!)                        │
  └─────────────────────────────────────────────────────────────────────────┘
```

### What to Put in Each Block — Design Principles

```
  TRY BLOCK:
  ─────────────────────────────────────────────────────────────────────────
  • Only the code that MIGHT raise an exception
  • Keep it as SHORT as possible — one risky operation per try block
  • Do NOT put all your code in one giant try — you lose precision
    (you won't know which line actually failed)
  • Think: "What specific line here could raise an exception?"

  Bad (too broad):                      Good (targeted):
  try:                                  name = input("Name: ")    # safe
      name  = input("Name: ")           age  = input("Age: ")     # safe
      age   = input("Age: ")            try:
      age   = int(age)                      age = int(age)         # risky
      print(f"Hi {name}, {age}")        except ValueError:
  except:                                   print("Age must be a number.")
      print("Something went wrong")     print(f"Hi {name}, {age}")

  ─────────────────────────────────────────────────────────────────────────
  EXCEPT BLOCK:
  ─────────────────────────────────────────────────────────────────────────
  • Catch the MOST SPECIFIC exception type you expect — not bare except
  • Give a MESSAGE the user can actually act on ("Enter a number" not "Error")
  • Do NOT catch exceptions you have no way to handle — let them bubble up
  • Use 'as e' to also print Python's own error message for debugging

  ─────────────────────────────────────────────────────────────────────────
  ELSE BLOCK (optional — runs only when NO exception occurred):
  ─────────────────────────────────────────────────────────────────────────
  • Code that DEPENDS on the try block succeeding
  • Keeps success path clearly separated from error path
  • Exceptions raised in else are NOT caught by the above except blocks

  ─────────────────────────────────────────────────────────────────────────
  FINALLY BLOCK (optional — always runs, no matter what):
  ─────────────────────────────────────────────────────────────────────────
  • Cleanup: close open files, database connections, reset state flags
  • Logging: record that an operation was attempted
  • Guaranteed to run even if the exception is NOT handled
```

````python
# Ex 14.5 - Basic try-except: exception occurs → caught → program continues
distance = 120
time = 0
fuel_consumed = 10

try:
    speed = distance / time          # ZeroDivisionError here
    print(f"Speed = {speed} mph")   # SKIPPED
except ZeroDivisionError:
    print("Exception caught! Cannot divide by zero.")

# Execution continues NORMALLY after the try-except block
mileage = distance / fuel_consumed
print(f"Mileage = {mileage} MPG")  # This line DOES run!
````

````python
# Ex 14.6 - No exception in try block → except block is SKIPPED
distance = 120
time = 2           # time is valid — no error
fuel_consumed = 10

try:
    speed = distance / time
    print(f"Speed = {speed} mph")   # executes normally
except ZeroDivisionError:
    print("Exception caught!")      # SKIPPED — no exception occurred

mileage = distance / fuel_consumed
print(f"Mileage = {mileage} MPG")
````

````python
# Ex 14.7 - Exception BEFORE the try block: the try-except cannot help!
# Only code INSIDE the try block is protected.
distance = 120
time = 0
fuel_consumed = 0

try:
    # This error would be handled IF it were in the try block:
    mileage_demo = distance / 10     # safe — not the real problem
    print(f"Safe demo mileage = {mileage_demo} MPG")
except ZeroDivisionError:
    print("Handled inside try block.")

# BUT if mileage = distance / fuel_consumed  was BEFORE the try, it would crash:
print("\nDemonstrating error before try block:")
try:
    risky_before_try = distance / fuel_consumed  # in try — so it IS caught here
    print(f"Mileage = {risky_before_try}")
except ZeroDivisionError as e:
    print(f"Inside try: caught → {e}")

print("Fuel Level = Low")  # this runs because we wrapped everything
````

### Printing Exception Information

Instead of printing a custom message, you can print the actual exception information by using `as e`:

```python
except ExceptionName as e:
    print(e)   # prints the error message from Python
```

````python
# Ex 14.8 - Printing exception information using 'as e'
distance = 120
time = 0
fuel_consumed = 10

try:
    speed = distance / time
    print(f"Speed = {speed} mph")
except ZeroDivisionError as e:
    print("Exception Information:", e)   # prints: division by zero

mileage = distance / fuel_consumed
print(f"Mileage = {mileage} MPG")
````

### Multiple except Blocks

A single `try` block can have **multiple `except` blocks** — one for each exception type that might occur.

```python
try:
    # code that may raise different exceptions
except ZeroDivisionError:
    # handle ZeroDivisionError
except ValueError:
    # handle ValueError
except TypeError:
    # handle TypeError
```

```
  MULTIPLE EXCEPT BLOCKS FLOW:

  try
   │
   ├── ZeroDivisionError?  → except ZeroDivisionError: (specific message)
   │
   ├── ValueError?         → except ValueError: (specific message)
   │
   └── No exception?       → skip ALL except blocks
```

````python
# Ex 14.9 - try with multiple except blocks
def calculate_speed(distance, time):
    try:
        speed = int(distance) / int(time)
        print(f"Speed = {speed} mph")
    except ZeroDivisionError:
        print(f"Can't divide {distance} by {time}.")
    except ValueError:
        print("Please provide integer values only!")

calculate_speed(120, 2)       # no exception
calculate_speed(120, 0)       # ZeroDivisionError
calculate_speed(120, 'two')   # ValueError
````

### Single except Block Handling Multiple Exceptions

You can also handle **multiple exception types in one `except` block** by grouping them in parentheses (a tuple):

```python
except (ZeroDivisionError, ValueError, TypeError) as e:
    print("Error:", e)
```

```
  Trade-off:
  ┌──────────────────────────────────┬──────────────────────────────────────┐
  │  Multiple except blocks          │  Single except with tuple            │
  ├──────────────────────────────────┼──────────────────────────────────────┤
  │  More code                       │  Less code                           │
  │  Custom message per exception    │  Same message for all exceptions     │
  │  Better for diverse errors       │  Better for similar errors           │
  └──────────────────────────────────┴──────────────────────────────────────┘
```

### Which Handler Style Should I Choose?

```
┌──────────────────────────────────┬─────────────────────────────────────────────────────┐
│  Style                           │  When to Use It                                     │
├──────────────────────────────────┼─────────────────────────────────────────────────────┤
│  Multiple separate except blocks │  • You want a DIFFERENT message/action per error    │
│                                  │  • The errors mean different things to the user     │
│  except ZeroDivisionError:       │  • Most explicit, most readable approach            │
│      print("Cannot divide by 0") │  • Best practice for professional code              │
│  except ValueError:              │                                                     │
│      print("Number required")    │                                                     │
├──────────────────────────────────┼─────────────────────────────────────────────────────┤
│  Single except with a tuple      │  • The errors are similar — same recovery action    │
│                                  │  • You want shorter, less repetitive code           │
│  except (ZeroDivisionError,      │  • Works well when both produce the same message    │
│           ValueError) as e:      │                                                     │
│      print("Input error:", e)    │                                                     │
├──────────────────────────────────┼─────────────────────────────────────────────────────┤
│  Default bare except:            │  • As a LAST RESORT catch-all after specific blocks │
│                                  │  • When code could raise an unpredictable exception │
│  except:                         │  ⚠️ MUST always be the LAST except block            │
│      print("Unexpected error")   │  ⚠️ Use sparingly — it hides the actual error type  │
├──────────────────────────────────┼─────────────────────────────────────────────────────┤
│  Catch a parent class            │  • Catch a whole FAMILY of related exceptions       │
│                                  │  • ArithmeticError catches ZeroDivisionError,       │
│  except ArithmeticError:         │    OverflowError, FloatingPointError automatically  │
│      ...                         │  • Exception catches ALL non-system exceptions      │
└──────────────────────────────────┴─────────────────────────────────────────────────────┘

  ORDER RULE — always most specific first, least specific last:
  ┌────────────────────────────────────────────────────────────────┐
  │  except ZeroDivisionError:   ← most specific (first)          │
  │  except ValueError:          ← specific                       │
  │  except ArithmeticError:     ← parent catches its children    │
  │  except Exception:           ← catches almost everything      │
  │  except:                     ← catch-all (MUST be last)       │
  └────────────────────────────────────────────────────────────────┘
```

## 🧹 Lesson 3 — The finally Block

### Why is finally Needed?

In real projects, after completing tasks you must always clean up — close database connections, release file handles, reset state. But:

```
  Why NOT put cleanup in the try block?
  → Not guaranteed to execute: if an exception is raised at line 2,
    the cleanup code at line 5 is SKIPPED.

  Why NOT put cleanup in the except block?
  → Not guaranteed to execute: if there is NO exception,
    the except block never runs → cleanup is SKIPPED.

  Solution: FINALLY BLOCK
  The finally block ALWAYS executes:
  ┌───────────────────────────────────────────────────────────┐
  │  Case 1: No exception        → try + finally execute      │
  │  Case 2: Exception handled   → try + except + finally     │
  │  Case 3: Exception NOT handled → try + finally execute    │
  │          (program still crashes, but cleanup ran first)   │
  └───────────────────────────────────────────────────────────┘
```

### Syntax

```python
try:
    # risky code
except ExceptionName:
    # handler
finally:
    # always runs — cleanup code here
```

### Real-World Use Cases for `finally`

```
  THE FUNDAMENTAL PROMISE:
  "No matter what happens — success, caught exception, or uncaught exception —
   the finally block WILL execute before control leaves this try structure."

  ─────────────────────────────────────────────────────────────────────────
  1. FILE HANDLING — always close the file handle to prevent leaks
     try:
         f = open("report.txt", "r")
         data = f.read()
     except FileNotFoundError:
         print("File not found!")
     finally:
         f.close()   # always runs — prevents file handle leak

  2. DATABASE CONNECTION — always release the connection
     try:
         conn = db.connect()
         result = conn.execute(query)
     except DatabaseError as e:
         print("Query failed:", e)
     finally:
         conn.close()   # always releases — even if query crashed

  3. STATE RESET — always restore variables to a safe state
     Power = True
     try:
         run_machine()
     except MachineError as e:
         print("Machine error:", e)
     finally:
         Power = False   # always powers off — safe shutdown guaranteed

  4. LOGGING — always record that an operation was attempted
     try:
         process_payment(amount)
     except PaymentError:
         print("Payment failed.")
     finally:
         log_attempt(amount, timestamp=now())   # always logged

  ─────────────────────────────────────────────────────────────────────────
  WITHOUT FINALLY — the danger of skipped cleanup:

  try:
      f = open("file.txt")
      data = process(f)    ← exception here!
      f.close()            ← NEVER REACHED → file handle leaks into memory
  except:
      print("Error")       ← f.close() not here either → still leaks
  ─────────────────────────────────────────────────────────────────────────
```

````python
# Ex 14.13 - Case 1: No exception → try runs, then finally runs
distance = 120
time = 2          # valid time — no exception
fuel_consumed = 10
Power = True

try:
    speed = distance / time
    print(f"Speed = {speed} mph")
except ZeroDivisionError as e:
    print("Exception:", e)
finally:
    # cleanup: reset all state regardless of success or failure
    Power = False
    distance = 0
    time = 0
    print(f"Power={Power}, Distance={distance} miles, Time={time} hrs")

print("\nCase 1: No exception — try block ran, then finally ran.")
````

````python
# Ex 14.14 - Case 2: Exception raised AND handled → all 3 blocks run
distance = 120
time = 0          # exception will be raised
fuel_consumed = 10
Power = True

try:
    speed = distance / time      # ZeroDivisionError here
    print(f"Speed = {speed} mph")  # SKIPPED
except ZeroDivisionError as e:
    print("Exception:", e)         # runs — exception handled
finally:
    Power = False
    distance = 0
    time = 0
    print(f"Power={Power}, Distance={distance}, Time={time}")  # STILL runs!

print("\nCase 2: Exception was handled — all 3 blocks (try, except, finally) ran.")
````

````python
# Ex 14.15 - Case 3: Exception raised but NOT handled → finally still runs
# (but program eventually crashes)
distance = 120
time = 0          # ZeroDivisionError
Power = True

try:
    speed = distance / time      # ZeroDivisionError
    print(f"Speed = {speed} mph")
except ValueError as e:          # WRONG handler — ZeroDivision is not ValueError
    print("ValueError:", e)      # will NOT run
finally:
    Power = False
    distance = 0
    time = 0
    print(f"Finally ran! Power={Power}, Distance={distance}, Time={time}")

# The outer try is just to show the notebook — in real code, program crashes:
print("\nCase 3: Exception NOT handled — finally ran anyway, then crash.")
````

### Nested try-except-finally Blocks

Python allows `try-except-finally` blocks to be **nested inside other `try` blocks**.

```python
try:
    # outer try
    try:
        # inner try
    except InnerException:
        # handles inner exception
except OuterException:
    # handles outer exception (or inner if inner didn't handle it)
finally:
    # always runs
```

```
  NESTED BLOCK CASES:
  ┌───────────────────────────────────────────────────────────────────┐
  │  Case 1: No exception anywhere                                    │
  │   → outer try + inner try + inner finally + outer finally run    │
  │                                                                   │
  │  Case 2: Exception in OUTER try (before inner try)               │
  │   → outer except + outer finally run                             │
  │                                                                   │
  │  Case 3: Exception in INNER try — handled by inner except        │
  │   → inner except + outer finally run                             │
  │                                                                   │
  │  Case 4: Exception in INNER try — NOT handled by inner except    │
  │   → bubbles up to outer except + outer finally run              │
  └───────────────────────────────────────────────────────────────────┘
```

````python
# Ex 14.16 - Nested blocks: Case 1 — no exception anywhere
distance = 120; time = 2; fuel_consumed = 10; Power = True

try:                                     # OUTER try
    speed = distance / time
    print(f"Speed   = {speed} mph")

    try:                                 # INNER try
        mileage = distance / fuel_consumed
        print(f"Mileage = {mileage} MPG")
    except ZeroDivisionError:
        print(f"Inner: Can't divide {distance} by {fuel_consumed}.")

except:                                  # OUTER default except
    print("Outer Default Exception!")
finally:                                 # OUTER finally
    Power = False; distance = 0; time = 0
    print(f"Power={Power}, Distance={distance}, Time={time}")
````

````python
# Ex 14.17 - Case 2: Exception in outer try → outer except handles it
distance = 120; time = 0; fuel_consumed = 10; Power = True

try:                                     # OUTER try
    speed = distance / time              # ZeroDivisionError HERE (outer level)
    print(f"Speed = {speed} mph")       # SKIPPED
    try:                                 # INNER try — never reached
        mileage = distance / fuel_consumed
        print(f"Mileage = {mileage} MPG")
    except ZeroDivisionError:
        print("Inner handler")
except:                                  # OUTER default handles it
    print("Outer Default Exception caught ZeroDivisionError!")
finally:
    Power = False; distance = 0; time = 0
    print(f"Power={Power}, Distance={distance}, Time={time}")
````

````python
# Ex 14.18 - Case 3: Exception in inner try — inner except handles it
distance = 120; time = 2; fuel_consumed = 0; Power = True

try:
    speed = distance / time              # OK
    print(f"Speed   = {speed} mph")
    try:
        mileage = distance / fuel_consumed  # ZeroDivisionError HERE (inner)
        print(f"Mileage = {mileage} MPG")  # SKIPPED
    except ZeroDivisionError:           # INNER except handles it
        print(f"Inner: Can't divide {distance} by {fuel_consumed}.")
except:
    print("Outer Default Exception!")
finally:
    Power = False; distance = 0; time = 0
    print(f"Power={Power}, Distance={distance}, Time={time}")
````

````python
# Ex 14.19 - Case 4: Exception in inner try — inner CANNOT handle → bubbles to outer
distance = 120; time = 2; fuel_consumed = 0; Power = True

try:
    speed = distance / time              # OK
    print(f"Speed   = {speed} mph")
    try:
        mileage = distance / fuel_consumed  # ZeroDivisionError
        print(f"Mileage = {mileage} MPG")
    except ValueError:                  # WRONG — can't handle ZeroDivision
        print("Inner: ValueError handler")
except:                                 # OUTER default catches the bubbled error
    print("Outer Default Exception catches the bubbled-up error!")
finally:
    Power = False; distance = 0; time = 0
    print(f"Power={Power}, Distance={distance}, Time={time}")
````

### The else Block in Exception Handling

The `else` block executes **only when NO exception occurred** in the `try` block.

```python
try:
    # risky code
except ExceptionName:
    # runs only IF exception occurred
else:
    # runs only IF NO exception occurred
finally:
    # always runs
```

```
  ELSE BLOCK LOGIC:
  ┌───────────────────────────────────────────────────────────────────┐
  │  No exception  →  try block ✓  →  else block ✓  →  finally ✓   │
  │  Exception     →  try block ✗  →  except block  →  finally ✓   │
  │                                →  else block SKIPPED             │
  └───────────────────────────────────────────────────────────────────┘

  Use case: put code that depends on try succeeding into else.
  This keeps exception-handling and success logic clearly separated.
```

### Full Pattern Guide — try / except / else / finally

```
  COMPLETE BLOCK DECISION TABLE:
  ─────────────────────────────────────────────────────────────────────────────────────
  Block      │  Runs when?                         │  Use it for
  ───────────┼─────────────────────────────────────┼─────────────────────────────────────
  try        │  Always (attempted first)            │  The risky code that might fail
  except     │  Only if matching exception in try   │  Handling / recovering from error
  else       │  Only if NO exception occurred       │  Code that depends on try succeeding
  finally    │  ALWAYS — success or failure         │  Cleanup, logging, state reset
  ─────────────────────────────────────────────────────────────────────────────────────

  VALID COMBINATIONS:
  ┌──────────────────────────────────────────────────────┬──────────────┐
  │  Combination                                         │  Valid?      │
  ├──────────────────────────────────────────────────────┼──────────────┤
  │  try + except                                        │  ✓ Yes       │
  │  try + finally                                       │  ✓ Yes       │
  │  try + except + finally                              │  ✓ Yes       │
  │  try + except + else                                 │  ✓ Yes       │
  │  try + except + else + finally  (full pattern)       │  ✓ Yes       │
  │  try alone (no except, no finally)                   │  ✗ No        │
  │  except alone (no try)                               │  ✗ No        │
  │  else without except                                 │  ✗ No        │
  │  default except: BEFORE specific except              │  ✗ SyntaxError│
  └──────────────────────────────────────────────────────┴──────────────┘

  FULL PATTERN EXAMPLE — Sipho's Grade Processor:
  ─────────────────────────────────────────────────────────────────────────
  student    = "Sipho Dlamini"
  mark_input = "85"
  processed  = False

  try:
      mark = int(mark_input)             # risky: could be ValueError
      if not (0 <= mark <= 100):
          raise ValueError("Mark out of range")
  except ValueError as e:
      print(f"Invalid mark for {student}: {e}")    # error path
  else:
      symbol = "A" if mark >= 80 else "B" if mark >= 70 else "C"
      print(f"{student}: {mark}% → {symbol}")      # success path only
      processed = True
  finally:
      status = "processed" if processed else "skipped"
      print(f"Record for {student}: {status}")      # ALWAYS runs
  ─────────────────────────────────────────────────────────────────────────
```

````python
# Ex 14.20 - else block: no exception → try + else + finally all run
distance = 120; time = 2; fuel_consumed = 10; Power = True

try:
    speed = distance / time           # succeeds
    print(f"Speed = {speed} mph")
except ZeroDivisionError:
    print(f"Can't divide {distance} by {fuel_consumed}.")
else:
    speed_in_kmph = speed * 1.60934   # runs ONLY because try succeeded
    print(f"Speed in km/h = {speed_in_kmph:.2f} km/h")
finally:
    Power = False; distance = 0; time = 0
    print(f"Power={Power}, Distance={distance}, Time={time}")
````

## 🛠️ Lesson 4 — Custom Exceptions

### Predefined vs Custom Exceptions

```
┌─────────────────────────────────┬──────────────────────────────────────────┐
│  Predefined Exceptions          │  Custom Exceptions                       │
├─────────────────────────────────┼──────────────────────────────────────────┤
│  Already available in Python    │  Defined by the programmer               │
│  e.g. ZeroDivisionError,        │  Inherit from the built-in Exception     │
│       ValueError, NameError     │  Used for project-specific scenarios     │
│  Handled by Python's PVM        │  Raised explicitly with the raise keyword │
└─────────────────────────────────┴──────────────────────────────────────────┘
```

### How to Create a Custom Exception

```
  STEP 1: Create a class that inherits from Exception

      class MyException(Exception):
          def __init__(self, message):
              self.message = message

  STEP 2: Raise it using the raise keyword where needed

      raise MyException("Something went wrong")

  STEP 3: Catch it like any other exception

      try:
          raise MyException("Something went wrong")
      except MyException as e:
          print(e.message)
```

### How to Design a Custom Exception — Step-by-Step

```
  STEP 1 — Inherit from Exception (or a relevant child class):
  ────────────────────────────────────────────────────────────────────────────
  class NegativeMarkError(Exception):     # always inherit from Exception
      def __init__(self, subject, mark):
          self.subject = subject
          self.mark    = mark
          self.message = f"Mark {mark} for {subject} cannot be negative"
          super().__init__(self.message)  # pass message to built-in Exception

      def __str__(self):
          return self.message             # what print(e) shows

  STEP 2 — Raise it explicitly with the raise keyword:
  ────────────────────────────────────────────────────────────────────────────
  def add_mark(subject, mark):
      if mark < 0:
          raise NegativeMarkError(subject, mark)   # raise keyword required
      print(f"{subject}: {mark}%")

  STEP 3 — Catch it like any built-in exception:
  ────────────────────────────────────────────────────────────────────────────
  try:
      add_mark("Maths", -5)
  except NegativeMarkError as e:
      print(f"Error: {e}")              # calls __str__ → our custom message
      print(f"Subject: {e.subject}")    # access custom attribute
      print(f"Bad mark: {e.mark}")      # access custom attribute
```

### Custom Exception Hierarchies

```
  BUILD A HIERARCHY WHEN:
  • Your project has many domain-specific errors related to one area
  • You want to catch ALL related errors with a single handler (parent)
  • You also want specific handling for individual types

  EXAMPLE HIERARCHY:
      AppError(Exception)               ← catch everything app-related
      │
      ├── ValidationError(AppError)     ← catch all validation errors
      │    ├── MissingFieldError(ValidationError)
      │    ├── InvalidEmailError(ValidationError)
      │    └── OutOfRangeError(ValidationError)
      │
      └── DatabaseError(AppError)       ← catch all database errors
           ├── ConnectionError(DatabaseError)
           └── QueryError(DatabaseError)

  Catching a parent class catches ALL children:
      except ValidationError  →  catches Missing, InvalidEmail, OutOfRange too
      except AppError         →  catches everything in the hierarchy

┌─────────────────────────────────────┬──────────────────────────────────────────────────┐
│  Use BUILT-IN Exception when:       │  Create CUSTOM Exception when:                   │
├─────────────────────────────────────┼──────────────────────────────────────────────────┤
│  Standard exceptions already        │  The error is specific to your business domain   │
│  describe the problem accurately    │  You need custom attributes (subject, amount)    │
│  e.g. ValueError for bad input type │  You want a hierarchy of related error classes   │
└─────────────────────────────────────┴──────────────────────────────────────────────────┘
```

````python
# Ex 14.28 - Custom Exception: NegativeError
class NegativeError(Exception):
    def __init__(self, data):
        self.data = data

# Simulate user input with a negative value
for distance in [100, -50, 200]:
    try:
        print(f"Input: {distance}")
        if distance < 0:
            raise NegativeError(distance)    # explicitly raise custom exception
        print(f"  Valid distance: {distance} miles")
    except NegativeError as e:
        print(f"  NegativeError: You provided {e.data}. Please provide positive values only.")
````

````python
# Ex 14.29 - Multiple Custom Exceptions for different scenarios
class NegativeDistanceException(Exception):
    def __init__(self, message):
        self.msg = message

class NegativeTimeException(Exception):
    def __init__(self, message):
        self.msg = message

def check_speed_inputs(distance, time):
    try:
        if distance < 0:
            raise NegativeDistanceException("The distance you provided is negative!")
        elif time < 0:
            raise NegativeTimeException("The time you provided is negative!")
        else:
            speed = distance / time
            print(f"Speed = {speed} mph")
    except NegativeDistanceException as e:
        print(f"NegativeDistanceException: {e.msg}")
    except NegativeTimeException as e:
        print(f"NegativeTimeException: {e.msg}")

check_speed_inputs(100,  2)   # valid
check_speed_inputs(-50,  2)   # NegativeDistanceException
check_speed_inputs(100, -2)   # NegativeTimeException
````

  <h3 style='color: #6c757d; margin-top: 0;'>📋 Key Takeaways — Chapter 14: Exception Handling</h3>
  <table style='width:100%; border-collapse: collapse;'>
    <tr style='background:#e9ecef;'><th style='padding:8px; text-align:left;'>Concept</th><th style='padding:8px; text-align:left;'>Key Point</th></tr>
    <tr><td style='padding:8px;'><strong>Error</strong></td><td style='padding:8px;'>Any unexpected condition that disrupts program execution. Two types: Syntax Error (before run) and Runtime Error (during run).</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>Syntax Error</strong></td><td style='padding:8px;'>Invalid grammar (missing colon, wrong indent). Python detects it BEFORE the program runs — cannot be caught with try-except; programmer must fix it.</td></tr>
    <tr><td style='padding:8px;'><strong>Runtime Error / Exception</strong></td><td style='padding:8px;'>Occurs DURING execution (division by zero, index out of range). CAN be caught and handled. Every exception is an instance of an exception class.</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>Normal vs Abnormal Flow</strong></td><td style='padding:8px;'>Normal: all statements run → expected output. Abnormal: exception at line N → all remaining lines SKIPPED → program terminates.</td></tr>
    <tr><td style='padding:8px;'><strong>Exception Handling</strong></td><td style='padding:8px;'>Does NOT fix the exception — provides an alternative path so the program avoids abnormal termination. Code after try-except continues to run.</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>try-except</strong></td><td style='padding:8px;'><code>try</code> wraps risky code. If an exception matches the <code>except ExceptionType</code>, the handler runs and execution continues after the block.</td></tr>
    <tr><td style='padding:8px;'><strong>Multiple except blocks</strong></td><td style='padding:8px;'>One except per exception type → custom message per error. Or group in a tuple: <code>except (ValueError, TypeError)</code> → single shared handler.</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>Default except</strong></td><td style='padding:8px;'>Bare <code>except:</code> catches ANY exception not matched above. MUST be the LAST except block — placing it first causes <code>SyntaxError</code>.</td></tr>
    <tr><td style='padding:8px;'><strong>finally</strong></td><td style='padding:8px;'>Runs ALWAYS — whether the try succeeded, an exception was caught, or even if it wasn't caught. Use for cleanup: close files, reset state, release resources.</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>else</strong></td><td style='padding:8px;'>Runs ONLY if the try block completed with NO exception. Use it to separate success-path code from the error-handling code — makes intent clearer.</td></tr>
    <tr><td style='padding:8px;'><strong>Nested try</strong></td><td style='padding:8px;'>An exception in the inner try that its except can't handle BUBBLES UP to the outer except. The outer finally still runs before the crash.</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>Valid combinations</strong></td><td style='padding:8px;'><code>try+except</code> ✓ | <code>try+finally</code> ✓ | <code>try+except+else</code> ✓ | <code>try+except+else+finally</code> ✓ | <code>else</code> without <code>except</code> ✗</td></tr>
    <tr><td style='padding:8px;'><strong>Custom Exception</strong></td><td style='padding:8px;'>Inherit from <code>Exception</code>: <code>class MyError(Exception): ...</code>. Raise with <code>raise MyError("message")</code>. Catch like any other exception.</td></tr>
    <tr style='background:#f8f9fa;'><td style='padding:8px;'><strong>Exception hierarchy</strong></td><td style='padding:8px;'><code>BaseException → Exception → ArithmeticError → ZeroDivisionError</code>. Catching a parent class also catches all its children.</td></tr>
    <tr><td style='padding:8px;'><strong>as e</strong></td><td style='padding:8px;'><code>except ZeroDivisionError as e: print(e)</code> — prints Python's own error message. Useful for debugging without hardcoding the message.</td></tr>
  </table>
