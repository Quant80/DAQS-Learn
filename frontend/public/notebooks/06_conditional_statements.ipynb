## 🔀 Lesson 1 — Introduction to Conditional Statements

**Conditional Statements** are statements executed based on a condition. They allow a program to make decisions — executing different blocks of code depending on whether a condition is `True` or `False`.

In Python, there are **four types** of conditional statements:

```
┌─────────────────────────────────────────────────────────┐
│           CONDITIONAL STATEMENTS IN PYTHON              │
├─────────────────────────────────────────────────────────┤
│                                                         │
│   1. if statement                                       │
│      └─ executes a block only when condition is True    │
│                                                         │
│   2. if else statement                                  │
│      └─ one block for True, another for False           │
│                                                         │
│   3. Nested if statement                                │
│      └─ an if statement inside another if statement     │
│                                                         │
│   4. if elif else statement                             │
│      └─ multiple conditions evaluated in order          │
│                                                         │
└─────────────────────────────────────────────────────────┘
```

<div style='border: 2px solid #0077cc; border-radius: 12px; padding: 15px; background: #e8f4fd; margin: 15px 0;'>
<strong style='color: #0077cc;'>💡 What is Indentation?</strong>
<ul style='margin: 8px 0 0;'>
<li><strong>Indentation</strong> refers to the spaces at the beginning of a code line.</li>
<li>Python uses indentation to indicate a <strong>block of code</strong> — unlike other languages that use curly braces <code>{}</code>.</li>
<li>All statements inside a conditional block must be indented by the same amount (typically 4 spaces or 1 tab).</li>
<li>Incorrect indentation causes an <code>IndentationError</code>.</li>

## ✅ Lesson 2 — The if Statement

The `if` statement executes a block of code **only when the condition is** `True`. If the condition is `False`, the block is skipped entirely.

**Syntax:**
```python
if (condition):
    if block statement
next line of code
```

```
┌──────────────────────────────────────────┐
│           if Statement Flow              │
│                                          │
│            ┌─────────────┐               │
│            │  condition  │               │
│            └──────┬──────┘               │
│                   │                      │
│           ┌───────┴───────┐              │
│          True           False            │
│           │               │              │
│    ┌──────▼──────┐        │              │
│    │ execute     │        │              │
│    │ if block    │        │              │
│    └──────┬──────┘        │              │
│           └───────────────┘              │
│                   │                      │
│           ┌───────▼───────┐              │
│           │  next code    │              │
│           └───────────────┘              │
└──────────────────────────────────────────┘
```

````python
# Ex 6.1 — if statement
a = int(input("Enter your age\n"))

if a > 17:
    print("Pass for interview.")

print("Thank you!")
````

## 🔗 Lesson 4 — Logical Operators & Compound Conditions

**Logical Operators** perform logical operations on operands and return `True` or `False`. They are used to build **compound conditions** — combinations of more than one simple condition.

In Python there are **three** logical operators:

```
┌─────────────┬───────────────────────────────────────────────────────┐
│  Operator   │  Description                                          │
├─────────────┼───────────────────────────────────────────────────────┤
│   and       │  True only if BOTH arguments are True                 │
│   or        │  True if AT LEAST ONE argument is True                │
│   not       │  Complement (opposite) of the boolean value           │
└─────────────┴───────────────────────────────────────────────────────┘
```

**Truth Tables:**

```
  AND Table          OR Table          NOT Table
┌───┬───┬─────┐   ┌───┬───┬─────┐   ┌───┬───────┐
│ A │ B │ A&B │   │ A │ B │ A|B │   │ A │ not A │
├───┼───┼─────┤   ├───┼───┼─────┤   ├───┼───────┤
│ T │ T │  T  │   │ T │ T │  T  │   │ T │   F   │
│ T │ F │  F  │   │ T │ F │  T  │   │ F │   T   │
│ F │ T │  F  │   │ F │ T │  T  │   └───┴───────┘
│ F │ F │  F  │   │ F │ F │  F  │
└───┴───┴─────┘   └───┴───┴─────┘
```

<div style='border: 2px solid #0077cc; border-radius: 12px; padding: 15px; background: #e8f4fd; margin: 15px 0;'>

<ul style='margin: 8px 0 0;'>

<li>These operators evaluate boolean types and return boolean results.</li>
<li>They allow you to combine multiple relational operators into a single condition.</li>

## 🪆 Lesson 5 — Nested if Statement

A **nested if statement** is an `if` statement placed inside another `if` statement. It allows a program to decide among multiple options based on **multiple conditions evaluated in layers**.

**Syntax:**
```python
if (condition 1):
    if (condition 2):
        if block statement of condition 2
    else:
        else block statement of condition 2
else:
    else block statement of condition 1
next line of code
```

```
┌─────────────────────────────────────────────────┐
│            Nested if Statement Flow             │
│                                                 │
│           ┌──────────────┐                      │
│           │ condition 1  │                      │
│           └──────┬───────┘                      │
│                  │                              │
│          ┌───────┴────────┐                     │
│         True            False                   │
│          │                │                     │
│   ┌──────▼──────┐   ┌─────▼──────┐              │
│   │ condition 2 │   │ else block │              │
│   └──────┬──────┘   └────────────┘              │
│          │                                      │
│   ┌──────┴────────┐                             │
│  True           False                           │
│   │               │                             │
│ ┌─▼──────┐  ┌─────▼─────┐                      │
│ │ inner  │  │ inner     │                      │
│ │ if blk │  │ else blk  │                      │
│ └────────┘  └───────────┘                      │
└─────────────────────────────────────────────────┘
```

**Ex 6.6** — Nested if: age range AND GPA check:

````python
# Ex 6.6 — Nested if statement
a = int(input("Enter your age\n"))
c = float(input("Enter your GPA\n"))

if a > 17 and a < 45:
    if c > 9:
        print("Accepted.")
    else:
        print("Waiting List...")
else:
    print("Application Denied.")
````

## 📊 Lesson 6 — The if elif else Statement

The `if elif else` statement handles **multiple conditions** evaluated in written order. The conditions are evaluated in sequence — as soon as one becomes `True`, its block executes and the remaining conditions are **skipped**.

**Syntax:**
```python
if (condition):
    if block statement
elif:
    elif block statement
else:
    else block statement
next line of code
```

```
┌──────────────────────────────────────────────────────┐
│             if elif else Statement Flow              │
│                                                      │
│           ┌──────────────────┐                       │
│           │   condition 1    │                       │
│           └────────┬─────────┘                       │
│                    │                                 │
│           ┌────────┴─────────┐                       │
│          True              False                     │
│           │                  │                       │
│   ┌───────▼──────┐   ┌───────▼──────────┐            │
│   │  if block    │   │   condition 2    │            │
│   └──────────────┘   └───────┬──────────┘            │
│                              │                       │
│                     ┌────────┴─────────┐             │
│                    True              False            │
│                     │                  │             │
│              ┌──────▼──────┐   ┌───────▼──────┐      │
│              │ elif block  │   │  else block  │      │
│              └─────────────┘   └──────────────┘      │
└──────────────────────────────────────────────────────┘
```

**Ex 6.7** — `if elif else` with age AND GPA compound conditions:

````python
# Ex 6.7 — if elif else statement
a = int(input("Enter your age\n"))
c = float(input("Enter your GPA\n"))

if a > 17 and a < 45 and c > 9:
    print("Accepted.")
elif a > 17 and a < 45 and c > 7 and c < 9:
    print("Pass for interview.")
else:
    print("Application Denied.")
````

## 💡 Lesson 7 — More Examples: Largest Number & Variable Swapping

### Finding the Largest Number

Conditional statements are commonly used to find the maximum of two or three numbers.

**Ex 6.8** — Finding largest of two numbers (basic `if else`):

````python
# Ex 6.8 — if else statement (largest of two)
a = float(input("Enter 1st No: "))
b = float(input("Enter 2nd No: "))

if a > b:
    print(a, "is the largest No.")
else:
    print(b, "is the largest No.")
````

<div style='border: 2px solid #0077cc; border-radius: 12px; padding: 15px; background: #e8f4fd; margin: 15px 0;'>
<strong style='color: #0077cc;'>💡 The Ternary (Conditional) Operator:</strong>
<ul style='margin: 8px 0 0;'>
<li>Python supports a compact one-line conditional called the <strong>ternary operator</strong>.</li>
<li>Syntax: <code>[value_if_True] if (condition) else [value_if_False]</code></li>
<li>It is useful for simple assignments based on a condition without writing a full if else block.</li>
</ul>
</div>

**Ex 6.9** — Largest number using the ternary operator:

````python
# Ex 6.9 — if else using ternary (conditional) operator
a = float(input("Enter 1st No: "))
b = float(input("Enter 2nd No: "))

max = a if a > b else b

print("Largest No.=", max)
````

````python
# Ex 6.10 — if elif else (handles equal numbers case)
a = float(input("Enter 1st No: "))
b = float(input("Enter 2nd No: "))

if a > b:
    print(a, "is the largest No.")
elif a == b:
    print(a, "and", b, "are equal.")
else:
    print(b, "is the largest No.")
````

**Ex 6.11** — Half nested if: largest of three numbers (has a bug for equal values — see Ex 6.12):

````python
# Ex 6.11 — Half Nested if (largest of three)
# Bug: fails for inputs like (1, 1, 2) — else always prints b without checking c
a = float(input("Enter 1st no: "))
b = float(input("Enter 2nd no: "))
c = float(input("Enter 3rd no: "))

if a > b:
    if a > c:
        print("largest no.= ", a)
    else:
        print("largest no.= ", c)
else:
    print("largest no.= ", b)
````

<div style='border: 2px solid #fd7e14; border-radius: 12px; padding: 15px; background: #fff8f0; margin: 15px 0;'>
<strong style='color: #fd7e14;'>⚠️ Bug Fix:</strong> Ex 6.11 gives wrong output for inputs like (1, 1, 2), (1, 2, 2), (2, 1, 2), or (2, 2, 2) because the <code>else</code> branch always prints <code>b</code> without checking <code>c</code>. Ex 6.12 fixes this using <code>&gt;=</code> comparisons and a full nested if.

**Ex 6.12** — Full nested if: correct largest of three (handles all equal cases):

````python
# Ex 6.12 — Full Nested if (correct for all inputs)
a = float(input('Enter 1st No: '))
b = float(input('Enter 2nd No: '))
c = float(input('Enter 3rd No: '))

if(a >= b):
    if(a >= c):
        largest = a
    else:
        largest = c
else:
    if(b >= c):
        largest = b
    else:
        largest = c

print("The largest No is", largest)
````

### Variable Swapping — 4 Techniques

Swapping exchanges the values of two variables. Python offers four techniques:

```
┌────────────────────────────────────────────────────────────┐
│              Variable Swapping Techniques                  │
├────────────────────┬───────────────────────────────────────┤
│  Technique         │  Code                                 │
├────────────────────┼───────────────────────────────────────┤
│  Third variable    │  temp=a; a=b; b=temp                  │
│  Assignment op     │  a, b = b, a                          │
│  + and - ops       │  a=a+b; b=a-b; a=a-b                  │
│  XOR (^) op        │  a=a^b; b=a^b; a=a^b                  │
└────────────────────┴───────────────────────────────────────┘
```

**Ex 6.13** — Swap using a third (temporary) variable:

````python
# Ex 6.13 — Swap using third variable
a, b = 1, 2

print("Before Swap a=", a, "b=", b)
temp = a
a = b
b = temp
print("After Swap a=", a, "b=", b)
````

````python
# Ex 6.14 — Swap using Assignment operator (the Pythonic way)
a, b = 1, 2

print("Before Swap a=", a, "b=", b)
a, b = b, a
print("After Swap a=", a, "b=", b)
````

````python
# Ex 6.15 — Swap using + and - operators
a, b = 1, 2

print("Before Swap a=", a, "b=", b)
a = a + b   # a = 1+2 = 3
b = a - b   # b = 3-2 = 1  (original a)
a = a - b   # a = 3-1 = 2  (original b)
print("After Swap a=", a, "b=", b)
````

````python
# Ex 6.16 — Swap using * and / operators
a, b = 1, 2

print("Before Swap a=", a, "b=", b)
a = a * b
b = a / b
a = a / b
print("After Swap a=", a, "b=", b)
````

## ⚙️ Lesson 8 — Bitwise Operators

**Bitwise Operators** are operators used to perform calculations directly on the **binary bit-level representation** of integers.

<div style='border: 2px solid #fd7e14; border-radius: 12px; padding: 15px; background: #fff8f0; margin: 15px 0;'>
<strong style='color: #fd7e14;'>⚠️ Important:</strong> Bitwise operators work <strong>only on integers</strong>. They cannot be applied to floats or strings.
</div>

There are **six** bitwise operators in Python:

```
┌──────────────────┬──────────┬─────────────────────────────────────────┐
│  Operator        │  Symbol  │  Description                            │
├──────────────────┼──────────┼─────────────────────────────────────────┤
│  Bitwise AND     │    &     │  1 if both bits are 1                   │
│  Bitwise OR      │    |     │  1 if at least one bit is 1             │
│  Bitwise NOT     │    ~     │  Flips all bits (complement)            │
│  Bitwise XOR     │    ^     │  1 if bits are different                │
│  Right Shift     │   >>     │  Shifts bits right (divides by 2^n)     │
│  Left Shift      │   <<     │  Shifts bits left (multiplies by 2^n)   │
└──────────────────┴──────────┴─────────────────────────────────────────┘
```

```
  Shift Operations on integer 16 (binary: 00010000):
  ┌──────────────────────────────────────────────┐
  │  16 >> 1  →  8   (00001000)  divide by 2    │
  │  16 << 1  →  32  (00100000)  multiply by 2  │
  └──────────────────────────────────────────────┘
```

### Case Study — The Bus Stop Problem

Bus stops are numbered where each stop is **double** the previous: 0 → 1 → 2 → 4 → 8 → 16 → 32. Given the current stop number, find the previous and next stops.

**Ex 6.17** — Bus Stop Problem using arithmetic operators:

````python
# Ex 6.17 — Bus Stop Problem by Arithmetic operators
c = int(input("enter current stop\n"))

p = c / 2
p = int(p)
n = c * 2

print("Previous bus stop is", p)
print("Current bus stop is", c)
print("Next bus stop is", n)
````

**Ex 6.18** — Bus Stop Problem using **bitwise shift** operators (elegant solution!):

````python
# Ex 6.18 — Bus Stop Problem using bitwise operators
c = int(input("Enter current stop\n"))

p = c >> 1   # right shift = divide by 2
n = c << 1   # left shift = multiply by 2
print("Previous bus stop is", p)
print("Current bus stop is", c)
print("Next bus stop is", n)
````

**Ex 6.19** — Swapping using **XOR (^)** operator (no temporary variable needed!):

````python
# Ex 6.19 — Swapping Using XOR (^) operator
a = 1
b = 2

print("Before Swap a=", a, "b=", b)
a = a ^ b
b = a ^ b
a = a ^ b
print("After Swap a=", a, "b=", b)
````

## 🔍 Lesson 9 — Membership Operators

**Membership Operators** check whether an element is **present in** or **absent from** a sequence. The sequence can be a string, list, tuple, set, or dictionary.

There are **two** membership operators:

```
┌────────────────┬──────────────────────────────────────────────────────┐
│  Operator      │  Description                                         │
├────────────────┼──────────────────────────────────────────────────────┤
│  in            │  Returns True if element IS found in the sequence    │
│  not in        │  Returns True if element is NOT found in sequence    │
└────────────────┴──────────────────────────────────────────────────────┘
```

**Ex 6.20** — Membership operators with a string (case-sensitive!):

````python
# Ex 6.20 — Membership Operators on a string
text = "Welcome"

print("Well" in text)     # False — 'Well' not in 'Welcome'
print("wel" in text)      # False — case sensitive: 'w' != 'W'
print("Wel" in text)      # True  — 'Wel' IS in 'Welcome'
print("come" not in text) # False — 'come' IS in 'Welcome', so 'not in' is False
````

**Ex 6.21** — Membership operators with a list:

````python
# Ex 6.21 — Membership Operators on a list
class_10A = ["John", "Eden", "Bob"]

print("john" in class_10A)      # False — case sensitive: 'john' != 'John'
print("Lucas" not in class_10A) # True  — Lucas is not in the list
````

````python
# Practical — Membership Operators in a real scenario
student = input("Enter student name to check: ")
class_10A = ["John", "Eden", "Bob", "Alice", "Mary"]

if student in class_10A:
    print(f"{student} is in Class 10A.")
else:
    print(f"{student} is NOT in Class 10A.")
````

## 🧮 Lesson 10 — The eval() Function

The `eval()` function is a **built-in Python function** that takes a **string as input**, evaluates it as a Python expression, and returns the result.

```
┌────────────────────────────────────────────────────┐
│         How eval() Works                           │
│                                                    │
│   eval( '10 + 10' )  →  20                         │
│   eval( '2 ** 8'  )  →  256                        │
│   eval( '1 > 2'   )  →  False                      │
│   eval( '2 << 3'  )  →  16                         │
│                                                    │
│   Input: STRING → eval() → Output: RESULT          │
└────────────────────────────────────────────────────┘
```

</div>
