## 🏭 Lesson 1 — Introduction to Object-Oriented Programming

### The 3 Programming Paradigms

A **programming paradigm** is a style or methodology for writing programs.

```
┌─────────────────────────────────────────────────────────────────────────────┐
│                      PROGRAMMING PARADIGMS                                  │
├──────────────────────────┬──────────────────────────────────────────────────┤
│  Paradigm                │  Description                                     │
├──────────────────────────┼──────────────────────────────────────────────────┤
│  Functional Oriented     │  Programs built around FUNCTIONS. Functions      │
│                          │  always return the same output for same input.   │
├──────────────────────────┼──────────────────────────────────────────────────┤
│  Procedural Oriented     │  Step-by-step instructions (like a recipe).     │
│                          │  Emphasizes executing instructions in order.     │
├──────────────────────────┼──────────────────────────────────────────────────┤
│  Object Oriented (OOP)   │  Programs built around OBJECTS — reusable        │
│                          │  pieces with data (attributes) and actions       │
│                          │  (methods). Python supports both OOP and         │
│                          │  functional styles.                              │
└──────────────────────────┴──────────────────────────────────────────────────┘
```

### The TV Factory Analogy

OOP concepts are easiest to understand through a real-world analogy:

```
  TV FACTORY ANALOGY
  ═══════════════════════════════════════════════════════════════

  CLASS  ──►  Blueprint / Template  (the design for making a TV)
              Defines:  brand, model, screen_size  ← attributes
                        turn_on(), change_channel() ← behaviours

  OBJECT ──►  Actual TV manufactured from the blueprint
              salon_tv   = Television("Sony",    "X80K", 65)
              bedroom_tv = Television("Samsung", "N5300", 32)

              Each TV is a SEPARATE object with its OWN data
              but shares the same CLASS structure.

  CONSTRUCTOR ──►  The factory line that builds each TV object.
                   Runs AUTOMATICALLY when a new object is created.
  ═══════════════════════════════════════════════════════════════
```

### Key OOP Terms

```
┌────────────────┬──────────────────────────────────────────────────────────┐
│  Term          │  Definition                                              │
├────────────────┼──────────────────────────────────────────────────────────┤
│  Class         │  Blueprint / template for creating objects               │
│  Object        │  A specific instance of a class                          │
│  Attribute     │  Variable inside a class (data)                          │
│  Method        │  Function inside a class (behaviour)                     │
│  Constructor   │  Special method auto-called when object is created       │
│  self          │  Default variable that refers to the current object      │
└────────────────┴──────────────────────────────────────────────────────────┘
```

### What is `self`?

`self` is a variable that always refers to **the current object** calling the method. It lets methods access their own object's data.

```
  Why use self?
  ┌─────────────────────────────────────────────────────────────┐
  │  1. Access instance variables  →  self.brand, self.volume   │
  │  2. Differentiate instance vs local variables               │
  │  3. Call other methods on same object  →  self.method()     │
  │  4. Identify which specific object is being acted on        │
  └─────────────────────────────────────────────────────────────┘

  Note: 'self' is a convention, not a keyword.
        You could call it 'this' or 'tv' — but 'self' is standard.
```

````python
# Ex 13.1 - Define a class (no output yet — no object created)
class Television:
    brand = "Sony"

    def turn_on(self):
        print(self.brand, "TV is on.")

print("Class defined. No output until we create an object.")
````

````python
# Ex 13.2 - Creating an object
my_tv = Television()
print(my_tv)   # shows memory address by default
````

````python
# Ex 13.3 - __str__() gives a human-readable string for the object
class Television:
    brand = "Sony"

    def turn_on(self):
        print(self.brand, "TV is on.")

    def __str__(self):
        return f"This is a {self.brand} Television object"

my_tv = Television()
print(my_tv)           # now calls __str__
print(type(my_tv))     # Ex 13.4
````

````python
# Ex 13.5 - Calling methods and accessing variables
my_tv.turn_on()                    # via object name
Television.turn_on(my_tv)          # via class name (equivalent)
print("Brand:", my_tv.brand)       # accessing class variable
````

## 🔧 Lesson 2 — Constructors

A **constructor** is a special method (`__init__`) that runs **automatically** the moment an object is created. It is used to set up the object's initial state.

```
  Syntax:

      def __init__(self, [parameters]):
          # initialization code

  ┌──────────────────────────┬────────────────────────────────────────────┐
  │  Constructor Type        │  Description                               │
  ├──────────────────────────┼────────────────────────────────────────────┤
  │  Non-parameterized       │  No arguments — uses default values        │
  │  Parameterized           │  Takes arguments — sets custom values      │
  └──────────────────────────┴────────────────────────────────────────────┘

  Method vs Constructor:
  ┌─────────────────────────┬──────────────────────────────────────────────┐
  │  Method                 │  Constructor                                 │
  ├─────────────────────────┼──────────────────────────────────────────────┤
  │  Does operations        │  Initializes instance variables              │
  │  Any name               │  Must be named __init__                      │
  │  Called explicitly      │  Auto-called at object creation              │
  └─────────────────────────┴──────────────────────────────────────────────┘
```

````python
# Ex 13.6 - Non-parameterized Constructor (runs automatically)
class Television:
    def __init__(self):
        print("Constructor executed automatically at TV creation!")

my_tv = Television()   # constructor fires here
````

````python
# Ex 13.8 - Non-parameterized constructor prints self (the object reference)
class Television:
    def __init__(self):
        print("Constructor executed.")
        print("self refers to:", self)

my_tv = Television()
````

````python
# Ex 13.9 - Parameterized Constructor — pass values at creation
class Television:
    def __init__(self, brand, model, screen_size):
        self.brand = brand
        self.model = model
        self.screen_size = screen_size
        print(f"This TV is {brand} {screen_size}" + '" ' + f"{model}")

# Create MULTIPLE objects from the SAME class
salon_tv   = Television("Sony",    "Class X80K",       65)
bedroom_tv = Television("Samsung", "Class N5300 Series", 32)
````

## 📦 Lesson 3 — Class Variables and Methods

### Three Types of Variables in a Class

```
┌────────────────────────────────────────────────────────────────────────────┐
│                      TYPES OF CLASS VARIABLES                              │
├───────────────────────┬──────────────────┬─────────────────────────────────┤
│  Type                 │  Scope           │  Changes per object?            │
├───────────────────────┼──────────────────┼─────────────────────────────────┤
│  Instance Variable    │  Object level    │  YES — each object has its own  │
│  Static Variable      │  Class level     │  NO  — shared by all objects    │
│  Local Variable       │  Method level    │  Only exists inside the method  │
└───────────────────────┴──────────────────┴─────────────────────────────────┘

  Instance Variable examples: brand, model, screen_size (differ per TV)
  Static Variable examples:   display="LED" (all TVs use LED)
  Local Variable examples:    min_volume, max_volume (temp, in one method)
```

### Three Types of Methods in a Class

```
┌──────────────────────────────────────────────────────────────────────────────┐
│                      TYPES OF CLASS METHODS                                  │
├───────────────────────┬────────────────────┬─────────────────────────────────┤
│  Type                 │  Decorator         │  First Parameter                │
├───────────────────────┼────────────────────┼─────────────────────────────────┤
│  Instance Method      │  (none)            │  self  — acts on object data    │
│  Class Method         │  @classmethod      │  cls   — acts on class data     │
│  Static Method        │  @staticmethod     │  (none) — utility function      │
└───────────────────────┴────────────────────┴─────────────────────────────────┘
```

````python
# Ex 13.10 - Instance variables declared in a constructor
class Television:
    def __init__(self, brand, model, screen_size):
        self.brand = brand           # instance variable
        self.model = model           # instance variable
        self.screen_size = screen_size  # instance variable

salon_tv   = Television("Sony",    "Class X80K",        65)
bedroom_tv = Television("Samsung", "Class N5300 Series", 32)

print(salon_tv.__dict__)    # shows all instance variables as dict
print(bedroom_tv.__dict__)
````

````python
# Ex 13.11/12 - Instance variables in methods and via object name
class Television:
    def __init__(self, brand, model, screen_size):
        self.brand = brand
        self.model = model
        self.screen_size = screen_size

    def display_details(self):
        self.resolution = "4K"    # instance variable created in method
        self.display_type = "LED"
        print("TV Details")
        print(f"  Model      : {self.brand} {self.model}")
        print(f"  Screen     : {self.screen_size}\"")
        print(f"  Display    : {self.resolution} {self.display_type}")

my_tv = Television("Sony", "Class X80K", 65)
my_tv.price = 2400              # instance variable via object name
my_tv.display_details()
print(my_tv.__dict__)
````

````python
# Ex 13.14/21 - Static variables — shared by ALL objects
class Television:
    display   = "LED"           # static variable
    resolution = "4K"           # static variable
    width     = 16
    height    = 9
    brand     = "Sony"
    model     = "Class X80K"
    netflix_web_url = "https://netflix.com"

    def __init__(self):
        print("Display :", self.display, Television.resolution)  # two access styles

    def aspect_ratio(self):
        print("Aspect Ratio is", self.width, ":", Television.height)

    @classmethod
    def display_details(cls):
        print("TV Model :", cls.brand, Television.model)

    @staticmethod
    def connect_to_netflix():
        print("Connecting to", Television.netflix_web_url)

my_tv = Television()
my_tv.aspect_ratio()
my_tv.display_details()       # class method
my_tv.connect_to_netflix()    # static method
Television.connect_to_netflix()  # can also call via class name (no object needed)
````

````python
# Ex 13.22 - Local Variables — exist only inside the method
class Television:
    volume = 15

    def volume_up(self):
        min_volume, max_volume = 0, 100   # LOCAL variables
        if Television.volume == max_volume:
            Television.volume = max_volume
        else:
            Television.volume += 1
        print("Volume :", Television.volume)

my_tv = Television()
my_tv.volume_up()

# Local variable 'max_volume' is NOT accessible outside:
try:
    print(max_volume)  # NameError!
except NameError as e:
    print("NameError:", e, "— local variables only exist inside their method")
````

### Setter and Getter Methods

```
  Setter (mutator)  — SETS a value to an instance variable
      def set_channel(self, channel): self.channel = channel

  Getter (accessor) — GETS (returns) the value of an instance variable
      def get_channel(self): return self.channel

  Why use them? Validate data before storing it.
  Example: only allow channels that actually exist.
```

````python
# Ex 13.24 - Setter and Getter methods
class Television:
    def __init__(self):
        self.channels = {1: "Aljezzira", 2: "BBC", 3: "CNN"}

    def change_channel(self, channel):   # SETTER — validates first
        if channel >= 1 and channel <= len(self.channels):
            self.channel = channel
            channel_name = self.get_channel()   # uses getter
            print("Now watching:", channel_name)
        else:
            print("Channel does not exist!")

    def get_channel(self):   # GETTER — returns the current channel name
        return self.channels[self.channel]

my_tv = Television()
my_tv.change_channel(2)   # valid
my_tv.change_channel(5)   # invalid
````

## 🔒 Lesson 4 — Encapsulation, Name Mangling & Nested Classes

### What is Encapsulation?

**Encapsulation** restricts direct access to an object's internal data, protecting it from accidental or unauthorised changes.

```
  Benefits of Encapsulation:
  ┌─────────────────────────────────────────────────────────────┐
  │  1. Data Hiding    — users can't see internal details       │
  │  2. Code Reusability — encapsulated code is easier to test  │
  │  3. Security       — prevents unauthorised data access      │
  └─────────────────────────────────────────────────────────────┘

  Access Modifiers in Python (via naming convention):
  ┌───────────────────┬──────────┬──────────────────────────────────────────┐
  │  Access Level     │  Prefix  │  Accessible from                         │
  ├───────────────────┼──────────┼──────────────────────────────────────────┤
  │  Public           │  (none)  │  Anywhere (inside or outside class)      │
  │  Protected        │  _       │  Inside class and its subclasses         │
  │  Private          │  __      │  Only inside the class                   │
  └───────────────────┴──────────┴──────────────────────────────────────────┘

  Example:
      self.channel  = "BBC"   # Public  — open to all
      self._volume  = 15      # Protected — convention: treat with care
      self.__brand  = "Sony"  # Private — name-mangled to _Television__brand
```

````python
# Ex 13.25 - Public, protected, and private attributes & methods
class Television:
    def __init__(self):
        self.channel  = "BBC"   # Public
        self._volume  = 15      # Protected
        self.__brand  = "Sony"  # Private

    def turn_on(self):
        print(f"The {self.__brand} TV is now on.")  # private OK inside class

    def _change_channel(self, channel):   # protected method
        print(f"Channel changed to {channel}.")

    def __adjust_volume(self, volume):    # private method
        print(f"Volume adjusted to {volume}.")

my_tv = Television()
my_tv.turn_on()                   # works — public method
print(my_tv.channel)              # works — public attribute
````

````python
# Ex 13.26 - Protected method can be called from outside (but shouldn't be)
my_tv._change_channel("CNN")   # works but discouraged

# Ex 13.27 - Private method CANNOT be called from outside
try:
    my_tv.__adjust_volume(30)  # AttributeError!
except AttributeError as e:
    print("AttributeError:", e)
````

````python
# Ex 13.31 - Accessing private variable via a Getter method
class Television:
    def __init__(self):
        self.channel  = "BBC"
        self._volume  = 15
        self.__brand  = "Sony"   # private

    def turn_on(self):
        print(f"The {self.__brand} TV is on.")

    def get_brand(self):          # getter exposes private safely
        return self.__brand

my_tv = Television()
print("Brand:", my_tv.get_brand())   # access via getter — correct way
````

### Nested Classes

A **nested class** is a class defined inside another class. The inner class is only accessible through the outer class.

```
  Television (outer class)
  │
  ├── volume, channels    (Television attributes)
  ├── turn_on()           (Television method)
  └── Remote (nested class)
       ├── battery_percentage
       ├── change_channel()
       └── increase_volume()
```

````python
# Ex 13.32 - Nested Classes
class Television:
    def __init__(self):
        self.volume   = 15
        self.channels = {1: "Aljezzira", 2: "BBC", 3: "CNN"}
        self.remote   = self.Remote(self)   # create nested object

    def turn_on(self):
        print("The TV is on.")

    class Remote:    # nested class
        def __init__(self, television):
            self.battery_percentage = 100
            self.television = television

        def change_channel(self, channel):
            print(f"Changing channel to {self.television.channels[channel]}")

        def increase_volume(self):
            self.television.volume += 1
            print("Increasing volume:", self.television.volume)

tv = Television()
tv.turn_on()
tv.remote.change_channel(2)
tv.remote.increase_volume()
````

````python
# Ex 13.33 - Garbage Collection
# Python automatically frees memory for objects with no references.
import gc

print("Garbage collector enabled?:", gc.isenabled())  # True by default
gc.disable()
print("After disable:", gc.isenabled())                # False
gc.enable()
print("After enable:",  gc.isenabled())                # True
````

## 🧬 Lesson 5 — Inheritance

**Inheritance** lets a new class (child) reuse all the variables, methods, and constructors of an existing class (parent).

```
  Benefits:
  • Code reusability  — don't rewrite methods that already exist
  • Less development time
  • Reduced code repetition

  Terminology:
  ┌────────────────────────────┬────────────────────────────────┐
  │  Existing class (parent)   │  New class (child)             │
  ├────────────────────────────┼────────────────────────────────┤
  │  Super class               │  Sub class                     │
  │  Base class                │  Derived class                 │
  │  Parent class              │  Child class                   │
  └────────────────────────────┴────────────────────────────────┘

  3 Types of Inheritance:
  ┌───────────────────────────────────────────────────────────────┐
  │  Single      →  1 parent  →  1 child                         │
  │  Multi-level →  Grandparent → Parent → Child (chain)         │
  │  Multiple    →  2+ parents → 1 child                         │
  └───────────────────────────────────────────────────────────────┘

  Syntax: class Child(Parent):
```

### Single Inheritance

```
  Television (parent)
      │
      └──► Flat_TV (child) — inherits turn_on(), change_channel()
                             adds its own HDMI_mode()
```

````python
# Ex 13.35 - Single Inheritance
class Television:            # PARENT
    def __init__(self):
        self.channels = {1: "Aljezzira", 2: "BBC", 3: "CNN"}

    def turn_on(self):
        print("The TV is on.")

    def change_channel(self, channel):
        print(f"Changing to {self.channels[channel]}")

class Flat_TV(Television):   # CHILD — inherits from Television
    def HDMI_mode(self):
        self.mode = "HDMI"
        print("Mode :", self.mode)

my_flat_tv = Flat_TV()
my_flat_tv.turn_on()          # inherited from Television
my_flat_tv.change_channel(3)  # inherited from Television
my_flat_tv.HDMI_mode()        # own method
````

### Multilevel Inheritance

```
  Television (grandparent)
      │
      └──► Flat_TV (parent)  — adds HDMI_mode()
                │
                └──► Smart_TV (child) — adds connect_to_netflix()

  Smart_TV inherits from BOTH Flat_TV AND Television.
```

````python
# Ex 13.36 - Multilevel Inheritance
class Television:
    def __init__(self):
        self.channels = {1: "Aljezzira", 2: "BBC", 3: "CNN"}

    def turn_on(self):         print("The TV is on.")
    def change_channel(self, c): print(f"Changing to {self.channels[c]}")

class Flat_TV(Television):
    def HDMI_mode(self):       print("Mode : HDMI")

class Smart_TV(Flat_TV):      # inherits from Flat_TV (and Television)
    def connect_to_netflix(self): print("Connecting to Netflix...")

my_smart_tv = Smart_TV()
my_smart_tv.turn_on()            # from Television
my_smart_tv.change_channel(3)    # from Television
my_smart_tv.HDMI_mode()          # from Flat_TV
my_smart_tv.connect_to_netflix() # own method
````

### Multiple Inheritance

```
  Television (parent 1)    Smart_Speaker (parent 2)
        │                         │
        └─────────────────────────┘
                     │
                 Smart_TV (child)

  When BOTH parents have the same method name,
  the ORDER in the class definition decides which one is used:

  class Smart_TV(Television, Smart_Speaker)  → Television.method wins
  class Smart_TV(Smart_Speaker, Television)  → Smart_Speaker.method wins
```

````python
# Ex 13.37 - Multiple Inheritance
class Television:
    def __init__(self):
        self.volume   = 15
        self.channels = {1: "Aljezzira", 2: "BBC", 3: "CNN"}

    def turn_on(self):         print("The TV is on.")
    def change_channel(self, c): print(f"Channel: {self.channels[c]}")

class Smart_Speaker:
    def increase_volume(self):
        self.volume += 1
        print("Volume :", self.volume)

class Smart_TV(Television, Smart_Speaker):  # inherits from BOTH
    def connect_to_netflix(self): print("Connecting to Netflix...")

my_smart_tv = Smart_TV()
my_smart_tv.turn_on()
my_smart_tv.change_channel(3)
my_smart_tv.increase_volume()    # from Smart_Speaker
my_smart_tv.connect_to_netflix()
````

````python
# Ex 13.38 & 13.39 - Order of inheritance determines method priority
class Television:
    def __init__(self): self.volume = 15
    def increase_volume(self):
        self.volume += 1
        print("TV Volume :", self.volume)

class Smart_Speaker:
    def increase_volume(self):
        self.volume += 1
        print("Speaker Volume :", self.volume)

# Case 1: Television first → TV volume
class Smart_TV_A(Television, Smart_Speaker):
    pass

# Case 2: Smart_Speaker first → Speaker volume
class Smart_TV_B(Smart_Speaker, Television):
    pass

print("Case 1 (Television first):")
Smart_TV_A().increase_volume()

print("\nCase 2 (Smart_Speaker first):")
Smart_TV_B().increase_volume()
````

## 🦆 Lesson 6 — Polymorphism

**Polymorphism** = *poly* (many) + *morphs* (forms). It is the ability of an object to take on many forms.

```
  Types of Polymorphism in Python:
  ┌────────────────────────────────────────────────────────────────┐
  │  1. Overriding                                                 │
  │     1.1 Method Overriding   — child redefines parent method    │
  │     1.2 Constructor Overriding — child redefines __init__      │
  │                                                                │
  │  2. Overloading (not native — use alternatives)               │
  │     2.1 Method Overloading    — same name, diff params        │
  │     2.2 Constructor Overloading                               │
  │     2.3 Operator Overloading  — +, * used for different types │
  │                                                                │
  │  3. Duck Typing Philosophy                                     │
  └────────────────────────────────────────────────────────────────┘
```

### 1.1 Method Overriding

````python
# Ex 13.40 - Method Overriding — child replaces parent's method
class Smart_Speaker:
    def __init__(self): self.volume = 15

    def increase_volume(self):
        self.volume += 1
        print(f"Smart Speaker Volume : {self.volume}")

class Smart_TV(Smart_Speaker):
    def increase_volume(self):    # OVERRIDES parent's method
        self.volume += 1
        print(f"Smart TV Volume : {self.volume}")

my_smart_tv = Smart_TV()
my_smart_tv.increase_volume()    # child's version is called
````

### The `super()` Function

`super()` lets you call the **parent class constructor, methods, or variables** from inside the child class.

```
  Use cases:
  ┌────────────────────────────────────────────────────────────────┐
  │  super().__init__()       → call parent's constructor          │
  │  super().method_name()    → call parent's method               │
  │  super().variable_name    → access parent's static variable    │
  └────────────────────────────────────────────────────────────────┘
```

````python
# Ex 13.42 - Calling parent constructor from child using super()
class Smart_Speaker:
    def __init__(self):
        self.volume = 15
        print("Default Smart Speaker Volume :", self.volume)

class Smart_TV(Smart_Speaker):
    def __init__(self):
        self.volume = 20
        super().__init__()    # also runs parent __init__
        print("Smart TV Volume :", self.volume)

my_smart_tv = Smart_TV()
# Note: super().__init__() resets self.volume to 15
````

````python
# Ex 13.43 - Calling parent method from child method using super()
class Smart_Speaker:
    volume = 15
    def increase_volume(self):
        self.volume += 1
        print("Smart Speaker Volume :", self.volume)

class Smart_TV(Smart_Speaker):
    volume = 20
    def increase_volume(self):
        super().increase_volume()   # call parent method first
        print("Smart TV Volume :", self.volume)

my_smart_tv = Smart_TV()
my_smart_tv.increase_volume()
````

````python
# Ex 13.45 - Accessing parent's variable using super()
class Smart_Speaker:
    volume = 15

class Smart_TV(Smart_Speaker):
    volume = 20
    def increase_volume(self):
        self.volume = super().volume + 1   # use PARENT's volume (15), not own (20)
        print("Smart TV Volume :", self.volume)

Smart_TV().increase_volume()   # 15 + 1 = 16
````

### 2.3 Operator Overloading

Python already supports operator overloading — the `+` and `*` operators work differently depending on the data type:

````python
# Ex 13.52 - '+' Operator Overloading
print(10 + 20)                   # arithmetic addition
print("Python" + "Programming") # string concatenation
print([1, 2, 3] + [4, 5, 6])    # list concatenation

print()
# Ex 13.53 - '*' Operator Overloading
print(10 * 20)             # multiplication
print("Python" * 3)        # string repetition
print([1, 2, 3] * 3)       # list repetition
````

### 3. Duck Typing Philosophy

> *"If it walks like a duck, swims like a duck, and quacks like a duck — then it probably is a duck."*

**Duck typing** means the type of an object is less important than whether it has the expected methods. Python doesn't check *what* the object is — only *what it can do*.

````python
# Ex 13.54 - Duck Typing — one Remote, controls both TV and Speaker
class Television:
    volume = 20
    def increase_volume(self):
        self.volume += 1
        print("Smart TV Volume :", self.volume)

class Speaker:
    volume = 15
    def increase_volume(self):
        self.volume += 1
        print("Smart Speaker Volume :", self.volume)

class Remote:
    def volume_up(self, device):    # accepts ANY object with increase_volume()
        device.increase_volume()    # duck typing — doesn't check type

remote = Remote()
remote.volume_up(Television())  # works on TV
remote.volume_up(Speaker())     # works on Speaker too!
````

## 🎭 Lesson 7 — Abstraction and Interfaces

### What is Abstraction?

**Abstraction** simplifies complex systems by showing only the essential details and hiding implementation. Like a TV remote — you press a button without knowing the internal circuitry.

```
  Method Types:
  ┌───────────────────┬────────────────────────────────────────────────┐
  │  Implemented      │  Has both name AND body  →  concrete method    │
  │  Unimplemented    │  Has only name, NO body  →  abstract method    │
  └───────────────────┴────────────────────────────────────────────────┘

  Abstract Class Rules:
  ┌─────────────────────────────────────────────────────────────────────┐
  │  • Import ABC and abstractmethod from abc module                    │
  │  • Class must inherit from ABC                                      │
  │  • Decorate abstract methods with @abstractmethod                   │
  │  • Abstract methods must use pass in the body                       │
  │  • Child class MUST implement all abstract methods                  │
  │  • Cannot create objects directly from an abstract class            │
  └─────────────────────────────────────────────────────────────────────┘

  Abstract Class vs Interface:
  ┌──────────────────────┬───────────────────────────────────────────────┐
  │  Abstract Class      │  Some implemented + some abstract methods     │
  │                      │  Use when: partial implementation is known    │
  ├──────────────────────┼───────────────────────────────────────────────┤
  │  Interface           │  ONLY abstract methods (no implementation)   │
  │                      │  Use when: NO implementation details known   │
  └──────────────────────┴───────────────────────────────────────────────┘
```

````python
# Ex 13.55 - Abstract Class with Abstract Method
from abc import ABC, abstractmethod

class Television(ABC):       # abstract class
    def __init__(self):
        self.volume = 15

    def power_on(self):      # concrete method — has body
        print("The TV is powered on.")

    @abstractmethod
    def increase_volume(self):  # abstract method — no body
        pass

# Cannot create Television object directly:
try:
    tv = Television()
except TypeError as e:
    print("Cannot instantiate abstract class:", e)
````

````python
# Child class MUST implement all abstract methods
class SmartTV(Television):
    def increase_volume(self):     # implements the abstract method
        self.volume += 1
        print("Smart TV Volume :", self.volume)

    def decrease_volume(self):
        self.volume -= 1
        print("Smart TV Volume :", self.volume)

my_smart_tv = SmartTV()        # can create child class objects
my_smart_tv.power_on()         # inherited concrete method
my_smart_tv.increase_volume()  # implemented abstract method
my_smart_tv.decrease_volume()
````

````python
# Ex 13.57 - Interface (abstract class with ONLY abstract methods)
from abc import ABC, abstractmethod

class Television(ABC):       # interface — all methods are abstract
    def __init__(self):
        self.volume   = 15
        self.channels = {1: "Aljezzira", 2: "BBC", 3: "CNN"}

    def power_on(self):      # unimplemented (interface-style)
        pass

    @abstractmethod
    def increase_volume(self):
        pass

class SmartTV(Television):   # must implement all abstract methods
    def power_on(self):
        print("The TV is powered on.")

    def increase_volume(self):
        self.volume += 1
        print("Smart TV Volume :", self.volume)

    def decrease_volume(self):
        self.volume -= 1
        print("Smart TV Volume :", self.volume)

my_smart_tv = SmartTV()
my_smart_tv.power_on()
my_smart_tv.increase_volume()
my_smart_tv.decrease_volume()
````
