## 📊 Lesson 1 — Introduction to Data Types

All programming exists to **create data and perform operations on it**. Before a computer can process data, it needs to know what *kind* of data it is — a number, text, a collection, a yes/no answer?

> **Data Type** is the type of data to be stored in a variable.

Python groups data types into two main categories:

| Category | Created by | Examples |
|---|---|---|
| **Built-in** | Python itself — already available | `int`, `str`, `list`, `bool`, `None`… |
| **User-defined** | You, the programmer | `class`, modules, custom arrays |

### Python's Built-in Data Type Classification Tree

```
┌─────────────────────────────────────────────────────────────────┐
│                   Python Built-in Data Types                    │
└─────────────────────────────────────────────────────────────────┘
                              │
          ┌───────────────────┼──────────────────────┐
          │                   │                      │
        None               Bool               Numeric & Sequences
     (absence of         True / False              │
       a value)                         ┌──────────┴──────────┐
                                        │                     │
                                    Numeric            Sequences /
                                        │              Collections
                              ┌─────────┼──────┐             │
                             int      float  complex   ┌──────┼──────┐
                           (whole)  (decimal) (a+bj)  str  list  tuple
                                                          set   dict
```

```
┌─────────────┬──────────────────────────────────┬────────────────────┐
│  Data Type  │  Description                     │  Example           │
├─────────────┼──────────────────────────────────┼────────────────────┤
│  None       │  No value / absence of value     │  None              │
│  bool       │  True or False                   │  True, False       │
│  int        │  Whole numbers                   │  5, -12, 9999      │
│  float      │  Decimal numbers                 │  3.14, -0.5, 9.81  │
│  complex    │  Real + imaginary parts          │  3+6j, 2-5.5j      │
│  str        │  Text / characters               │  "Hello", 'World'  │
│  list       │  Ordered, mutable collection     │  [1, 'a', 3.0]     │
│  tuple      │  Ordered, immutable collection   │  (1, 'a', 3.0)     │
│  set        │  Unordered, no duplicates        │  {1, 2, 3}         │
│  dict       │  Key-value pairs                 │  {'name': 'John'}  │
└─────────────┴──────────────────────────────────┴────────────────────┘
```

### The `type()` Function

Use `type()` to check the data type of any variable:

```
  variable_name ─────► type(variable_name) ─────► <class 'data_type'>

  Example:
  x = 42
  type(x)  ───►  <class 'int'>
```

## 🔢 Lesson 2 — Numeric Data Types

Python supports three numeric types: **int**, **float**, and **complex**. It also has **bool** which is technically a subclass of int.

```
┌──────────────┬──────────────────────────────────────────────────┐
│  Data Type   │  Description                                     │
├──────────────┼──────────────────────────────────────────────────┤
│  bool        │  True (=1) or False (=0) — Yes/No answers       │
│  int         │  ...-2, -1, 0, 1, 2... (no size limit!)         │
│  float       │  3.14, -0.5, 9.81 (decimals)                   │
│  complex     │  3+6j, 2-5.5j (real + imaginary)               │
└──────────────┴──────────────────────────────────────────────────┘
```

### 1. `bool` — Boolean (True or False)

The **`bool`** data type has exactly **two possible values**: `True` and `False`.

Python internally stores `True` as `1` and `False` as `0` — so you can do arithmetic with booleans!

```
  True  ──────────►  1  (integer value)
  False ──────────►  0  (integer value)

  So:  True + True  = 2
        True + False = 1
        False + False = 0
```

````python
# Ex 4.2 — Bool data type
a = True
b = False
print("Gravity = 9.81", a)
print("1 + 2 = 5 is", b)
print(type(a))
print(type(b))
````

````python
# True is 1 and False is 0 under the hood
print(True + True)     # 1 + 1 = 2
print(True + False)    # 1 + 0 = 1
print(False + False)   # 0 + 0 = 0
print(int(True))       # 1
print(int(False))      # 0
````

### 2. `int` — Integer (Whole Numbers)

Stores numbers **without** a decimal point. Python's `int` has **no size limit** — it can store numbers as large as your computer's RAM allows!

```
  int — Whole Numbers

  Negative ◄──────────────────► Positive
  ...-3, -2, -1, 0, 1, 2, 3...

  No size limit in Python!
  (Unlike C/Java which cap at 2,147,483,647)
```

````python
# Ex 4.3 — int data type
Sid = 16784
print("My student id is", Sid)
print(type(Sid))
````

````python
# Python int can be extremely large — no overflow!
big = 99999999999999999999999999999999999999999
print(big)
print(type(big))
print(2 ** 100)  # 2 to the power 100
````

### 3. `float` — Floating Point (Decimal Numbers)

Stores numbers **with** a decimal point. Used for measurements, grades, money, scientific values.

```
  Structure:  [integer part] . [fractional part]

  Example:      9   .   7   4
                │           │
          integer part   decimal part

  Always produces float:  10 / 2  →  5.0  (not 5)
  Floor division:         10 // 2 →  5    (int)
```

````python
# Ex 4.4 — float data type
grade = 9.74
print("My grade is", grade)
print(type(grade))
````

````python
# Division ALWAYS returns float, even for whole-number results
print(10 / 2)    # → 5.0  (float!)
print(10 // 2)   # → 5    (int — floor division)
print(type(10 / 2))
print(type(10 // 2))
````

### 4. `complex` — Complex Numbers

Stores numbers in the form **`a + bj`** where:
- `a` = real part
- `b` = imaginary part
- `j` = √−1 (the imaginary unit — Python uses `j` instead of `i`)

```
  complex — a + bj

        3  +  6j
        │     │
   real part  imaginary part
   (.real)    (.imag)

  Access parts:  n = 3+6j
                 n.real  →  3.0
                 n.imag  →  6.0
```

````python
# Ex 4.5 — complex data type
a = 3 + 6j
b = 2 - 5.5j
c = 2.4 + 7j
print(a, b, c)
print(type(a))
````

````python
# Access real and imaginary parts separately
n = 3 + 6j
print("Full complex number:", n)
print("Real part   (.real):", n.real)
print("Imaginary   (.imag):", n.imag)
print("Type:", type(n))
````

## 📝 Lesson 3 — Sequence and Collection Data Types

**Sequences** store **multiple values** inside a single variable. Python has five main collection types.

```
┌───────────┬───────────┬───────────┬─────────────┬───────────────────┐
│  Type     │ Brackets  │ Ordered?  │ Duplicates? │ Mutable?          │
├───────────┼───────────┼───────────┼─────────────┼───────────────────┤
│  str      │ " " or '' │    Yes    │     Yes     │ No  (immutable)   │
│  list     │  [ ]      │    Yes    │     Yes     │ Yes (changeable)  │
│  tuple    │  ( )      │    Yes    │     Yes     │ No  (immutable)   │
│  set      │  { }      │    No     │     No      │ Yes (changeable)  │
│  dict     │  {k:v}    │  Yes 3.7+ │  keys: No   │ Yes (changeable)  │
└───────────┴───────────┴───────────┴─────────────┴───────────────────┘
```

### 1. `str` — String (Text)

A **string** is a group of characters enclosed in quotes.

```
┌──────────────────┬──────────────────────────────────────────────────────┐
│  Single  ' '     │  name = 'John'                                       │
│                  │  Use when: no apostrophe in text                     │
├──────────────────┼──────────────────────────────────────────────────────┤
│  Double  " "     │  book = "John's book"                                │
│                  │  Use when: text contains apostrophe (')              │
├──────────────────┼──────────────────────────────────────────────────────┤

│                  │  Use when: text has BOTH ' and "                     │

```

````python
# Ex 4.6 — String with single quotes
name = 'John'
print("My name is", name)
print(type(name))
````

````python
# Ex 4.7 — String with double quotes (contains apostrophe)
book = "John's book"
print("This is", book)
print(type(book))
````

````python
# Ex 4.8 — String with triple quotes (contains both apostrophe and double quote)
v = '''"John's book is awesome."'''
print(v)
print(type(v))
````

### 2. `list` — Ordered, Mutable Collection

A **list** stores multiple values in order inside square brackets `[ ]`. It can hold **different data types** and is **mutable** (changeable after creation).

```
  detail = [1, 'John', 9.5, 'pass', 1]
              │    │     │     │     │
  Index:     [0]  [1]   [2]   [3]   [4]
  Type:      int  str  float  str   int

  ✓ Different data types allowed
  ✓ Duplicates allowed (1 appears twice)
  ✓ Mutable — can add/remove/change items
  ✓ Ordered — items keep their position
```

**Syntax:** `name_of_list = [element, element, ...]`

````python
# Ex 4.10 — List data type
detail = [1, 'John', 9.5, 'pass', 1]
print("My details", detail)
print(type(detail))
````

### 3. `tuple` — Ordered, Immutable Collection

A **tuple** is like a list but **immutable** — once created, elements cannot be added, removed, or changed. Uses parentheses `( )`.

```
  detail = (1, 'John', 9.5, 'pass', 1)

  ✓ Different data types allowed
  ✓ Duplicates allowed (1 appears twice)
  ✗ Immutable — CANNOT change after creation
  ✓ Ordered — items keep their position

  Use tuple for: GPS coordinates, days of week,
  RGB colours, database records that must not change
```

**Syntax:** `name_of_tuple = (element, element, ...)`

````python
# Ex 4.11 — Tuple data type (Creating a Tuple)
detail = (1, 'John', 9.5, 'pass', 1)
print("My details", detail)
print(type(detail))
````

### 4. `set` — Unordered, No Duplicates

A **set** stores multiple values in curly braces `{ }`. It **automatically removes duplicate values** and has no guaranteed order.

```
  Before:  {1, 'John', 9.5, 'pass', 1}  ← duplicate 1
                              │
                       set removes 1
                              ▼
  After:   {'pass', 1, 9.5, 'John'}     ← no duplicate

  ✗ Duplicates NOT allowed — auto removed
  ✗ No guaranteed order
  ✓ Mutable — can add/remove items
```

**Syntax:** `name_of_set = {element, element, ...}`

````python
# Ex 4.12 — Set data type (notice the duplicate 1 is removed)
detail = {1, 'John', 9.5, 'pass', 1}
print("My details", detail)   # 1 appears only once!
print(type(detail))
````

### 5. `dict` — Dictionary (Key-Value Pairs)

A **dictionary** stores data as **key:value** pairs inside curly braces. Think of it like a real dictionary — look up a word (key) to find its definition (value).

```
  Info = {2: 'John',   1: 'Mike',   3: 'Dani'}
           │    │       │    │       │    │
          key value   key value   key value

  Like a table:
  ┌──────┬─────────┐
  │ Rank │  Name   │
  ├──────┼─────────┤
  │  1   │  Mike   │
  │  2   │  John   │
  │  3   │  Dani   │
  └──────┴─────────┘

  Access: Info[1]  →  'Mike'
  Keys must be unique; values can repeat
```

**Syntax:** `name_of_dictionary = {key: value, key: value, ...}`

````python
# Ex 4.13 — Dictionary data type (race results: rank → name)
Info = {2: 'John', 1: 'Mike', 3: 'Dani'}
print("Race", Info)
print(type(Info))
````

````python
# Access a value by its key
student = {'name': 'Sarah', 'age': 20, 'grade': 'A'}
print(student['name'])
print(student['grade'])
````

## 🔄 Lesson 4 — Type Conversion (Type Casting)

Based on requirements, developers can convert from one data type to another. This is called **type conversion** or **type casting**.

```
  Source Type ──────► conversion_function() ──────► New Type

  Example:  int  ──────►  float(65)  ──────►  65.0
            str  ──────►  int('123') ──────►  123
            list ──────►  tuple([1,2])  ────►  (1, 2)
```

### Python's Built-in Type Conversion Functions

```
┌──────────────────┬────────────────────────────────────────────────┐
│  Function        │  Converts to                                   │
├──────────────────┼────────────────────────────────────────────────┤
│  int(x)          │  Integer — removes decimal, truncates NOT rounds│
│  float(x)        │  Float — adds decimal point                    │
│  complex(x)      │  Complex — adds 0j imaginary part              │
│  bool(x)         │  Boolean — 0 → False, non-zero → True          │
│  str(x)          │  String — wraps value in quotes                │
│  list(x)         │  List — converts iterable to [ ]               │
│  tuple(x)        │  Tuple — converts iterable to ( )              │
│  set(x)          │  Set — converts iterable, removes duplicates   │
│  ord(character)  │  Int — returns ASCII value of a character      │
│  chr(number)     │  Str — returns character from ASCII number     │
└──────────────────┴────────────────────────────────────────────────┘
```

### 1. Convert to `float`

````python
# Ex 4.14 — Convert int to float
i = 65
itf = float(i)   # integer to float
print(itf)
print(type(itf))
````

### 2. Convert to `int`

````python
# Ex 4.15 — Convert float to int (TRUNCATES — does NOT round!)
f = 65.89
fti = int(f)   # float to int
print(fti)     # → 65, NOT 66!
print(type(fti))
````

````python
# Ex 4.16 — Convert string of digits to int
st = '123'
sti = int(st, 10)   # base 10 (decimal system)
print(sti)
print(type(sti))
````

  'A' = 65    'a' = 97    '0' = 48
  'B' = 66    'b' = 98    '1' = 49
  'Z' = 90    'z' = 122   '9' = 57

  ord('A') → 65   (character → number)
  chr(65)  → 'A'  (number → character)
  </pre>
</div>

````python
# Ex 4.17 — Convert character to its ASCII integer value using ord()
ch = 'A'   # Character to Int (ASCII)
cti = ord(ch)
print(cti)         # → 65
print(type(cti))
````

````python
# ASCII table demo — print letters and their codes
print("Char  ASCII")
print("─────────────")
for letter in ['A', 'B', 'C', 'D', 'E', 'Z']:
    print(f"  {letter}  →  {ord(letter)}")
````

### 3. Convert to `complex`

````python
# Ex 4.18 — Convert real number to complex (adds +0j imaginary part)
r = 6   # Real No. to Complex No.
rtc = complex(r)
print(rtc)         # → (6+0j)
print(type(rtc))
````

### 4. Convert to `str`

````python
# Ex 4.19 — Convert integer to string
i = 65
itst = str(i)   # Int to Str
print(itst)
print(type(itst))
````

````python
# Ex 4.20 — Convert ASCII number to character using chr()
nta = chr(65)   # ASCII Value to str
print(nta)      # → A
print(type(nta))
````

### 5. Convert to `list`

````python
# Ex 4.21 — Convert tuple to list
t = (1, 'John', 9.5, 'pass')
ttl = list(t)   # Tuple to List
print(ttl)
print(type(ttl))
````

````python
# Ex 4.22 — Convert set to list
s = {1, 'John', 9.5, 'pass'}
stl = list(s)   # Set to List
print(stl)
print(type(stl))
````

### 6. Convert to `tuple`

````python
# Ex 4.23 — Convert list to tuple
l = [1, 'John', 9.5, 'pass']
ltt = tuple(l)   # List to Tuple
print(ltt)
print(type(ltt))
````

````python
# Ex 4.24 — Convert set to tuple
s = {1, 'John', 9.5, 'pass'}
stt = tuple(s)   # Set to Tuple
print(stt)
print(type(stt))
````

### 7. Convert to `set`

````python
# Ex 4.25 — Convert list and tuple to set (duplicates removed automatically)
l = [1, 'John', 9.5, 'pass']
lts = set(l)   # List to Set
print(lts)
print(type(lts))

t = (1, 'John', 9.5, 'pass')
tts = set(t)   # Tuple to Set
print(tts)
print(type(tts))
````
